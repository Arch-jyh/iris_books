```python
#定义列向量线性组合函数
def col_LC(A,coefficients):
    m,p = A.shape
    #生成的一维数组,长度为m,值为0
    col_LC_result = np.zeros(m)

    for k in range(p):
        col_LC_result += coefficients[k] * A[:,k]
    return col_LC_result
```

```python
#定义矩阵乘法函数,列向量组合
def matrix_multiplication_col_LC(A,B):
    m,p_a = A.shape
    p_b,n = B.shape

    if p_a != p_b:
        raise ValueError('Dimensions do nor match')
    #省略了形参shape的参数名,直接传递了实参,按位置读取实参对应到对应位置的形参
    C = np.zeros((m,n))

    for j in range(n):
        coeffs = B[:,j]
        C[:,j] = col_LC(A,coeffs)

    return C
```

```python
A = np.array([[1,2,3],
              [4,5,6]])
A
```

```python
B = A.T
B
```
