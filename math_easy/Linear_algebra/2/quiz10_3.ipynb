```python
import numpy as np
```

```python
np.random.seed(0)
n_samples = 100
```

```python
#自变量数据
x = np.random.uniform(0,4,n_samples) #范围内均匀分布随机数,范围左闭右开
#因变量数据
y = np.sin(0.4*np.pi*x) + 0.4 * np.random.randn(n_samples)

#此时x,y都是1D数组 这里为了保持原题,如果x生成完后立刻执行y如果后面的randn的生成没有reshape将会发生广播,生成100*100的矩阵
```

```python
x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
degree = 3
X_poly = np.ones_like(x)
for i in np.arange(degree)+1:
    #这里注意,column自动将1D看成2D列向量了
    X_poly = np.column_stack((X_poly,x**i))
X_poly
```

```python
b = np.linalg.inv(X_poly.T @ X_poly) @ X_poly.T @ y #
b
```

```python
y_hat = X_poly@b
y_hat
```

```python
epsilon = y - y_hat
epsilon
```

```python
#GPT验证
import numpy as np
# 生成数据
np.random.seed(0)
n_samples = 100
x = np.random.uniform(0, 4, n_samples)
y = np.sin(0.4*np.pi * x) + 0.4*np.random.randn(n_samples)

# 设计矩阵 [1, x, x^2, x^3]
X = np.vander(x, 4, increasing=True)

# 最小二乘求系数 b
b = np.linalg.lstsq(X, y, rcond=None)[0]

# 预测和残差
y_hat = X @ b
eps = y - y_hat
b
```

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures #
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
n_samples = 100
#自变量数据
x = np.random.uniform(0,4,n_samples) #范围内均匀分布随机数,范围左闭右开
#因变量数据
y = np.sin(0.4*np.pi*x) + 0.4 * np.random.randn(n_samples)
```

```python
x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
degree = 3
pf = PolynomialFeatures(degree = degree)
X_poly = pf.fit_transform(x) #
X_poly
```

```python
lr = LinearRegression()
lr.fit(X_poly,y)
```

```python
coef = lr.coef_[0]
intercept = lr.intercept_[0]
#注意array是可变类型
coef
```

```python
b = np.hstack((intercept,coef[1:]))
b
```

```python
y_hat = lr.predict(X_poly) #
y_hat
```

```python
espilon = y - y_hat
espilon
```
