```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
undirected_G = nx.Graph() #蒋健无向图对象
```

```python
undirected_G.add_nodes_from(['a','b','c','d'])
```

```python
undirected_G.add_edges_from([('a','b'),
                             ('a','c'),
                             ('a','d'),
                             ('b','c'),
                             ('c','d')])#往对象里面存储数据,和前面添加节点一样,到这都不需要pos,仅仅保存节点间的关系数据
```

```python
pos = {'a':[0,0],
       'b':[1,0],
       'c':[1,1],
       'd':[0,1]}#创建位置,仅在绘图有用
```

```python
plt.figure(figsize = (6,6))
#自动获取当前的ax将数据填充进去,没有就创建一个
nx.draw_networkx(undirected_G,pos = pos,node_size = 188)
```

```python
#第一个函数是生成邻接矩阵 第二个函数是将矩阵转换为numpy数组类型
A = nx.adjacency_matrix(undirected_G).toarray()
A
```

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
undirected_G = nx.Graph()
```

```python
undirected_G.add_nodes_from(['a','b','c','d'])
```

```python
undirected_G.add_edges_from([('a','b',{'weight':3}),
                             ('a','c',{'weight':8}),
                             ('a','d',{'weight':5}),
                             ('b','c',{'weight':6}),
                             ('c','d',{'weight':2})])
```

```python
pos = {'a':[0,0],
       'b':[1,0],
       'c':[1,1],
       'd':[0,1]}
```

```python
plt.figure(figsize = (6,6))
nx.draw_networkx(undirected_G,pos = pos,node_size = 188)

edge_labels = nx.get_edge_attributes(undirected_G,'weight')
#绘制权重
nx.draw_networkx_edge_labels(undirected_G,pos,edge_labels = edge_labels)
```
