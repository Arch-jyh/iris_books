```python
import numpy as np
```

```python
np.random.seed(0)
n_samples = 100

X_vecs = np.random.randn(n_samples,6)

y = X_vecs @ np.array([[-3],[2],[-2],[3],[1],[-1]])
#1D数组与列向量相加,广播中会默认看成行向量导致列向量列方向广播,1D数组行方向广播
y = y + 1 + 0.25*np.random.randn(n_samples,1)
```

```python
y.shape
```

```python
X = np.column_stack((np.ones_like(y),X_vecs))
```

```python
X.shape
```

```python
b = np.linalg.inv(X.T @ X) @ X.T @ y
b
```

```python
y_hat = X @ b
y_hat.shape
```

```python
epsilon = y - y_hat
epsilon.shape
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
n_samples = 100

X_vecs = np.random.randn(n_samples,6)

y = X_vecs @ np.array([[-3],[2],[-2],[3],[1],[-1]])
#1D数组与列向量相加,广播中会默认看成行向量导致列向量列方向广播,1D数组行方向广播
y = y + 1 + 0.25*np.random.randn(n_samples,1)
```

```python
LR = LinearRegression()
```

```python
LR.fit(X_vecs,y)
```

```python
slope = LR.coef_
slope
```

```python
intercept = LR.intercept_
intercept
```

```python
#intercept自动视为一列,slope视为6列
b_LR = np.column_stack((intercept,slope))
b_LR = b_LR.T
b_LR
```

```python
y_hat = LR.predict(X_vecs)
y_hat.shape
```

```python
epsilen = y - y_hat
epsilen.shape
```
