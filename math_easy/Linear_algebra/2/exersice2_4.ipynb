```python
import numpy as np
```

```python
def matrix_multiplication(A,B):
    #获取形状
    m,p_A = A.shape #shape是一个自带属性,返回行数和列数
    p_B,n = B.shape
    #检测形状是否符合法则
    if p_A != p_B:
        #抛出错误
        raise ValueError('Dimensions do not match')

    #初始化结果矩阵C形状为(m,n)  初始值为0的全零矩阵
    C = np.zeros((m,n))

    #三层循环计算C
    #range不是生成元组(0,m) 生成range(0,m)表示这是range迭代器,是一个可迭代对象,迭代[0,10)的整数,左闭右开
    #np.arange(10)是返回所有数组中的数,是一个array数组
    for i in range(m):
        for j in range(n):
            for k in range(p_A):  #p_A = p_B  这里遍历了A的行与B的列
                C[i,j] += A[i,k] * B[k,j]
    return C                
```

```python
A = np.array([[1,2,3],
              [4,5,6]])
A
```

```python
B = A.T
B
```

```python
matrix_multiplication(A,B)
```

```python
matrix_multiplication(B,A)
```
