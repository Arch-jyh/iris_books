```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
directed_G = nx.DiGraph()
```

```python
directed_G.add_nodes_from(['s','c','r'])
```

```python
directed_G.add_edges_from([('s','s',{'weight':0.7}),
                           ('s','c',{'weight':0.25}),
                           ('s','r',{'weight':0.05}),
                           ('c','c',{'weight':0.3}),
                           ('c','s',{'weight':0.45}),
                           ('c','r',{'weight':0.25}),
                           ('r','r',{'weight':0.15}),
                           ('r','s',{'weight':0.55}),
                           ('r','c',{'weight':0.3})])#最好用weight,因为内置算法自动识别情况下weight这个键的名称
```

```python
pos = {'s':[1,np.sqrt(3)],
       'c':[0,0],
       'r':[2,0]}
```

```python
plt.figure(figsize = (6,6))
nx.draw_networkx(directed_G,pos = pos,node_size = 500,
                connectionstyle = 'arc3, rad = 0.1')
edge_labels = nx.get_edge_attributes(directed_G,'weight')#这里不填字符串表示取那种边的属性就不会检索,不是自动的
nx.draw_networkx_edge_labels(directed_G,pos,edge_labels = edge_labels,
                            connectionstyle='arc3,rad=0.1')#绘图伙计相同,绘制标签层覆盖在边上方
```

```python
A = nx.adjacency_matrix(directed_G).toarray()
A
```
