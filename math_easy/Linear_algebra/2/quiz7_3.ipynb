```python
import numpy as np
from scipy.linalg import lu
```

```python
A = np.array([[2,5,8,7],
              [5,2,2,8],
              [7,5,6,6],
              [5,4,4,8]])
```

```python
p,l,u = lu(A)
```

```python
np.allclose(A,p@l@u)
```

```python
p
```

```python
p,_,_ = lu(A,p_indices = True)
```

```python
p
```

```python
np.allclose(A,l[p,:]@u)
```

```python
rng = np.random.default_rng()
```

```python
A = rng.uniform(low = -4,high = 4,size = [3,2,4,8])
p,l,u = lu(A)
```

```python
p.shape,l.shape,u.shape
```

```python
np.allclose(A,p@l@u)
```

```python
PL,U = lu(A,permute_l = True)
np.allclose(A,PL@U)
```
