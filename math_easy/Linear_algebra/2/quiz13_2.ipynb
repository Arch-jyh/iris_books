```python
#Q1 cholesky分解
```

```python
import numpy as np
```

```python
A = np.array([[5/4,-3/4],
              [-3/4,5/4]])
```

```python
R_T = np.linalg.cholesky(A) #默认返回下三角矩阵
```

```python
R_T
```

```python
R = R_T.T
R
```
