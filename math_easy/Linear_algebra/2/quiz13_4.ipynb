```python
#Q1 马氏距离
```

```python
from scipy.spatial import  distance
```

```python
#协方差矩阵
iv = [[1,0.5,0.5],
      [0.5,1,0.5],
      [0.5,0.5,1]]
```

```python
distance.mahalanobis([1,0,0],[0,1,0],iv)
```

```python
distance.mahalanobis([0,2,0],[0,1,0],iv)
```

```python
distance.mahalanobis([2,0,0],[0,1,0],iv)
```

```python
#Q2 标准化欧氏距离
```

```python
distance.seuclidean([1,0,0],[0,1,0],[0.1,0.1,0.1])#第三个参数为各个分量的方差,不传入矩阵,只传入对角线元素了
```

```python
distance.seuclidean([1,0,0],[0,1,0],[1,0.1,0.1])
```

```python
distance.seuclidean([1,0,0],[0,1,0],[10,0.1,0.1])
```

```python
#Q6 一维数据之间的余弦距离
#因为距离一定是正的,cos有负的情况需要用1-cos 如果cos = 1距离为0(同向) cos = -1 距离为2(反向) cos = 0距离为1(正交)
#其实是一种关于余弦的距离,而不是简单的余弦就是距离
```

```python
distance.cosine([1,0,0],[0,1,0])
```

```python
distance.cosine([100,0,0],[0,1,0])
```

```python
distance.cosine([1,1,0],[0,1,0])
```
