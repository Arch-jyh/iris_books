```python
#矩阵乘法内积视角
```

```python
import numpy as np
```

```python
def matrix_multiplication_inner(A,B):
    m,p_A = A.shape
    p_B,n = B.shape

    if p_A != p_B:
        raise ValueError('Dimensions do not match')

    C = np.zeros((m,n))

    for i in range(m):
        for j in range(n):
            row_i_A = A[i,:]
            col_j_B = B[:,j]
            C[i,j] = np.dot(row_i_A,col_j_B)
    return C
```

```python
A = np.array([[1,2,3],
              [4,5,6]])
A
```

```python
B = A.T
B
```

```python
matrix_multiplication_inner(A,B)
```

```python
matrix_multiplication_inner(B,A)
```
