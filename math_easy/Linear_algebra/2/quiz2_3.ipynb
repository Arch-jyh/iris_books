```python
q2_arr3 = np.random.randint(1,10,(10,5))
q2_arr3
```

```python
def diagonal(array):
    q2_diag_1d = np.array([])
    if array.shape[0] >= array.shape[1]:
        for i in np.arange(0,array.shape[1]):
            q2_diag_1d = np.append(q2_diag_1d,array[i,i])
        print(q2_diag_1d)

    elif array.shape[0] < array.shape[1]:
        for i in np.arange(0,array.shape[0]):
            q2_diag_1d = np.append(q2_diag_1d,array[i,i])  
        print(q2_diag_1d)

    return q2_diag_1d
```

```python
diagonal(q2_arr1)
```

```python
diagonal(q2_arr2)
```

```python
diagonal(q2_arr3)
```

```python
q3_arr1 = diagonal(q2_arr1)
q3_arr2 = diagonal(q2_arr2)
q3_arr3 = diagonal(q2_arr3)
```

```python
def build_diagonal(array_1d):
    #shape返回一维数组,需要按照位置提取
    shape = array_1d.shape[0]
    zero_arr = np.zeros((shape,shape))
    for i in np.arange(0,shape):
        zero_arr[i,i] = array_1d[i]
        q3_diag_2d = zero_arr
    print(q3_diag_2d)
    return(q3_diag_2d)
```

```python
build_diagonal(q3_arr1)
```

```python
build_diagonal(q3_arr2)
```

```python
build_diagonal(q3_arr3)
```
