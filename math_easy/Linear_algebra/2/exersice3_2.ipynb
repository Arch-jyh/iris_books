```python
def matrix_multiplication_outer(A,B):
    #获取形状
    m,p_A = A.shape
    p_B,n = B.shape

    #检查形状是否匹配
    if p_A != p_B:
        raise ValueError('Dimensions do not match')

    #初始化结果矩阵
    C = np.zeros((m,n))

    for k in range(p_A):
        a_k = A[:,[k]]
        b_k = B[[k],:]
        C += (a_k @ b_k)
        
    return C
```

```python
A = np.array([[1,2,3],
              [4,5,6]])
A
```

```python
B = A.T
B
```

```python
matrix_multiplication_outer(A,B)
```

```python
matrix_multiplication_outer(B,A)
```
