```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[3,1],
              [1,3]])
```

```python
lambdas ,V = np.linalg.eig(A)
#数学上,不用保证lambda的顺序,V中的单位向量在子空间的方向±都可以
#python中根据函数的内置函数计算确是确定的结果
```

```python
lambdas
```

```python
Lambda = np.diag(lambdas)
Lambda
```

```python
lambda_1 = lambdas[0]
lambda_1
```

```python
v_1 = V[:,[0]]
v_1
```

```python
A @ v_1
```

```python
lambda_1 * v_1
```

```python
lambda_2 = lambdas[1]
lambda_2
```

```python
v_2 = V[:,[1]]
v_2
```

```python
A @ v_2
```

```python
lambda_2 * v_2
```

```python
V @ Lambda @ np.linalg.inv(V)
```

```python
fig,ax = plt.subplots()

theta_array = np.linspace(0,2*np.pi,721)
x1_array = np.cos(theta_array)
x2_array = np.sin(theta_array)
unit_circle = np.column_stack((x1_array,x2_array))

A_unit_circle = unit_circle @ A.T #行才是一组变化后的点
A_v1 = A@v_1 #这里还是列向量
A_v2 = A@v_2

plt.quiver(0,0,A_v1[0],A_v1[1],
           angles = 'xy',scale_units = 'xy',
           #zorder表示深度层级,让其显示在最上面
           scale = 1,color = '#FF00CF',zorder = 1e5)
plt.quiver(0,0,v_1[0],v_1[1],
           angles = 'xy',scale_units = 'xy',
           scale = 1,color = '#FF00CF',zorder = 1e5)

plt.quiver(0,0,A_v2[0],A_v2[1],
           angles = 'xy',scale_units = 'xy',
           scale = 1,color = '#FFBD00',zorder = 1e5)
plt.quiver(0,0,v_2[0],v_2[1],
           angles = 'xy',scale_units = 'xy',
           scale = 1,color = '#FFBD00',zorder = 1e5)

plt.plot(A_unit_circle[:,0],A_unit_circle[:,1],c = 'k',ls = '--')
plt.plot(unit_circle[:,0],unit_circle[:,1],c = 'k',ls = '--')

plt.axvline(x = 0,color = 'k',zorder = 0)#这里深度层级在最下面
plt.axhline(y = 0,color = 'k',zorder = 0)
#刻度参数设置 axis表示作用的轴这里同时作用xy.which表示主次刻度,这里表示都设置
    #最后两个参数表示隐藏底部和左侧可读
ax.tick_params(axis = 'both',which = 'both',length = 0,labelbottom = False,labelleft = False)
ax.set_aspect(1)
lim = 2;ax.set_xlim([-lim,lim]);ax.set_ylim([-lim,lim])
plt.xticks(np.arange(-lim,lim));plt.yticks(np.arange(-lim,lim))#上面隐藏了,这里无法显示
ax.grid(linestyle = '--',linewidth = 0.15,color = [0.8,0.8,0.8])
```
