```python
import numpy as np
```

```python
A = np.array([[5/4,-3/4],
              [-3/4,5/4]])
G_1 = A.T @ A
G_2 = A @ A.T

lambdas_G_1,V_G_1 = np.linalg.eigh(G_1)
lambdas_G_2,U_G_2 = np.linalg.eigh(G_2)
print(lambdas_G_1)
print(V_G_1)
print(lambdas_G_2)
print(U_G_2)

print('')

U,S,V_T = np.linalg.svd(A)
print(U)
print(S)
print(V_T)
```
