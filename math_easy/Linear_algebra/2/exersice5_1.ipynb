```python
import numpy as np
```

```python
def inverse_2x2_A(A):
    """
    A = [[a,b],
         [c,d]]
    """
    a,b = A[0]
    c,d = A[1]

    det = a*d - b*c

    if det == 0:
        print('矩阵不可逆,行列式为0')
        return None

    inv_A = np.array([[d,-b],[c,-a]]) / det
    return inv_A
```

```python
A = [[1,2],
     [3,4]]
A = np.array(A)
A
```

```python
inverse_2x2_A(A)
```

```python
B = [[0,1],
     [0,2]]
B = np.array(B)
B
```

```python
inverse_2x2_A(B)
```
