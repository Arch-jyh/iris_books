```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
theta = np.pi/3
A = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta), np.cos(theta)]])
```

```python
a1 = A[:,0]
a2 = A[:,1]
```

```python
#单位圆
theta_array = np.linspace(0,2*np.pi,721) #范围是两端都包含,是左闭右闭包括0和2pi
x1_array = np.cos(theta_array)
x2_array = np.sin(theta_array)
```

```python
fig,ax = plt.subplots()
ax.plot(x1_array,x2_array,lw = 0.2,c = 'k',ls = '--')

ax.quiver(0,0,a1[0],a1[1],
          #箭头方向按照xy坐标系表示,箭头长度也用xy坐标系表示
          angles = 'xy',scale_units = 'xy',
          #缩放因子为1,按照实际向量长度缩放.
          scale = 1,color = 'r')
ax.quiver(0,0,a2[0],a2[1],
          angles = 'xy',scale_units = 'xy',
          scale = 1,color = '#92D050')


grid_range = np.arange(-10,11)
for i in grid_range:
    start = i*a1 + grid_range[0]*a2
    end = i*a1 +grid_range[-1]*a2
    ax.plot([start[0],end[0]],[start[1],end[1]],color = 'k' ,lw = 0.2)

for j in grid_range:
    start = grid_range[0]*a1 + j*a2
    end = grid_range[-1]*a1 + j *a2
    ax.plot([start[0],end[0]],[start[1],end[1]],color = 'k',lw = 0.2)

ax.set_xlim(-4,4)
ax.set_ylim(-4,4)
ax.set_aspect('equal')
ax.axhline(0,color = 'k',linewidth = 1)
ax.axvline(0,color = 'k',linewidth = 1)
ax.grid(c = '0.88')
```
