```python
import numpy as np
```

```python
A = np.array([[1,1],
              [2,4]]) #系数矩阵
```

```python
b = np.array([[35],
              [94]]) #常数列向量
```

```python
np.linalg.det(A)#用行列式判断是否可逆
```

```python
A_inv = np.linalg.inv(A)
A_inv
```

```python
A_inv@A
```

```python
A@A_inv
```

```python
A_inv@b
```
