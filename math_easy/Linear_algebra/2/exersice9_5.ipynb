```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
np.random.seed(88888)
#rand表示随机,n为normal正态分布
    #randn是旧式写法 新写法直接使用nromal
A = np.random.randn(12,8)
```

```python
Q_complete,R_complete = np.linalg.qr(A,mode = 'complete')
```

```python
fig,axs = plt.subplots(1,5,figsize = (12,3))

plt.sca(axs[0])
#heatmap不需要给出单独的xy数据,接收的参数就是二维矩阵
ax = sns.heatmap(A,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidths = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('A')

plt.sca(axs[1])
plt.title('=')
plt.axis('off')

plt.sca(axs[2])
ax = sns.heatmap(Q_complete,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidths = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('Qc')

plt.sca(axs[3])
plt.title('@')
plt.axis('off')

plt.sca(axs[4])
ax = sns.heatmap(R_complete,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidths = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('Rc')
```

```python
fig,axs = plt.subplots(1,5,figsize = (12,3))

#配置当前活动坐标轴后面的所有plt都是作用在这个轴上
plt.sca(axs[0])
#axs包含了很多的轴,上面已经配置了轴后面的语句将会默认在此轴下执行
#因为配置了当前轴,所以sns.heatmap将会作用在当前轴上(axs[0])然后返回一个作用后的当前轴
#当前轴没有变只是用了一个新的变量名进行表示,所以后面可以直接用ax.进行配置axs[0]
#axis和numpy数组是可变对象 所以返回的轴和传入的轴相同.
    #像是int等python自带数值类型是不可变的
ax = sns.heatmap(Q_complete.T @ Q_complete,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidths = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
#如果没有上一步的ax = 的赋值操作,将可以使用axs[0].set_aspect进行比例调整
    #或plt.gca().set_aspect
ax.set_aspect('equal')
#标题是轴对象的属性,如果使用ax则是ax.set_title  如果写ax.title则会返回标题对象
#plt.title是一个针对状态机的语法,先获取当前轴然后再执行当前轴的方法(上同的.set_title)
    #自动执行plt.gca()返回当前轴然后执行.set_title()
#操作ax轴上使用plt是为了方便.这是针对状态机所设计的语法
    #其实创建fig和ax对象后,不需要使用plt即可 plt使用状态机所设计的语法是为了方便
    #plt.这种配置当前轴的操作是语法糖,将底层复杂语法包装成简单语法
plt.title('I')

plt.sca(axs[1])
plt.title('=')
plt.axis('off')

plt.sca(axs[2])
#重新指向变量名ax,这里是axs[2]
ax = sns.heatmap(Q_complete.T,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k' ,linewidth = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('Qc.T')

plt.sca(axs[3])
plt.title('@')
plt.axis('off')

plt.sca(axs[4])
ax = sns.heatmap(Q_complete,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidths = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('Qc')
```

```python
fig,axs = plt.subplots(1,5,figsize = (12,3))

plt.sca(axs[0])
ax = sns.heatmap(Q_complete@Q_complete.T,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidths = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('I')

plt.sca(axs[1])
plt.title('=')
plt.axis('off')

plt.sca(axs[2])
ax = sns.heatmap(Q_complete,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidth = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('Qc')

plt.sca(axs[3])
plt.title('@')
plt.axis('off')

plt.sca(axs[4])
ax = sns.heatmap(Q_complete.T,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidths = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('Qc.T')
```

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
np.random.seed(8888)
A = np.random.randn(12,8)
```

```python
Q,R = np.linalg.qr(A)
```

```python
fig,axs = plt.subplots(1,5,figsize = (12,3))

plt.sca(axs[0])
ax = sns.heatmap(A,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidths = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('A')

plt.sca(axs[1])
plt.title('=')
plt.axis('off')

plt.sca(axs[2])
ax = sns.heatmap(Q,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidth = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('Q')

plt.sca(axs[3])
plt.title('@')
plt.axis('off')

plt.sca(axs[4])
ax = sns.heatmap(R,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidth = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('R')
```

```python
fig,axs = plt.subplots(1,5,figsize = (12,3))

plt.sca(axs[0])
ax = sns.heatmap(Q.T @ Q,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidth = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('I')

plt.sca(axs[1])
plt.title('=')
plt.axis('off')

plt.sca(axs[2])
ax = sns.heatmap(Q.T,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidth = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('Q.T')

plt.sca(axs[3])
plt.title('@')
plt.axis('off')

plt.sca(axs[4])
#cmap中的'bwr'是蓝色到白色再到红色
#kws是keywords,表示关键字参数,传入'方向':水平'的字典键值对
ax = sns.heatmap(Q,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidth = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('Q')
```

```python
fig,axs = plt.subplots(1,5,figsize = (12,3))

plt.sca(axs[0])
ax = sns.heatmap(Q@Q.T,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidths = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('I')

plt.sca(axs[1])
plt.title('=')
plt.axis('off')

plt.sca(axs[2])
ax = sns.heatmap(Q,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidth = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('Q')

plt.sca(axs[3])
plt.title('@')
plt.axis('off')

plt.sca(axs[4])
ax = sns.heatmap(Q.T,cmap = 'bwr',vmax = 2.5,vmin = -2.5,
                 linecolor = 'k',linewidth = 0.1,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('Q.T')
```
