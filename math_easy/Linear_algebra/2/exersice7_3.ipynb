```python
import numpy as np
```

```python
def gauss_elimination(A,b):
    #转换为浮点数计算
    A = A.astype(float)
    b = b.astype(float)
    n = len(b) #计算数量

    #构造增广矩阵
    #stack要求拼接矩阵的维度一致,将b从1D行向量转换为2D列向量
    #1D2D是编程上的说法,为了方便描述矩阵的形状与数学概念对接.
    Ab = np.hstack([A,b.reshape(-1,1)])

    #原始矩阵A > 上三角
    for i in range(n):
        #行交换,选择最大值作为主元.
        #交换后放主对角线的列主元
        max_row = np.argmax(np.abs(Ab[i:,i])) + i
        Ab[[i,max_row]] = Ab[[max_row,i]]

        Ab[i] = Ab[i]/Ab[i,i]

        for j in range(i+1,n):
            Ab[j] = Ab[j] - Ab[j,i] * Ab[i]

    x = np.zeros(n)
    for i in range(n-1,-1,-1):
        x[i] = Ab[i,-1] - np.dot(Ab[i,i+1:n],x[i+1:n])

    return x
```

```python
A = np.array([[1,3,6],
              [2,7,14],
              [0,2,5]])
b = np.array([25,58,19])
```

```python
x = gauss_elimination(A,b)
x
```
