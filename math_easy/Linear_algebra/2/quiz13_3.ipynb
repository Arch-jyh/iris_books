```python
A = np.array([[1,0],
              [0,2]])
lambdas,V = np.linalg.eig(A)
print(lambdas)
print(V)
```

```python
A = np.array([[-1,0],
              [0,-2]])
lambdas,V = np.linalg.eig(A)
print(lambdas)
print(V)
```

```python
A = np.array([[1,-1],
              [-1,1]])
lambdas,V = np.linalg.eig(A)
print(lambdas)
print(V)
```
