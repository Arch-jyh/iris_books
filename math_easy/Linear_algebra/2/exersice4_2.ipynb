```python
import numpy as np
import matplotlib.pyplot as plt
#从一个专业3D绘图模块中导入一个指定绘图类
#mpl_toolkits指的是matplotlib的一个扩展工具包
#mplot3d是matplotlib的3d绘图模块 然后代入模块汇总的art3d绘图模块
#import后导入一个三维多边形集合的类-将3d面片放到3d坐标绘制-接受一组三维坐标,生成多边形集合对象
#ploy是多边形polygon 3d是three-dimensional collection表示集合或者容器
#绘制面片,不一定拼接成几何体
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

```python
A = np.array([[2,0,2],
              [2,2,0],
              [0,2,2]])
```

```python
a1 = A[:,0]
a2 = A[:,1]
a3 = A[:,2]
```

```python
O = np.array([0,0,0])
P1 = a1
P2 = a2
P3 = a3
P4 = a1+a2
P5 = a1+a3
P6 = a2+a3
P7 = a1+a2+a3

vertices = [O,P1,P2,P3,P4,P5,P6,P7]
```

```python
#确定每一面,每一个面是4个点
faces = [
    [O,P1,P4,P2], #底面
    [O,P2,P6,P3], #侧面
    [O,P1,P5,P3], #侧面
    [P7,P4,P1,P5], #上面
    [P7,P4,P2,P6], #侧面
    [P7,P5,P3,P6]  #背面
]
```

```python
#可视化
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

#绘制向量箭头
#quiver这里可绘制多根箭头,用于向量场 arrow只能画一根箭头
#挨着传入向量单一点坐标可以绘制一个向量
#传入多个数组,判定数组长度匹配后,对应索引的取值为一个点坐标,可以绘制多个向量
ax.quiver(0,0,0,*a1,color = 'red',linewidth = 2)
ax.quiver(0,0,0,*a2,color = 'green',linewidth = 2)
ax.quiver(0,0,0,*a3,color = 'blue',linewidth = 2)

#绘制六面体
face_collection = Poly3DCollection(faces,alpha = 0.3,
                                   facecolor = 'yellow',
                                   edgecolor = 'black',ls = '--')
ax.add_collection3d(face_collection)

#图像装饰
ax.set_xlim([0,3])
ax.set_ylim([0,3])
ax.set_zlim([0,3])
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')
ax.view_init(elev = 30 , azim = 30)
ax.set_proj_type('ortho')
ax.set_box_aspect([1,1,1]) #比例
```
