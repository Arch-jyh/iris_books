```python
def matrix_equality(A,B):
    m,p = A.shape
    q,n = B.shape

    if p == q:
    #A@B
        AB = np.zeros((m,n))
        for i in range(m):
            for j in range(n):
                for k in range(p):
                    AB[i,j] += A[i,k] * B[k,j]

    if m == n:
        #B@A
        BA = np.zeros((q,p))
        for i in range(q):
            for j in range(p):
                for k in range(n):
                    BA[i,j] += B[i,k] * A[k,j]

    return np.array_equal(AB,BA)
```

```python
A = np.array([[1,2],
              [3,4]])
B = np.array([[5,6],
              [7,8]])
```

```python
matrix_equality(A,B)
```

```python
def matrix_multiplication(A,B):
    #获取形状
    m,p_A = A.shape 
    p_B,n = B.shape
    if p_A != p_B:
        raise ValueError('Dimensions do not match')

    C = np.zeros((m,n))
    
    for j in range(n):
        for i in range(m):
            for k in range(p_A):
                C[i,j] += A[i,k] * B[k,j]
    return C                
```

```python
A = np.array([[1,2,3],
              [4,5,6]])
A
```

```python
B = A.T
B
```

```python
matrix_multiplication(A,B)
```

```python
matrix_multiplication(B,A)
```
