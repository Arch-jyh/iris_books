```python
import numpy as np
import seaborn as sns
```

```python
np.random.seed(88)
A = np.random.randint(0,10,(5,12))
A
```

```python
sns.heatmap(A,cmap = 'RdYlBu_r',square = True,
            cbar = False,annot = True,vmin = 0,vmax = 9,
            linecolor = 'w',linewidth = 0.25)
```

```python
A_T = A.T
A_T
```

```python
sns.heatmap(A_T,cmap = 'RdYlBu_r',square = True,
             cbar = False,annot = True,vmin= 0,vmax = 9,
             linecolor = 'w',linewidth = 0.25)
```
