```python
import numpy as np
```

```python
def cal_minor(A,row_idx,col_idx):
    A_ij = np.delete(np.delete(A,row_idx,axis = 0), col_idx,axis = 1)
    return A_ij
```

```python
def _best_axis_and_index(A):
    """
    返回('row' 或 'col' , index)
    优选选择0最多者;若并列,默认选行;若行中多行0值数量相同,仍然并列,取最小索引 #
    """

    n = A.shape[0]
    #A == 0 是求布尔矩阵 求和中axis = 1表示横向压缩,是列的方向,结果是行的和
    #sum(array,axis..)是numpy自带的语法 既有函数形式也有方法形式
    #结果是长度为n的1D数组 axis = n 就是按照n的维度方向进行相加,知道没有东西可加,直到没有东西可加就降维了
    row_zero_counts = (A == 0).sum(axis = 1)
    col_zero_counts = (A == 0).sum(axis = 0)

    max_row_zeros = row_zero_counts.max()
    max_col_zeros = col_zero_counts.max()

    #行优先,相等默认取行
    if max_row_zeros >= max_col_zeros:
        #argmax返回最大值的索引 用int转换成整数型(numpy.int64 >>> python内置的int)
        idx = int(np.argmax(row_zero_counts))
        return 'row',idx

    else:
        idx = int(np.argmax(col_zero_counts))
        return 'col',idx
```
