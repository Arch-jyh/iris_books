```python
import numpy as np
```

```python
A = np.random.randint(0,10,(2,2))
B = np.random.randint(0,10,(2,2))
C = np.random.randint(0,10,(2,2))
D = np.random.randint(0,10,(2,2))
#分配名字,打印方便
mats = [('A',A),('B',B),('C',C),('D',D)]
```

```python
def perms(items):
    if not items:
        return[[]]
    res = []

    for i in range(len(items)):
        #取出第i索引元素
        head = items[i]
        #取出除了i索引元素外的所有元素
        tail = items[:i] + items[i+1:]
        #对剩下的元素做全排列 perms返回的类型是列表
        for p in perms(tail):
            #i索引元素放在前面
            #这里的加法不是矩阵加法,是列表拼接所以给head带上[]表示列表,然后拼接
            res.append([head] + p)
    return res
```

```python
def chain_product(seq):#输入的是一种顺序的组合
    #两个[]表示取两次索引第一次取列表中的元组,第二次取元组中的元素(那个大写的变量,是一个array数组)
    M = seq[0][1]
    #这里将元组取出 从除了输入元组的i为1开始取,不与上面重叠
    #_是一个变量名,但是这里没用,渐渐习惯成不用的变量附在这个变量吗上  这里被赋值的是name,但是不用
    #mat是实际的array数组赋值
    for _,mat in seq[1:]:
        #矩阵乘法按照顺序连乘
        M = M@mat
    return M
```

```python
all_orders = perms(mats)
print('总排列数:',len(all_orders))
```

```python
for seq in all_orders:
    #join表示链接符号,用恰面的字符串作为链接符号,加入到join中的给出的字符串形成新的字符串
    order_str = '@'.join(name for name,_ in seq)
    #计算
    prod = chain_product(seq)
    print(order_str)
    print(prod,'\n')
```
