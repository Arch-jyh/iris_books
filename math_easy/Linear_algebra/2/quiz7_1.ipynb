```python
def intersect(a_11,a_12,a_13,b_1,a_21,a_22,a_23,b_2,a_31,a_32,a_33,b_3):
    A = np.array([[a_11,a_12,a_13],
                  [a_21,a_22,a_23],
                  [a_31,a_32,a_33]])
    b = np.array([[b_1],
                  [b_2],
                  [b_3]])
    if np.linalg.det(A) != 0 :
        x_intersection = np.linalg.solve(A,b)
        x1_intersect,x2_intersect,x3_intersect = x_intersection[0],x_intersection[1],x_intersection[2]
    else:
        x1_intersect,x2_intersect,x3_intersect = None,None,None

    return x1_intersect,x2_intersect,x3_intersect
```

```python
intersect(1,2,3,1,3,0,1,2,1,2,1,3)
```

```python
intersect(1,1,0,2,1,0,-1,1,0,1,-1,0)
```
