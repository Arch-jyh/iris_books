```python
import numpy as np
```

```python
def row_LC(B,coefficients):
    p,n = B.shape
    row_LC_result = np.zeros(n) #初始化组合结果

    for k in range(p):
        row_LC_result += coefficients[k] * B[k,:]

    return row_LC_result
```

```python
def matrix_multiplication_row_LC(A,B):
    m,p_A = A.shape
    p_B,n = B.shape

    if p_A != p_B:
        raise ValueError('Dimensions do not match')

    #传递shape形参(还有其他参数所以mn加括号)
    C = np.zeros((m,n))

    for i in range(m):
        coeffs = A[i,:]
        C[i,:] = row_LC(B,coeffs)

    return C
```

```python
A = np.array([[1,2,3],
              [4,5,6]])
A
```

```python
B = A.T
B
```

```python
matrix_multiplication_row_LC(A,B)
```

```python
matrix_multiplication_row_LC(B,A)
```
