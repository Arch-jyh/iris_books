```python
import numpy as np
```

```python
np.random.seed(0)
```

```python
#rand是生成0-1之间的均匀随机数 然后再放大10倍变成范围为0-10
#0-1是数学上的标准化随机数范围,然后再按需线性变换区间或分布
x = np.random.rand(50)*10 
x
```

```python
true_slope = 0.8
```

```python
true_intercept = 2
```

```python
noise = np.random.rand(50)*2-1  #生成[-1,1]之间的均匀随机数
noise
```

```python
y = true_slope*x + true_intercept + noise
y
```

```python
X = np.column_stack((np.ones_like(x),x))
```

```python
b = np.linalg.inv(X.T@X)@X.T@y
b
```

```python
y_hat = X@b
```

```python
noise_new = y - y_hat
noise_new
```
