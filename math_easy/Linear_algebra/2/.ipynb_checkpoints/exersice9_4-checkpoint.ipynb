```python
A = np.array([
    [2,2,0],
    [2,0,2],
    [0,2,2]])
```

```python
a1 = A[:,[0]]
a2 = A[:,[1]]
a3 = A[:,[2]]
```

```python
eta_2 = a2 - q1@q1.T @a2
```

```python
a2_proj_q1 = q1@q1.T@a2
a2_proj_q1
```

```python
def gram_schmidt_qr(A):
    m,n = A.shape
    Q = np.zeros((m,n))
    R = np.zeros((n,n))

    for i in range(n):
        #将提取出来的1D数组重新配置成2D
            #降低为1D是munpy默认操作,节约内存
        ai = A[:,i].reshape(-1,1)
        for j in range(i):
            #同理提取出的1D数组配置成2D
            qj = Q[:,j].reshape(-1,1)
            #item表示将numpy数组数组转换为标量数字
            #是numpy自带方法.用于仅有一个元素的数组,取出值本身
            R[j,i] = (qj.T@ai).item()
            ai = ai - R[j,i]*qj #正交投影剔除

        norm = np.linalg.norm(ai)
        if norm < 1e-10:
            raise ValueError(f"第{i+1}列向量线性相关,无法正交化")

        Q[:,i] = (ai/norm).flatten()
        R[i,i] = norm
    return Q,R
```
