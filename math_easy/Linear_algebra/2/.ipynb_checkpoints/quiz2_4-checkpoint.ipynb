```python
import numpy as np
```

```python
def matrix_equality(A,B):
    m,p = A.shape
    q,n = B.shape

    if p == q:
    #A@B
        AB = np.zeros((m,n))
        for i in range(m):
            for j in range(n):
                for k in range(p):
                    AB[i,j] += A[i,k] * B[k,j]

    if m == n:
        #B@A
        BA = np.zeros((q,p))
        for i in range(q):
            for j in range(p):
                for k in range(n):
                    BA[i,j] += B[q,k] * A[k,p]

    return AB == BA
```

```python
A = np.array([[1,2],
              [3,4]])
B = np.array([[5,6],
              [7,8]])
```

```python
matrix_equality(A,B)
```
