```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
directed_G = nx.DiGraph()
```

```python
directed_G.add_nodes_from(['a','b','c','d'])
```

```python
directed_G.add_edges_from([('a','b'),
                           ('b','a'),
                           ('a','c'),
                           ('d','a'),
                           ('b','c'),
                           ('c','d'),
                           ('d','c')])
```

```python
pos = {'a':[0,0],
       'b':[1,0],
       'c':[1,1],
       'd':[0,1]}
```

```python
plt.figure(figsize = (6,6))
nx.draw_networkx(directed_G,pos = pos,node_size = 188,
                 #自动解析字符串(参数说明方法是固定的) arc是链接样式,弧线型是第三种  rad是弯曲半径
                 connectionstyle = 'arc3,rad = 0.1')
```

```python
A = nx.adjacency_matrix(directed_G).toarray()
A
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
adjacency_matrix = np.array([[0,1,1,0],
                             [1,0,1,0],
                             [0,0,0,1],
                             [1,0,1,0]])
```

```python
#邻接矩阵创建有向图
derected_G = nx.DiGraph(adjacency_matrix,nodetype = int)
mapping = {0:'a',1:'b',2:'c',3:'d'}
directed_G = nx.relabel_nodes(directed_G,mapping)
```
