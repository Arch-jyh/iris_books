```python
#q2 
#np.block(arrays)   注意接收的是arrays 单个高维array有形状不能拼接  传递array组成的列表按照他们的形状拼接
#按照给出的列表进行拼接,顺序: 最内层用-1拼接 外一层用-2拼接 一直到最外层
    #不看传入参数最外层维度,直接按照[]对应层数拼接,可用[]指定拼接哪一层
#除了当前轴长度可以不同,其他轴长度必须相同(不做广播)
    #维度不足的块,不足位置对应高纬度数组轴长度也为1,补充前导轴 这个逻辑可以用在标量 (1,1,2)与(3)  后者变为(1,1,3)
#axis = -1永远指的是最后一个维度,不用看行列方向(高维行列方向不准确)
#其他:
    #仅用标量调用与np.array()作用相同
```

```python
import numpy as np
A = np.eye(2)*2
B = np.eye(3)*3
np.block([
    [A,np.zeros((2,3))],
    [np.zeros((3,2)),B]])
```

```python
np.block([1,2,3])
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
np.block([a,b,10]) #函数设计之初传入一个表示整体的参数arrays因需要确定列表结构并确定轴顺序
```

```python
A = np.ones((2,2),int)
B = 2*A
np.block([A,B])
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
np.block([[a],[b]])
```

```python
A = np.ones((2,2),int)
B = A*2
np.block([[A],[B]])
```

```python
a = np.array(0)
b = np.array([1])
np.block([a])  #与输入值对应位置的拼接维度与产生维度相同
```

```python
np.block([b])
```

```python
np.block([[a]])
```

```python
np.block([[0]])
```

```python
A = np.array([[0,5],
              [3,4],
              [5,0]])
B = A.T
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
lincolors1 = [[0.9,0.5,0.2],
             [0.2,0.9,0.5],
             [0.2,0.5,0.9]]
lincolors2 = [[0.9,0.5,0.2],
              [0.9,0.2,0.9]]
```

```python
fig1 = plt.figure(figsize = (6,12))
ax1 = fig1.add_subplot(211)
for i in np.arange(0,3):
    ax1.quiver(0,0,A[i,0],A[i,1],color = lincolors1[i],
               angles = 'xy' , scale_units = 'xy',
               scale = 1,zorder = 1000
              )
ax1.set_xlim(0,5)
ax1.set_ylim(0,5)
ax1.set_xticks(np.arange(0,5.1,1))
ax1.set_yticks(np.arange(0,5.1,1))
ax1.grid(True,linestyle = '--',linewidth = 0.5,color = '0.8')
ax1.set_aspect('equal')


ax2 = fig1.add_subplot(212,projection = '3d')
for i in np.arange(0,2):
    ax2.quiver(0,0,0,A[0,i],A[1,i],A[2,i],color = lincolors2[i],lw = 0.5
              ) 
ax2.set_xlim(0,5)
ax2.set_ylim(0,5)
ax2.set_zlim(0,5)
ax2.set_xticks(np.arange(0,5.1,1))
ax2.set_yticks(np.arange(0,5.1,1))
ax2.set_zticks(np.arange(0,5.1,1))
ax2.grid(True,linestyle = '--',linewidth = 0.5,color = '0.8')
ax2.set_aspect('equal')
ax2.view_init(elev = 30,azim = 35)
```

```python
fig2 = plt.figure(figsize = (6,12))

ax1 = fig2.add_subplot(211,projection = '3d')
for i in np.arange(0,2):
    ax1.quiver(0,0,0,B[i,0],B[i,1],B[i,2],color = lincolors2[i],lw = 0.5
              ) 
ax1.set_xlim(0,5)
ax1.set_ylim(0,5)
ax1.set_zlim(0,5)
ax1.set_xticks(np.arange(0,5.1,1))
ax1.set_yticks(np.arange(0,5.1,1))
ax1.set_zticks(np.arange(0,5.1,1))
ax1.grid(True,linestyle = '--',linewidth = 0.5,color = '0.8')
ax1.set_aspect('equal')
ax1.view_init(elev = 30,azim = 35)


ax2 = fig2.add_subplot(212)
for i in np.arange(0,3):
    ax2.quiver(0,0,B[0,i],B[1,i],color = lincolors1[i],
               angles = 'xy' , scale_units = 'xy',
               scale = 1,zorder = 1000
              )
ax2.set_xlim(0,5)
ax2.set_ylim(0,5)
ax2.set_xticks(np.arange(0,5.1,1))
ax2.set_yticks(np.arange(0,5.1,1))
ax2.grid(True,linestyle = '--',linewidth = 0.5,color = '0.8')
ax2.set_aspect('equal')
```
