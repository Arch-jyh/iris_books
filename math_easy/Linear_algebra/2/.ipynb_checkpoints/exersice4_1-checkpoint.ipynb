```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[2,1],
              [1,3]])
```

```python
a1 = A[:,0]
a2 = A[:,1]
```

```python
origin = np.array([0,0]) #配置起点
P1 = origin
P2 = a1
P3 = a1+a2
P4 = a2
parallelogram = np.array([P1,P2,P3,P4])
```

```python
#jupyter会自动在创建fig的cell中运行当前cell的绘图程序后自动展示
#jupyter自动检查挡墙call中有没有绘图对象fig,如果有就调用display(fig)进行显示
#如果后面的ax.等方法分call进行绘制,jupyter检查时这个call中没有fig就不会绘制,ax.只是方法修改了轴对象,而不是绘图对象(结果对象)
#jupyter这样配置是逻辑是每个call只展示最后一个结果对象(fig绘图对象)
#后面的call中使用了ax或者plt等方法后可以输入fig结果对象进行输出,像是a = 1 (换行) a 输出1,因为a是结果对象.
fig,ax = plt.subplots()

ax.arrow(*origin,*a1,head_width = 0.2,
         length_includes_head = True,color = 'red')
ax.arrow(*origin,*a2,head_width = 0.2,
         length_includes_head = True,color = 'green')

ax.fill(parallelogram[:,0],parallelogram[:,1],
        edgecolor = 'k',linestyle = '--',
        #alpha是图像第四个通道,表示合成时的权重,控制前景和背景颜色的比例,叫做混合系数,数学爱用alpha表示.自然表示了表示透明度
        alpha = 0.3,facecolor = 'yellow')

ax.set_aspect('equal')
```
