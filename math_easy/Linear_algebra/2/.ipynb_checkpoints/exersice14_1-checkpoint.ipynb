```python
import numpy as np
```

```python
A = np.array([[0,1],
              [1,1],
              [1,0]])
```

```python
ATA = A.T @ A
```

```python
eigvals_V,V = np.linalg.eigh(ATA)
```

```python
eigvals_V
```

```python
#argsort默认返回升序排序后的索引顺序
#赋值操作复制给本身会覆盖改变原数组,-A只能是生成新数组
idx = np.argsort(-eigvals_V)

eigvals_V = eigvals_V[idx]

V = V[:,idx]
```

```python
eigvals_V
```

```python
V
```

```python
#首先取元素最大值逐个与0比较 防止有0值因为浮点误差产生负值而被开根号,直接比较大小取0值
    #每次从参数1中按照顺序取出一个值与后面的数组比较
    #这里是用到了广播规则,后者0广播变化成了与前者数组相同的结构,按照索引位置依次对比两个数组上的元素大小
        #返回一个同形状数组,记录对比下的最大值
singular_vals = np.sqrt(np.maximum(eigvals_V,0))
```

```python
singular_vals
```

```python
#构造S
```

```python
S = np.zeros_like(A,dtype = float) #dtype的d表示data
```

```python
#原地填充到主对角线上 参数2的1D数组长度需要≤S的对角线长度
#fill原意为填充,不需要填满,参数2的1D数组长度小于被填充矩阵对角线长度,剩余部分不填
np.fill_diagonal(S,singular_vals)
S
```

```python
#计算AAT
```

```python
AAT = A @ A.T
```

```python
eigvals_U , U = np.linalg.eigh(AAT)
```

```python
eigvals_U , U
```

```python
idx = np.argsort(-eigvals_U)

eigvals_U = eigvals_U[idx]

U = U[:,idx]
```

```python
idx
```

```python
eigvals_U,U
```

```python
#还原A
```

```python
A_reconstructed = U@S@V.T
#四舍五入保留5位小数
np.round(A_reconstructed,5)
```

```python
#用numpy
```

```python
U,S,V = np.linalg.svd(A,full_matrices = True)
```

```python
U,S,V
```

```python
S_s = np.zeros_like(A,dtype = float)
np.fill_diagonal(S_s,S)
```

```python
A_reconstructed_2 = U @ S_s @ V.T
A_reconstructed_2
```
