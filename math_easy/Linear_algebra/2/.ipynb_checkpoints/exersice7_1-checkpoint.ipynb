```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def intersect(a_11,a_12,b_1,a_21,a_22,b_2):
    A = np.array([[a_11,a_12],
                  [a_21,a_22]])

    b = np.array([b_1,b_2])

    #判断是否有唯一解
    if np.linalg.det(A) != 0:
        x_intersection = np.linalg.solve(A,b)
        x1_intersect,x2_intersect = x_intersection[0],x_intersection[1]
    else:
        x1_intersect,x2_intersect = None,None
    return x1_intersect,x2_intersect
```

```python
def visualize(a_11,a_12,b_1,a_21,a_22,b_2):
    #定义绘图范围
    x1 = np.linspace(-4,4,400)
    x2 = np.linspace(-4,4,400)
    xx1,xx2 = np.meshgrid(x1,x2)

    #计算每一个点的方程的值
    zz1 = a_11*xx1 + a_12*xx2 - b_1
    zz2 = a_21*xx1 + a_22*xx2 - b_2

    #绘制等高线
    plt.figure(figsize = (5,5))
    #levels被传入了[0]值.只绘制值为0的等高线.levels只在给的值的位置绘制等高线,而不是自动分级绘制
    plt.contour(xx1,xx2,zz1,levels = [0],colors = 'red',linewidths = 2,linestyles = 'solid')
    plt.contour(xx1,xx2,zz2,levels = [0],colors = 'blue',linewidths = 2,linestyles = 'solid')

    #计算交点
    x1_intersect,x2_intersect = intersect(a_11,a_12,b_1,a_21,a_22,b_2)

    #绘制交点
    if x1_intersect is not None and x2_intersect is not None:
        plt.scatter(x1_intersect,x2_intersect,color = 'black',
                    marker = 'o',s = 100,label = 'Intersection')

    #装饰
    plt.xlabel('$x_1$',fontsize = 12)
    plt.ylabel('$x_2$',fontsize = 12)
    plt.grid(True,linestyle = '--',color = '0.8')
    plt.axhline(0,color = 'black',linewidth = 1)
    plt.axvline(0,color = 'black',linewidth = 1)
    #返回当前ax,调用ax的函数
    plt.gca().set_aspect('equal',adjustable = 'box')
```

```python
a_11,a_12,b_1 = 1,-1, 2
a_21,a_22,b_2 = 3, 3, 0
```

```python
visualize(a_11,a_12,b_1,a_21,a_22,b_2)
```
