```python
import numpy as np
```

```python
def scale_2D(x1,x2,s1,s2):
    x = np.array([[x1],
                  [x2]])
    s = np.array([[s1,0],
                  [0,s2]])

    x_s = s@x
    print(x_s)
```

```python
scale_2D(1,2,3,2)
```
