```python
import numpy as np
```

```python
A = np.array([[1,2,3],
              [4,5,6]])
A
```

```python
B = A.T
B
```

```python
C = A@B
C
```

```python
A[[0],:]
```

```python
B[:,[0]]
```

```python
A[[0],:] @ B[:,[0]]
```

```python
rows,cols = C.shape
C_manual = np.zeros((rows,cols))
```

```python
for i in range(rows):
    for j in range(cols):
        c_ij = A[[i],:] @ B[:,[j]]  #得到1x1的二维矩阵
        c_ij = c_ij[0][0]  #提取二维矩阵中的标量
        C_manual[i,j] = c_ij #将标量赋予给正确位置
```
