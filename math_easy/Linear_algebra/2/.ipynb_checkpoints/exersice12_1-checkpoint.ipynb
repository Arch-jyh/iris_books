```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
data = load_iris()
#设置列标签为特征名
X_df_iris = pd.DataFrame(data.data,columns = data.feature_names)
```

```python
target = data.target
target_names = []
for i in np.arange(len(target)):
    target_name_idx = target[i]
    target_name = data.target_names[target_name_idx]
    target_names.append(target_name)

X_df_iris['species'] = target_names
```

```python
X_df_iris.head()
```

```python
X_df_iris.tail()
```

```python
X_df_iris.columns
```

```python
#iloc是按照行列下标提取[行,列]
#这是df数据,不是numpy数组,所以使用iloc提取数据 interger location(整数位置索引)
X_df = X_df_iris.iloc[:,:4]

X_df.columns = ['Sepal length (cm)',
                'Sepal width (cm)',
                'Petal lenght (cm)',
                'Petal widht (cm)']
```

```python
X_df.head()
```

```python
fig,ax = plt.subplots(figsize = (3,6))
sns.heatmap(X_df.iloc[:,:4],ax = ax,
            yticklabels = [],
            annot = False,
            cmap = 'viridis',
            square = False,
            cbar_kws = {'shrink':0.5})
```

```python
sns.pairplot(X_df,diag_kind = 'kde')
```

```python
#pandas中默认按照列求均值 
#得到1D数组后自动从df数组自动封装转换到series一维数组
    #是数组+索引的形式,左边的竖着这是索引,这里的索引是用的索引名,当然也可以用iloc进行整数索引(和numpy规则不一样)
mean_X = X_df.mean()
mean_X
```

```python
n = len(X_df)
all_ones_n = np.ones((n,1))
```

```python
all_ones_n.T @ X_df
```

```python
all_ones_n.T @ X_df/n
```

```python
g = sns.pairplot(X_df)
for i in range(len(g.axes)):
    for j in range(len(g.axes)):
        if i != j:
            ax = g.axes[i,j]
            ax.scatter(mean_X.iloc[j],mean_X.iloc[i],
                       color = 'red',marker = 'x',s = 188,zorder = 5)
            ax.axhline(y = mean_X.iloc[i],c = 'r',ls = '--')
            ax.axvline(x = mean_X.iloc[j],c = 'r',ls = '--')

        else:
            ax = g.axes[i,j]
            ax.axvline(x = mean_X.iloc[j],c = 'r',ls = '--')
```

```python
#中心化
```

```python
X_df_demean = X_df - X_df.mean()
X_df_demean.columns = ['Sepal length, demean (cm)',
                      'Sepal width, demean (cm)',
                      'Petal length, demena (cm)',
                      'Petal width, demean (cm)']
```

```python
mean_X_demean = X_df_demean.mean()
mean_X_demean
```

```python
fig,ax = plt.subplots(figsize = (3,6))
sns.heatmap(X_df_demean,ax = ax,
            yticklabels = [],
            annot = False,#控制是否在方格中绘制数值  annotate缩写,表示注释
            cmap = 'viridis',
            square = False,
            cbar_kws = {'shrink':0.5})#控制颜色条长度比例 0.5为原来的一半
```

```python
g = sns.pairplot(X_df_demean)

#range是迭代器不是生成两个数
for i in range(len(g.axes)):
    for j in range(len(g.axes)):
        if i != j:
            ax = g.axes[i,j]
            ax.scatter(mean_X_demean.iloc[j],mean_X_demean.iloc[i],
                        color = 'red',marker = 'x',s = 188,zorder = 5)
```

```python
#协方差矩阵
```

```python
Sigma = X_df.cov() #协方差
#.str表示使用名称形式的标签(列索引)接口然后进行操作
#regex表示不是正则,是一种符号匹配字符串的语法
Sigma.columns = Sigma.columns.str.replace('(cm)','',regex = False)
Sigma.index = Sigma.index.str.replace('(cm)','',regex = False)
Sigma
```

```python
X_df_demean.T @ X_df_demean / (n-1)
```

```python
fig,ax = plt.subplots(figsize = (3,3))
sns.heatmap(Sigma,ax = ax,
            annot = True,
            cmap = 'viridis',
            square = True)
```
