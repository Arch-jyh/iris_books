```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx,yy = np.meshgrid(np.linspace(-8,8,1000),
                    np.linspace(-8,8,1000))
```

```python
zz = np.abs(xx) + np.abs(yy)
```

```python
e1 = np.array([1,0])
e2 = np.array([0,1])

color_e1 = [0,0,0]
color_e2 = [0,0,0]
```

```python
plt.figure(figsize = (7.5,6))
plt.quiver(0,0,e1[0],e1[1],
           #向量朝向用xy轴计算,向量长度按照数据坐标系来度量缩放,
           angles = 'xy' , scale_units = 'xy', #单位和方向都是基于坐标轴
           #不进行缩放,保持原大小
           scale = 1,color = color_e1,
           #添加箭头标签   zorder = 1000 表示箭头放在最上边
           label = 'e1',zorder = 1000)

plt.quiver(0,0,e2[0],e2[1],
           angles = 'xy', scale_units = 'xy',
           scale = 1,color = color_e2,
           label = 'e2',zorder = 1000)

#这里zz的值洗漱或者稠密都能画等值线,找出指定的水平线插值后将所有点连接成曲线,但是过于稀疏的可能导致不精确
#函数绘图思维:先用xxyyzz当做离散采样空间数据,再插值找出等值线经过位置,在计算坐标,最后按照屏幕像素渲染.
contour = plt.contour(xx,yy,zz,
                     levels = np.arange(1,9),
                      cmap = 'RdYlBu_r')

cbar = plt.colorbar(contour)
plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axhline(0,color = 'k',linewidth = 0.5)
plt.axvline(0,color = 'k',linewidth = 0.5)
plt.xticks(np.arange(-8,9,2))
plt.yticks(np.arange(-8,9,4))
plt.grid(True,which = 'both',linestyle = '--',linewidth = 0.5,color = '0.8')

plt.gca().set_aspect('equal',adjustable = 'box')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx,yy = np.meshgrid(np.linspace(-8,8,1000),
                    np.linspace(-8,8,1000))
```

```python
p = 2
#**才表示幂运算,^不是幂运算而是按位异或运算  数学公式中^才是上角标,这是数学排版和纯文本的几号,非pyhon编程
zz = (xx**p + yy**p)**(1/p) 
```

```python
e1 = np.array([1,0])
e2 = np.array([0,1])

color_e1 = [0,0,0]
color_e2 = [0,0,0]
```

```python
plt.figure(figsize = (8,6))
plt.quiver(0,0,e1[0],e1[1],
           #scale_units scale两个参数需要搭配使用,一个确定缩放基准一个确定缩放比例
          angles = 'xy',scale_units = 'xy',
          scale = 1,color = color_e1,
          label = 'e1',zorder = 1000)

plt.quiver(0,0,e2[0],e2[1],
           angles = 'xy',scale_units = 'xy',
           scale = 1,color = color_e2,
           label = 'e2',zorder = 1000)

contour = plt.contour(xx,yy,zz,
                      levels = np.arange(1,9),
                      cmap = 'RdYlBu_r')

cbar = plt.colorbar(contour) #

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axhline(0,color = 'k',linewidth = 0.5)
plt.axvline(0,color = 'k',linewidth = 0.5)

plt.xticks(np.arange(-8,9,2))
plt.yticks(np.arange(-8,9,2))
#color参数为字符串,相当于其他参数'red'等颜色字符串参数的同一写法,这个数字表示灰度字符串数字参数
plt.grid(True,which = 'both',linestyle = '--',linewidth = 0.5,color = '0.8') 

plt.gca().set_aspect('equal',adjustable = 'box')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx,yy = np.meshgrid(np.linspace(-8,8,1000),
                    np.linspace(-8,8,1000))
#使用maximum
zz = np.maximum(np.abs(xx),np.abs(yy))
```

```python
e1 = np.array([1,0])
e2 = np.array([0,1])

color_e1 = [0,0,0]
color_e2 = [0,0,0]
```

```python
plt.figure(figsize = (8,6))

plt.quiver(0,0,e1[0],e1[1],
          angles = 'xy',scale_units = 'xy',
          scale = 1,color = color_e1,
          label = 'e1',zorder = 1000)
plt.quiver(0,0,e2[0],e2[1],
          angles = 'xy',scale_units = 'xy',
          scale = 1,color = color_e2,
          label = 'e1',zorder = 1000)

contour = plt.contour(xx,yy,zz,
                      levels = np.arange(1,9),
                     cmap = 'RdYlBu_r')

cbar = plt.colorbar(contour)

plt.xlim(-8,8);plt.ylim(-8,8)
plt.axhline(0,color = 'k',linewidth = 0.5);plt.axvline(0,color = 'k',linewidth = 0.5)
plt.xticks(np.arange(-8,9,2));plt.yticks(np.arange(-8,9,4))
#which = 'both表示主动画ticks的主次刻度线
#主刻度表示整数刻度或者是ticks配置刻度,次刻度一般为小数刻度或者ticks之间的刻度
plt.grid(True,linewidth = 0.5,color = '0.8',linestyle = '--',which = 'both')

plt.gca().set_aspect('equal',adjustable = 'box')
```
