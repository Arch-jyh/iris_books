```python
#可视化向量乘法
import numpy as np
import matplotlib.pyplot as plt
```

```python
a_vec = np.array([2,1])
```

```python
k = -2
k_a_vec = k*a_vec
```

```python
#可视化
plt.figure(figsize = (6,6))

#绘制向量
plt.arrow(0,0,a_vec[0],a_vec[1],
          head_width = 0.2,head_length = 0.3,
          #facecolor表示箭头填充颜色 edgecolor表示箭头边缘颜色
         fc = '#00B0F0',ec = '#00B0F0',label = 'a')
plt.arrow(0,0,k_a_vec[0],k_a_vec[1],
          head_width = 0.2,head_length = 0.3,
          fc = '#92D050',ec = '#92D050',label = 'k*a')

#装饰
#gca表示get current axes 表示获取当前活跃axes对象 因为后面的方法是axes对象的方法,不是plt的顶级函数
plt.gca().set_aspect('equal',adjustable = 'box')
plt.xlim(-5,5);plt.ylim(-5,5)
#配置竖屏和垂直线 horizontal vertical
plt.axhline(0,color = 'black',linewidth = 0.5)
plt.axvline(0,color = 'black',linewidth = 0.5)
plt.xticks(np.arange(-5,6))
plt.yticks(np.arange(-3,4))
plt.grid(color = 'gray',alpha = 0.8,linestyle = '-',linewidth = 0.25)
plt.legend()
```
