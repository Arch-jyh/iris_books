```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define vectors
origin = np.array([0, 0, 0])
green = np.array([0, 1, 0])
white = np.array([1, 1, 1])
yellow = np.array([1, 1, 0])

# Compute projections
proj_green_on_white = (np.dot(green, white) / np.dot(white, white)) * white
proj_green_on_yellow = (np.dot(green, yellow) / np.dot(yellow, yellow)) * yellow

# Plotting
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.set_box_aspect([1,1,1])

# Plot axes
ax.quiver(*origin, 1.2, 0, 0, color='gray', arrow_length_ratio=0.05)
ax.quiver(*origin, 0, 1.2, 0, color='gray', arrow_length_ratio=0.05)
ax.quiver(*origin, 0, 0, 1.2, color='gray', arrow_length_ratio=0.05)
ax.text(1.25, 0, 0, 'X', color='gray')
ax.text(0, 1.25, 0, 'Y', color='gray')
ax.text(0, 0, 1.25, 'Z', color='gray')

# Plot vectors
ax.plot([0, green[0]], [0, green[1]], [0, green[2]], color='green', label='Green')
ax.plot([0, white[0]], [0, white[1]], [0, white[2]], color='black', label='White')
ax.plot([0, yellow[0]], [0, yellow[1]], [0, yellow[2]], color='orange', label='Yellow')

# Plot projections
ax.scatter(proj_green_on_white[0], proj_green_on_white[1], proj_green_on_white[2], color='blue', s=50, label='Proj of Green on White')
ax.scatter(proj_green_on_yellow[0], proj_green_on_yellow[1], proj_green_on_yellow[2], color='red', s=50, label='Proj of Green on Yellow')

# Draw projection lines
ax.plot([green[0], proj_green_on_white[0]], [green[1], proj_green_on_white[1]], [green[2], proj_green_on_white[2]], color='blue', linestyle='--')
ax.plot([green[0], proj_green_on_yellow[0]], [green[1], proj_green_on_yellow[1]], [green[2], proj_green_on_yellow[2]], color='red', linestyle='--')

# Annotations
ax.text(proj_green_on_white[0], proj_green_on_white[1], proj_green_on_white[2], '1/3 of White', color='blue')
ax.text(proj_green_on_yellow[0], proj_green_on_yellow[1], proj_green_on_yellow[2], '1/2 of Yellow', color='red')

# Labels and legend
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.title('Projection of Green onto White (1/3) and Yellow (1/2)')
plt.show()
```

```python
import numpy as np
```

```python
a = np.array([3,4])
```

```python
def decompose(a,theta):
    v1 = np.array([np.cos(theta),np.sin(theta)])
    v2 = np.array([-np.sin(theta),np.cos(theta)])

    #标量投影
    a_v1 = np.dot(a,v1)
    a_v2 = np.dot(a,v2)

    #向量投影
    proj_a_v1 = a_v1*v1
    proj_a_v2 = a_v2*v2

    return proj_a_v1,proj_a_v2
```

```python
theta = 30
theta = theta/180*np.pi
```

```python
proj_a_v1,proj_a_v2 = decompose(a,theta)
```
