```python
#平行四边形加法
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a_vec = np.array([4,1])
b_vec = np.array([1,2])
```

```python
a_plus_b = a_vec + b_vec
```

```python
#可视化
plt.figure(figsize = (6,6))

#绘制向量a
#1,2个参数指定起点,3,4个参数指定方向和长度(xy的两个分量)
plt.quiver(0,0,
           a_vec[0],a_vec[1],
           #按照标准xy坐标轴绘制,让向量长度按照坐标轴刻度缩放,缩放比例为1(保持原样)
           angles = 'xy',scale_units = 'xy',scale = 1,
           #红色,图例标记为a
           color = 'r',label = 'a')

#绘制向量b
plt.quiver(0,0,
           b_vec[0],b_vec[1],
           angles = 'xy',scale_units = 'xy',scale = 1,
           color = 'b',label = 'b')

#绘制向量a+b
plt.quiver(0,0,
           a_plus_b[0],a_plus_b[1],
           angles = 'xy',scale_units = 'xy',scale = 1,
           color = 'pink',label = 'a+b')

#画出平行四边形另外两条线
#函数参数第一个为x_values 第二个为y_values
plt.plot([b_vec[0],a_plus_b[0]],
         [b_vec[1],a_plus_b[1]],'k--')

plt.plot([a_vec[0],a_plus_b[0]],
         [a_vec[1],a_plus_b[1]],'k--')

#装饰
#坐标比例标尺一直
plt.gca().set_aspect('equal',adjustable = 'box')
plt.xlim(0,5);plt.ylim(0,5)
#画出x,y轴的喝死基准线
plt.axhline(0,color = 'black',linewidth = 0.5)
plt.axvline(0,color = 'black',linewidth = 0.5)
#'-'表示实线
plt.grid(color = 'gray',alpha = 0.8,linestyle = '-',linewidth = 0.25)
#自动读取绘图时的label绘制图例
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a_vec = np.array([4,1])
b_vec = np.array([1,2])
```

```python
a_plus_b = a_vec + b_vec
```

```python
plt.figure(figsize = (6,6))

plt.quiver(0,0,a_vec[0],a_vec[1],
           angles = 'xy',scale_units = 'xy',scale=1,
           color = 'r',label = 'a')
plt.quiver(a_vec[0],a_vec[1],b_vec[0],b_vec[1],
           angles = 'xy',scale_units = 'xy' ,scale = 1,
           color = 'b',label= 'b')
plt.quiver(0,0,a_plus_b[0],a_plus_b[1],
           angles = 'xy',scale_units = 'xy',scale = 1,
           color = 'pink',label = 'a+b')

plt.gca().set_aspect('equal',adjustable = 'box')
plt.xlim(0,5);plt.ylim(0,5)
plt.axhline(0,color = 'black',linewidth = 5)
plt.axvline(0,color = 'black',linewidth = 5)
plt.grid(color = 'gray',alpha = 0.3,linestyle = '--',linewidth = 1)

plt.legend()
```
