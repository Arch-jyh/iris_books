```python
a = np.array([3,-4])
b = np.array([0,5])
```

```python
p_array = np.arange(1,10.1,0.5)
a_Lp_array = np.array([])
b_Lp_array = np.array([])
```

```python
for p in p_array:
    #作为对比最后一句引出方法函数概念,这里linalg没有使用()是因为就是一个模块的对象,然后找到这个对象的方法
    a_Lp = np.linalg.norm(a,ord = p)
    b_Lp = np.linalg.norm(b,ord = p)

    a_Lp_array = np.append(a_Lp_array,a_Lp)
    b_Lp_array = np.append(b_Lp_array,b_Lp)
```

```python
fig = plt.figure()

plt.plot(p_array,a_Lp_array,marker = 'o')
plt.plot(p_array,b_Lp_array,marker = 's')

plt.xlim(1,10)
plt.ylim(4,7)

plt.xticks(np.arange(1,10.1,0.5))
plt.yticks(np.arange(4,7.1,0.5))

plt.grid(True,linestyle = '--',c = '0.8')
#gca后面的()表示引用了gca并执行了这个函数返回一个axes对象,后面的.set是axes对象的方法
#方法也是函数,函数可以独立存在,但是方法必须依赖于对象
#plt.gca().set_aspect('equal',adjustable = 'box')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx,yy = np.meshgrid(np.linspace(-8,8,1000),
                    np.linspace(-8,8,1000))

zz1 = np.abs(xx)+np.abs(yy)
zz2 = np.sqrt(xx**2 + yy**2)
#maximum逐个元素比较两个数组,每个位置取较大值输出  输入两个参数必须形状一样
zz3 = np.maximum(np.abs(xx),np.abs(yy))

zz_all = np.array([zz1,zz2,zz3])

e1 = np.array([1,0])
e2 = np.array([0,1])
```

```python
fig,axs = plt.subplots(3,1,figsize = (8,18))
i = 0
for ax in axs:
    zz = zz_all[i]
    ax.quiver(0,0,e1[0],e1[1],
              angles = 'xy',scale_units = 'xy',
              scale = 1,color = [0,0,0],
              label = 'e1',zorder = 1000)
    ax.quiver(0,0,e2[0],e2[1],
          angles = 'xy',scale_units = 'xy',
          scale = 1,color = [0,0,0],
          label = 'e2',zorder = 1000)

    contourf = ax.contourf(xx,yy,zz,
                           levels = np.arange(1,9),
                           cmap = 'RdYlBu_r')

    #colorbar这个函数智能使用fig和plt绘制 是这两个其中的函数 前者会更干净因为指定了fig 后者需要多查找一次fig全局状态
    cbar = fig.colorbar(contourf,ax = ax)

    ax.set_xlim(-8,8);ax.set_ylim(-8,8)
    ax.axhline(0,color = 'k' , linewidth = 0.5)
    ax.axvline(0,color = 'k' , linewidth = 0.5)

    ax.set_xticks(np.arange(-8,9,2));ax.set_yticks(np.arange(-8,9,2))

    ax.grid(True,which = 'both',linestyle = '--',linewidth = 0.5,color = '0.8')

    ax.set_aspect('equal')

    i = i+1
```
