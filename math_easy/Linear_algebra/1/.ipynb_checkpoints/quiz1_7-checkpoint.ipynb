```python
a = np.array([3,4])
a_norm = np.linalg.norm(a)
```

```python
#错误的,因为这算的是正交投影,只有两个正交的向量,进行正交投影的到的投影向量相加才能得到原向量
for i in np.arange(10):
    #生成v1 v2向量
    v1 = np.random.uniform(-10,10,size=2)
    v2 = np.random.uniform(-10,10,size=2)

    #计算v1 v2模长
    v1_norm = np.linalg.norm(v1)
    v2_norm = np.linalg.norm(v2)

    #求v1v2的单位向量
    v1_unit = v1 / v1_norm
    v2_unit = v2 / v2_norm

    #计算a与v1v2的点积
    dot_a_v1 = np.dot(a,v1)
    dot_a_v2 = np.dot(a,v2)

    #计算a在v1v2方向上的投影
    proj_a_on_v1 = (dot_a_v1 / v1_norm) * v1_unit
    proj_a_on_v2 = (dot_a_v2 / v2_norm) * v2_unit


    print('第 %r 个关于a的随机方向向量分解:' %(i+1))
    print('随机向量v1为: [%.2f , %.2f] , 随机向量v2为: [%.2f , %.2f]' %(v1[0],v1[1],v2[0],v2[1]))
    print('在v1方向投影向量为: [%.2f , %.2f]' %(proj_a_on_v1[0],proj_a_on_v1[1]))
    print('在v2方向投影向量为: [%.2f , %.2f]' %(proj_a_on_v2[0],proj_a_on_v2[1]))
    print('\n')
```

```python
#正确的不是正交投影,而是向量分解,能够将a分解为两个向量 满足v1+v2 = a,给出10组[v1,v2]
#还有注意不是确定v1 求出v2 = a - v1 因为这样是将a通过v1拆解到v2 不是题目的知道a,v1,v2将a分解到v1,v2
#问题的核心是先知道v1 v2的单位向量,根据单位向量进行分解(不是正交投影)
```

```python
#以下是q4改正思路:  (但是两个分量需要分量方向单位向量乘以系数,有两个方向就有c1,c2两个系数)
#这个思路求两个系数需要用到solve,但是没学这个语法和数学解题思路.
```

```python
import numpy as np

a = np.array([3.,4.])

for i in range(10):
    # 随机生成不共线的 v1, v2
    while True:
        v1 = np.random.uniform(-10, 10, size=2)
        v2 = np.random.uniform(-10, 10, size=2)
        if abs(np.linalg.det(np.column_stack([v1, v2]))) > 1e-6:
            break

    V = np.column_stack([v1, v2])      # 2x2，列为基向量
    c = np.linalg.solve(V, a)          # 求系数 c1, c2
    a1, a2 = c[0]*v1, c[1]*v2          # 分解向量
    rec = a1 + a2

    print(f'#{i+1} v1={v1}, v2={v2}')
    print(f'c1={c[0]:.3f}, c2={c[1]:.3f}, 误差‖a-(a1+a2)‖={np.linalg.norm(a-rec):.2e}\n')
```
