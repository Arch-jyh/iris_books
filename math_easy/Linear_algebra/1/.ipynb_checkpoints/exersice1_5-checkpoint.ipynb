```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta_array = np.linspace(0,2*np.pi,300)
x1_array = np.cos(theta_array)
x2_array = np.sin(theta_array)
```

```python
#可视化
fig,ax = plt.subplots(figsize = (6,6))

#绘制单位圆
ax.plot(x1_array,x2_array,'b',linewidth = 1)

#设置XY轴比例相同
ax.set_aspect('equal')
ax.grid(True)
ax.set_xlim([-2,2])
ax.set_ylim([-2,2])
ax.axhline(0,color = 'k',linewidth = 0.5)
ax.axvline(0,color = 'k',linewidth = 0.5)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import hsv
```

```python
#创建数据
u = np.linspace(0,2*np.pi,37)
v = np.linspace(0,np.pi,19)

uu,vv = np.meshgrid(u,v)
r = 1

#计算球面坐标
xx = r*np.sin(vv)*np.cos(uu)
yy = r*np.sin(vv)*np.sin(uu)
zz = r*np.cos(vv)
```

```python
#网格可视化
fig = plt.figure(figsize = (6,6))
ax = fig.add_subplot(111,projection = '3d')

#绘制球面网格线
ax.plot_wireframe(xx,yy,zz,color = 'blue',linewidth = 0.5)

ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.set_zlim(-1,1)

ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.set_aspect('equal','box')
ax.set_proj_type('ortho')
```

```python
#绘制箭头图
fig = plt.figure(figsize = (6,6))
ax = fig.add_subplot(111,projection = '3d')

#归一化
u_norm = uu.ravel()/(2*np.pi) #为了与hsv colormap 对应

#获取 hsv colormap 对应的RGB颜色
colors = hsv(u_norm)

zeros = np.zeros_like(xx.ravel())

#绘制箭头图
ax.quiver(zeros,zeros,zeros,
          xx.ravel(),yy.ravel(),zz.ravel(),
          color = colors,length = 1,
          #这里第一个参数:传入向量全部缩放成单位向量然后再乘以指定长度,这种设计只反应方向,因为长度都因为length一致
          #第二个参数:箭头长度与箭杆长度比例 0表示没有箭头
          normalize =True,arrow_length_ratio = 0)

ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.set_zlim(-1,1)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
#保证三个方向等比例,box是显示在正方体里
ax.set_aspect('equal','box')
#正交投影,没有透视缩放
ax.set_proj_type('ortho')
```
