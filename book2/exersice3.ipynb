```python
def identify_axes(ax_dict,fontsize = 48):
    kw = dict(ha = 'center',va = 'center',fontsize = fontsize,color = 'darkgrey')
    for k,ax in ax_dict.items():
        #**kw是将字典中的参数一次性传入这个函数中
        #transform表示告诉轴的绘图这个绘图在哪里绘制 所以ax.transAxes表示提取出来的绘制的坐标系
        #k表示字典的key.定位已经由ax完成,k只是图中的标注,用位置参数,参数是绘制的字体就是传入的k本身,用作显示
        #transform表示坐标体系,这里传入的是归一化的坐标体系,所以0.5就是中间的意思
            #每一个轴还带了一个内部的归一化坐标系,开头就是Axes,所以这里表示归一化坐标系
            #transform表示选择并应用一种坐标系的转换方式
        ax.text(0.5,0.5,k,transform = ax.transAxes,**kw)
```

```python
np.random.seed(19680801)
hist_data = np.random.randn(1_500)

fig = plt.figure(layout = 'constrained')#自动压缩子图间距,但是布局不重叠
ax_array = fig.subplots(2,2,squeeze = False)#返回的是二维数组

ax_array[0,0].bar(['a','b','c'],[5,7,9])
ax_array[0,1].plot([1,2,3])#如果有一个数组,自动当做y值,x就是对应的点的索引(从0开始)
ax_array[1,0].hist(hist_data,bins = 'auto')
ax_array[1,1].imshow([[1,2],[2,1]])#颜色映射是cmap默认值决定

identify_axes(
    #先遍历行,后遍历列,ax_array是2D的
    {(j,k):a for j,r in enumerate(ax_array) for k,a in enumerate(r)},
)
```

```python
#从scipy的统计模块中导入正态分布对象
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

import os

if not os.path.isdir('Figures'):
    os.makedir('Figures')
```

```python
x_array = np.linspace(-5,5,200)
mu_array = [-2,-1,0,1,2]
sigma_array = [0.5,0.75,1,1.5,2]
mu_array_,sigma_array_ = np.meshgrid(mu_array,sigma_array)
```

```python
fig,axs = plt.subplots(nrows = 5,ncols = 5,figsize = (10,10))

for mu_idx,sigma_idx,ax in zip(mu_array_.ravel(),sigma_array_.ravel(),axs.ravel()):#按照索引抽取每一列表元素组成元组返回

    title_idx = 'mu = ' + str(mu_idx) + '; sigma = ' + str(sigma_idx)
    ax.plot(x_array,norm.pdf(x_array,mu_idx,sigma_idx),color = '#0070C0')#pdf是改了密度函数,生成正态分布

    #x表示x轴 然后让y1,y2之间填充,这里y2 = 0所以是y1 - 0之间填充
    ax.fill_between(x_array,norm.pdf(x_array,mu_idx,sigma_idx),0,color = '#DBEEF3')

    ax.set_yticklabels([])
    ax.set_yticks([])
    #整个y轴不可见,刻度,刻度文本,脊等.轴是更大的,包括脊,脊仅仅表示四个边
    ax.get_yaxis().set_visible(False)
    #设置y轴标签为空 等价ax.set_ylabel(None)  set是一个统一管理多属性的接口函数,可以一次性统一配置多个
    ax.set(ylabel = None)

    ax.set_title(title_idx)
    ax.set_xlim(-5,5)
    ax.set_ylim(0,1)
    ax.set_xticks([-5,0,5])
    #将脊配置为不可见,但存在 ax.spines['left'].set_visible(False)则是完全禁用,没有脊了
    ax.spines['left'].set_color('none')

    ax.spines.right.set_visible(False)
    ax.spines.top.set_visible(False)
    #将x刻度放在底部脊上,这个函数可以配置上下作业,但是针对x只能是上和底 y才能是左和右
    ax.xaxis.set_ticks_position('bottom')
    #只配置x轴,刻度线向内画
    ax.tick_params(axis = 'x',direction = 'in')
    
    ax.grid(False)

plt.tight_layout()#自动调整,让元素不重叠(子图间隙,标签位置等)
fig.savefig('Figures/高斯分布随mu,sigma变化.svg',format = 'svg')
```

```python
def plot_joint_marginal(xx1,xx2,surface,
                        x1,f_x1,
                        x2,f_x2,
                        title_txt,color = '#0070C0'):#函数名最后的单词是边缘的意思,表示边缘分布

    fig = plt.figure(figsize = (6,6))
    #创建的这一刻自动和当前fig绑定
    gs = gridspec.GridSpec(2,2,width_ratios = [3,1],height_ratios = [3,1])

    #主图
    #取一个单独的格子
    ax1 = plt.subplot(gs[0])

    ax1.contour(xx1,xx2,surface,20,cmap = 'RdYlBu_r')

    #将y轴的标签放在右边的脊上
    #注意是标签,标签是tick的标签,不是图片的title
    ax1.yaxis.set_label_position('right')
    ax1.set_xticks(np.arange(0,1.2,0.2))
    ax1.set_yticks(np.arange(0,1.2,0.2))
    ax1.set_title(title_txt)

    #Y小图
    ax2 = plt.subplot(gs[1])
    ax2.plot(f_x2,x2,color = color)
    #两个参数的话默认第三个参数为0,填充x2 - 0之间,填充的图形是闭合多边形(自交区相互消去)
    ax2.fill_between(f_x2,x2,edgecolor = 'none',facecolor = color,alpha = 0.2)

    ax2.set_xlabel('PDF')
    ax2.set_ylim(0,1)
    ax2.set_xlim(0,5)
    ax2.set_xticks(np.arange(6))
    ax2.set_yticks(np.arange(0,1.2,0.2))
    ax2.invert_xaxis()
    ax2.yaxis.tick_right()

    #X小图
    ax3 = plt.subplot(gs[2])
    ax3.plot(x1,f_x1,color = color)
    ax3.fill_between(x1,f_x1,edgecolor = 'none',facecolor = color,alpha = 0.2)
    ax3.set_ylabel('PDF')
    ax3.yaxis.set_label_position('left')
    ax3.set_xlim(0,1)
    ax3.set_ylim(0,5)
    ax3.set_yticks(np.arange(6))

    ax4 = plt.subplot(gs[3])
    ax4.set_visible(False)

    fig.savefig('Figures/' + title_txt + '.svg',format = 'svg')

    plt.show()
```

```python
x1 = np.linspace(0,1,201)
x2 = np.linspace(0,1,201)

xx1,xx2 = np.meshgrid(x1,x2)

xx3 = 1.0 - xx1 - xx2
#通过第一个参数检查每一个值,如果对填充第二个参数中的对应参数,如果不对执行第三参数中的对应参数
#np.nan会自动广播成相应的形状
xx3 = np.where(xx3 > 0.0,xx3,np.nan)
```

```python
alpha_1 = 2
alpha_2 = 2
alpha_3 = 2

alpha_0 = alpha_1 + alpha_2 + alpha_3

alphas = np.array([alpha_1,alpha_2,alpha_3])
rv = stats.dirichlet(alphas)

PDF_ff = rv.pdf(np.array(([xx1.ravel(),xx2.ravel(),xx3.ravel()])))

PDF_ff = np.reshape(PDF_ff,xx1.shape)

beta_dist = stats.beta

theta_array = np.linspace(0,1,500)

beta_1 = alpha_0 - alpha_1
f_x1 = beta_dist.pdf(
    x1,alpha_1,beta_1)
beta_2 = alpha_0 - alpha_2
f_x2 = beta_dist.pdf(x2,alpha_2,beta_2)

#%是旧的字符串格式化,0表示最小宽度, .0表示保留几位小数 后面的%(括号外的)意思是将后面的元组的值按照顺序填写到前面字符串中
title_txt = 'Dirichlet(%0.0f,%0.0f,%0.0f)'%(alpha_1,alpha_2,alpha_3)
plot_joint_marginal(xx1,xx2,PDF_ff,
                    x1,f_x1,
                    x2,f_x2,
                    title_txt)
```

```python
def squiggle_xy(a,b,c,d,i = np.arange(0.0,2*np.pi,0.05)):
    return np.sin(i*a)*np.cos(i*b),np.sin(i*c)*np.cos(i*d)
```

```python
#不启用自动布局算法,因为需要手动用spec控制布局
fig = plt.figure(figsize = (8,8),constrained_layout = False)
outer_grid = fig.add_gridspec(4,4,wspace = 0,hspace = 0)

for a in range(4):
    for b in range(4):
        #只能在fig下的gridspec继续切而不会在fig上面
        inner_grid = outer_grid[a,b].subgridspec(3,3,wspace = 0,hspace = 0)
        #然后根据里面的分配好的spec继续创建,创建语法不同于fig
        
        axs = inner_grid.subplots()
        #对多维数组进行遍历,返回索引元组(位置)和ax本身
        for (c,d),ax in np.ndenumerate(axs):
            ax.plot(*squiggle_xy(a+1,b+1,c+1,d+1))
            ax.set(xticks = [],yticks = [])

for ax in fig.get_axes():
    #ax检测自己所在那一层的spec,不是fig的spec
    ss = ax.get_subplotspec()
    #检测是否在第一列,如果在第一行返回True,表示绘制第一行的边框
    ax.spines.top.set_visible(ss.is_first_row())
    ax.spines.bottom.set_visible(ss.is_last_row())
    ax.spines.left.set_visible(ss.is_first_col())
    ax.spines.right.set_visible(ss.is_last_col())

fig.savefig('Figures/subgridspec嵌套子图.svg',format = 'svg')
plt.show()
```
