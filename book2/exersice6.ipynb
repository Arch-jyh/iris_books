```python
values = np.linspace(0,1,11)
#返回colormap对象
cmap = plt.get_cmap('RdYlBu_r')
#对返回的对应直接使用(),表示这是可调用对象,相当于将对象看成了函数
    #原因是对象是个类,实现了__call__,这个可以将对象看成函数直接调用,执行__call__下的函数
#这个函数相当于用value当做归一化的位置进行一次色彩映射,返回映射对应位置的颜色,返回值类型是array,每一列对应了RGBA
    #用RGB是因为这是针对硬件的,需要RGB
colors = plt.get_cmap('RdYlBu_r')(values)
colors
```

```python
#四舍五入到1位小数,第一个参数为输入,第二个参数为保留小数位数
#reshape中的-1表示自行判断行数
data = np.round(np.linspace(0,1,11).reshape(-1,1),1)

fig,ax = plt.subplots(figsize = (5,5))
#传入的data是(11,1)的大小,将按照二维数据大小自动产生y为11个格子,x为1个格子
    #格子在数据轴上的长度也为对应的大小为1,每个格子填充一个数据颜色
    #显示时候会调整x,使其不会和y的轴的长度比例一致,让底轴范围为0-1以充满整个绘图
sns.heatmap(data,
            xticklabels = False,
            yticklabels = False,
            cmap = colormap,
            annot = True,#显示每个格子中的对应的data数值,自动放置到显示出来的的格子的中心
            cbar = False,
            #fmt是annot的数字配置格式(这是表示不格式化)(不是字体,类似保留小数这种)显示格式为空将不进行格式转化直接显示原始字符串
            fmt = '',ax = ax)

for i,value in enumerate(data):
    #因为data是2d,取出的value是一行,哪怕只有一个元素,也需要从1d中取出来
    rgba_color = colormap(value[0])
    #tuple生成元组
    rgb_color = tuple(int(255 * c) for c in rgba_color[:3])
    ax.text(0,i+0.5,#这是数据坐标位置,锚点将会找这个坐标,不属于heatmap的内容,所以文字仅仅是叠加对象,锚点仅仅用于定位
            f'RGB: {rgb_color}',
            color = 'black',
            #锚点对齐方式,锚点在左侧,然后是左侧的中心
            ha = 'left',
            va = 'center')

plt.savefig('Figures/color mapping,1.svg')
plt.show()
```

```python
colormap = plt.get_cmap('RdYlBu_r')

data = np.round(np.linspace(0,1,21).reshape(-1,1),2)

fig,ax = plt.subplots(figsize = (5,5))
sns.heatmap(data,
            xticklabels = False,
            yticklabels = False,
            cmap = colormap,
            annot = True,
            cbar = False,
            fmt = '',ax = ax)

for i,value in enumerate(data):
    rgba_color = colormap(value[0])
    #c是一个数组,然后外部用tuple转换为元组
    rgb_color = tuple(int(255 * c) for c in rgba_color[:3])#
    ax.text(0,i+0.5,
            #将变量的值替换为字符串进行表示
            f'RGB: {rgb_color}',
            color = 'black',
            ha = 'left',va = 'center')
plt.savefig('Figures/color mapping,2.svg')
plt.show()
```
