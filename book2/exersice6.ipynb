```python
#四舍五入到1位小数,第一个参数为输入,第二个参数为保留小数位数
#reshape中的-1表示自行判断行数
data = np.round(np.linspace(0,1,11).reshape(-1,1),1)

fig,ax = plt.subplots(figsize = (5,5))
#传入的data是(11,1)的大小,将按照二维数据大小自动产生y为11个格子,x为1个格子
    #格子在数据轴上的长度也为对应的大小为1,每个格子填充一个数据颜色
    #显示时候会调整x,使其不会和y的轴的长度比例一致,让底轴范围为0-1以充满整个绘图
sns.heatmap(data,
            xticklabels = False,
            yticklabels = False,
            cmap = colormap,
            annot = True,#显示每个格子中的对应的data数值,自动放置到显示出来的的格子的中心
            cbar = False,
            #fmt是annot的数字配置格式(这是表示不格式化)(不是字体,类似保留小数这种)显示格式为空将不进行格式转化直接显示原始字符串
            fmt = '',ax = ax)

for i,value in enumerate(data):
    #因为data是2d,取出的value是一行,哪怕只有一个元素,也需要从1d中取出来
    rgba_color = colormap(value[0])
    #tuple生成元组
    rgb_color = tuple(int(255 * c) for c in rgba_color[:3])
    ax.text(0,i+0.5,#这是数据坐标位置,锚点将会找这个坐标,不属于heatmap的内容,所以文字仅仅是叠加对象,锚点仅仅用于定位
            f'RGB: {rgb_color}',
            color = 'black',
            #锚点对齐方式,锚点在左侧,然后是左侧的中心
            ha = 'left',
            va = 'center')

plt.savefig('Figures/color mapping,1.svg')
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
#lambdify是将符号转换为python函数 latex是将表达式转换为latex公式字符串
#必须接收sympy表达式,表达式中混有python运算也没事 但是不能纯粹的python运算(本来就能运行)
from sympy import lambdify,diff,exp,latex
#从符号库abc中导入x和y符号变量
from sympy.abc import x,y#x,y已经是符号变量了,如果用其进行变量赋值,符号变量会失效(x,y还是变量名,这时候已经绑定符号了)
#colormap模块,新街口名称为colormaps(更规范)
from matplotlib import cm
```

```python
num = 301
x_array = np.linspace(-3,3,num)
y_array = np.linspace(-3,3,num)
xx,yy = np.meshgrid(x_array,y_array)

f_xy = 3*(1-x)**2*exp(-(x**2) - (y+1)**2)\
    - 10*(x/5 - x**3 - y**5)*exp(-x**2 - y**2)\
    - 1/3*exp(-(x+1)**2 - y**2)

f_xy_fcn = lambdify([x,y],f_xy)
ff = f_xy_fcn(xx,yy)
```

```python
f_xy#只要表达式混有符号变量,就会变味符号表达式
```

```python
#用新创建的字典并传入
fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})
ax.set_proj_type('ortho')

surf = ax.plot_surface(xx,yy,ff,cmap = cm.RdYlBu,
                       linewidth = 0,antialiased = False)#最后是是否开启抗锯齿

ax.set_xlabel(r'$\it{x_1}$')
ax.set_ylabel(r'$\it{x_2}$')
#it是斜体命令,这是一个latex命令不是字符串的所以前面加r停止python字符串调用\,让latex工作
#$$下也有latex的指令也是通过\调用,python字符串下的指令也是通过\调用.调用前者需要加r停止python调用后进行工作
ax.set_zlabel(r'$\it{f}$($\it{x_1}$,4\it{x_2}$)')
ax.set_xlim(x_array.min(),x_array.max())
ax.set_ylim(y_array.min(),y_array.max())

ax.view_init(azim = -135,elev = 30)

ax.grid(False)

plt.rcParams['font.family'] = 'Times New Roman'#设置全局字体家族为一个字体养蛇,这是论文常用英文字体
plt.rcParams['font.size'] = '10'

#传入图像对象,会读取颜色映射和数值范围.表示高度收缩为50%,aspect表示长宽比(高宽比)
#colorbar会读取颜色信息,数据范围,colormap,norm归一化器
fig.colorbar(surf,shrink = 0.5,aspect = 20)
fig.savefig('Figures/用plot_surface()绘制二元函数曲面.svg',format = 'svg')
```

```python
cmap_arrays = ['RdYlBu_r','Blues_r','rainbow','viridis']

#21个点就是20个区间
#21个点对应的是等高线,20个区间对应的等高线之间的色带
levels = np.linspace(-10,10,21)

#循环绘制四幅图
for cmap_idx in cmap_arrays:
    fig,ax = plt.subplots()
    #极大值和极小值是孤立点,通常会落入上一个或者下一个区间中,不单独表示一层
    #填满指的是填满所有等高线之间的区间
    colorbar = ax.contourf(xx,yy,ff,levels = levels,cmap = cmap_idx)

    #cbar是fig的一部分不是ax的,所以需要指定哪个ax指定绘制在哪里
    cbar = fig.colorbar(colorbar,ax = ax)
    cbar.set_ticks([-10,-5,0,5,10])
    cbar.ax.set_title(r'$\it{f}$($\it{x_1}$,$\it{x_2}$)',fontsize = 8)

    ax.set_xlim(xx.min(),xx.max())
    ax.set_ylim(yy.min(),yy.max())

    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')

    plt.gca().set_aspect('equal',adjustable = 'box')

    title = 'Colormap = ' + str(cmap_idx)
    plt.title(title)

    fig.savefig('Figures/平面填充等高线_' + str(cmap_idx) + '.svg',format = 'svg')
```

```python
#定义函数,输入为色谱名称,函数可视化色谱所在色彩空间的位置(RGB)

def visualize_cm_in_RGB(cm_name_str):
    cm_name = mpl.colormaps[cm_name_str]
    color_codes = cm_name(np.linspace(0,1,200))

    fig = plt.figure(figsize = (6,6))

    azim_array = [38,0,-90,90]
    elev_array = [34,0,0,90]

    for idx,angles in enumerate(zip(azim_array,elev_array)):
        ax = fig.add_subplot(2,2,idx+1,projection = '3d')

        ax.scatter(color_codes[:,0],color_codes[:,1],color_codes[:,2],
                    c = color_codes,s = 4,alpha = 1)

        ax.set_xlim(0,1)
        ax.set_ylim(0,1)
        ax.set_zlim(0,1)
        ax.set_xticks([0,1])
        ax.set_yticks([0,1])
        ax.set_zticks([0,1])
        ax.set_xlabel('R')
        ax.set_ylabel('G')
        ax.set_zlabel('B')

        ax.xaxis.line.set_color((1.0,1.0,1.0,0.0))
        ax.yaxis.line.set_color((1.0,1.0,1.0,0.0))
        ax.zaxis.line.set_color((1.0,1.0,1.0,0.0))
    
        ax.xaxis.set_pane_color((1.0,1.0,1.0,0.0))
        ax.yaxis.set_pane_color((1.0,1.0,1.0,0.0))
        ax.zaxis.set_pane_color((1.0,1.0,1.0,0.0))

        plt.rcParams['grid.color'] = 'k'
        ax.set_proj_type('ortho')
        ax.set_box_aspect(aspect = (1,1,1))
        ax.view_init(azim = angles[0],elev = angles[1])

    fig.savefig('Figures/' + cm_name_str + '_色谱色号位置.svg',format = 'svg')
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.colors import rgb_to_hsv
```

```python
def visualize_cm_in_HSV(cm_name):
    fig = plt.figure(figsize = (6,6))
    azim_array = [38,0,-90,-90]
    elev_array = [34,0,0,90]

    x = np.linspace(0.0,1.0,500)

    #加一个维度,切片正常
    #保持了批次维度的习惯
    rgb = mpl.colormaps[cm_name](x)[np.newaxis,:,:3]

    #圆柱的点
    theta_array = np.linspace(0,2*np.pi,100)
    xline = np.sin(theta_array)
    yline = np.cos(theta_array)

    #可以处理不是2d的rgb但是最后一个维度必须是3,不能有透明度(这里是1,500,3)的形状
    #输出也会保持形状输出,还是(1,500,3)的形状
    HSV = rgb_to_hsv(rgb)
    #将最外层取出0索引,形状变回(500,3)
    HSV = HSV[0]
    HSV[:,0] = HSV[:,0] * 2 * np.pi
    HSV_xyz = np.copy(HSV)
    HSV_xyz[:,0] = HSV[:,1] * np.sin(HSV[:,0])
    HSV_xyz[:,1] = HSV[:,1] * np.cos(HSV[:,0])

    for idx,angles in enumerate(zip(azim_array,elev_array)):
        ax = fig.add_subplot(2,2,idx+1,projection = '3d')

        ax.plot(xline,yline,theta_array*0+0,'k')
        ax.plot(xline,yline,theta_array*0+1,'k')#
        ax.plot((0,0),(0,0),(0,1),color = 'k')

        ax.scatter(HSV_xyz[:,0],HSV_xyz[:,1],HSV_xyz[:,2],
                    c = rgb[0],s = 10,alpha = 1)

        ax.set_xlim(-1,1)
        ax.set_ylim(-1,1)
        ax.set_zlim(0,1)
        ax.set_xticks([-1,1])
        ax.set_yticks([-1,1])
        ax.set_zticks([0,1])
        ax.set_zlabel('V')

        plt.rcParams['grid.color'] = 'k'
        ax.set_proj_type('ortho')
        ax.set_box_aspect(aspect = (1,1,1))
        ax.view_init(azim = angles[0],elev = angles[1])

    fig.savefig('Figures/HSV_' + cm_name + '.svg',format = 'svg')
```

```python
cm_list = ['RdYlBu','viridis','Blues','cool','rainbow','jet','turbo','hsv']
```

```python
for idx,cm_idx in enumerate(cm_list):
    visualize_cm_in_HSV(cm_idx)
```

```python
cmaps = {}

#坡度引申为渐变,所以表示显示渐变条
gradient = np.linspace(0,1,256)
#渐变条像素为2*256
gradient = np.vstack((gradient,gradient))

#category是cmap的一些分类,一个category统一了一些类似的cm
def plot_color_gradients(category,cmap_list):
    nrows = len(cmap_list)
    figh = 0.35 + 0.15 + (nrows + (nrows - 1) * 0.1) * 0.22#图像高度
    fig,axs = plt.subplots(nrows = nrows + 1,figsize = (6.4,figh))#加一行作为下边距
    #调整图集参数,子图们只能在这个范围内
    #这里的坐标都是表示比例,且都是从上到下数的比例和从左到右数的比例,所以left和right读取的是一个方向的比例值
        #是按照
    fig.subplots_adjust(top = 1 - 0.35 / figh,bottom = 0.15/figh,left = 0.2,right = 0.99)
    axs[0].set_title(f'{category} colormaps',fontsize = 14)#用title当全部的标题,后面还会用这个子图显示cm

    #多出来的axs当成下边距离
    for ax,name in zip(axs,cmap_list):
        #使用imshow进行彩色映射  'auto'表示拉伸填满子图
        #imshow函数是为了显示像素图.可以进行cm颜色映射   如果图片本身有彩色信息,将会按照图片本身的彩色信息展示
            #仅仅分析矩阵外部两层当做图片格式,内部是色彩信息,复杂的直接用复杂信息色彩,简单的可以后面指定cm(本次的情况)
        ax.imshow(gradient,aspect = 'auto',cmap = mpl.colormaps[name])
        #text是独立的图层,虽然ax.text还是在ax下进行控制,但是设计之初留给axis一定的空间可以超过ax显示
            #fig.text才是另一个不在ax控制下显示text,是另一个空见了
        ax.text(-0.01,0.5,name,va = 'center',ha = 'right',fontsize = 10,
                #在[0,1]的轴坐标系中,转换到了归一化坐标轴
                transform = ax.transAxes)

    for ax in axs:
        ax.set_axis_off()

    cmaps[category] = cmap_list
```

```python
#一个子图colormap显示的条数(子图中的色彩数量)
#_表示内部变量,表示给自己用(当前模块,当前代码),不是外部用的
_DSUBS = {
    'Perceptually Uniform Sequential' : 5,
    'Sequential':6,
    'Sequential (2)':6,
    'Diverging':6,
    'Cyclic':3,
    'Qualitative':4,
    'Miscellaneous':6
}
#每个子图内colormap的水平间距
_DC = {
    'Perceptually Uniform Sequential':1.4,
    'Sequential':0.7,
    'Sequential (2)':1.4,
    'Diverging':1.4,
    'Cyclic':1.4,
    'Qualitative':1.4,
    'Miscellaneous':1.4
}

#.0明确这是浮点数,实际带不带都可以,保证不会出现整数问题
x = np.linspace(0.0,1.0,100)

for cmap_category,cmap_list in cmaps.items():#拆成键值一对数据分别传入
    #get(key,default)表示key存在返回key对应的值,不存在返回默认值6
        #表示每个子图中的放置数量,cmap对应的条数
    dsub = _DSUBS.get(cmap_category,6)
    #ceil无条件向上取整,目的是计算多少个subplot
        #因为一个category可能需要多个subplot,一个sub最多6个显示不全计算实际sub数量
    nsubplots = int(np.ceil(len(cmap_list)/dsub))

    #squeeze = False时钟返回二维数组结构
        #1*N或者M*1形状都会从2d变味1d,使用squeeze=False是为了保持2d形状,这是针对axs的.
            #fig的网格格式依旧是按照subplots的格式来,没事
        #永远为2d是为了简化判断ax维度(在不同绘图主题中一些ax数量可能为单数变成一个对象的问题(维度都没了))
    fig,axs = plt.subplots(nrows = nsubplots,squeeze = False,
                            figsize = (7,2.6*nsubplots))

    #.flat将二维axs展成1d方便遍历,.flat是一个1d迭代器,调用后根据外部循环进行迭代所有内部值
    #迭代器特点:不一次性返回全部元素,每次返回一个值,会记住位置
        #本就是一种魔术方法会自己调用
    for i,ax in enumerate(axs.flat):
        locs = []
        #遍历每个子图应该显示的多个cm
        #in后面属于切片操作,结果仍然是1d的列表,切片永远返回列表
        for j,cmap in enumerate(cmap_list[i*dsub:(i+1)*dsub]):
            #已经用x进行映射了,取前三列且多加一个维度
            rgb = mpl.colormaps[cmap](x)[np.newaxis,:,:3]
            #第一个()中表示用A到B的转换函数创建一个转换器,然后转换rgb函数,最后得到新的色彩空间
            lab = cspace_converter('sRGB1','CAM02-UCS')(rgb)

            #这个主题下的所有cm单独反转处理,因为这个图有自己的特点,亮度和其他图是反着的
            if cmap_category == 'Sequential':
                #y_是亮度曲线
                #开始:结束:步长  -1中-表示反向1表示步长 这里就是反向步长  切片中的0被舍弃没有取值,出来的数据自动是1d的
                #第二个轴是采样点的倒序
                #色彩格式第二个轴是亮度
                y_ = lab[0,::-1,0]
                #c_是颜色值,就是x的映射数据相应也取反,是需要映射的值,只有一个维度
                c_ = x[::-1]
            else:
                y_ = lab[0,:,0]#正常取
                c_ = x#正常取

            #获取该类别下每一个cm的距离,如果没有默认为1.4
            dc = _DC.get(cmap_category,1.4)
            #绘制散点  c是映射值(颜色)cmap是映射的目标map
            ax.scatter(x+j*dc,y_,c = c_,cmap = cmap,s = 300,linewidths = 0.0)

            if cmap_category in ('Perceptually Uniform Sequential','Sequential'):
                #这两个类别下,取x的最后一个元素
                #Sequential这类别开头亮度高,反转了显示是从底到高,所以最后的位置是最高的位置
                    #另外一个也是这种情况
                locs.append(x[-1] + j*dc)
            elif cmap_category in ('Diverging','Qualitative','Cyclic',
                                   'Miscellaneous','Sequential (2)'):
                #这几个类别下取中间元素
                #int直接截断小数部分取整  np.ceil是向上取整如果有小数的话(一直为正方向,-2.3>>>-2)
                #这组中中间高,两边低
                locs.append(x[int(x.size/2.)] + j*dc)

        #配置范围,x方向统一为第一个.
        ax.set_xlim(axs[0,0].get_xlim())
        #这个色彩空间没归一化,亮度就是0-100
        ax.set_ylim(0.0,100.0)

        #将x刻度移动到顶部
        ax.xaxis.set_ticks_position('top')
        #创建固定刻度位置器,仅仅在当前ax下存储了刻度位置信息
        #因为locs能用于其他地方可能混淆,但是ticker定位后只能用于刻度.为了工程管理
        ticker = mpl.ticker.FixedLocator(locs)
        #在循环中的哪个ax下就配置那个ax的刻度
        ax.xaxis.set_major_locator(ticker)
        #创建刻度标签文字内容
        #刻度怎么显示文字由formatter决定
            #传入字符串当做刻度标签进行处理(功能之一,所以叫fiexedformatter,传入字符串作为标签仅仅是常用功能)
                #其他功能是有多语言输出,数学公式输出等等
            #最后都是显示为刻度标签,但是有一些预处理功能
        formatter = mpl.ticker.FixedFormatter(cmap_list[i*dsub:(i+1)*dsub])
        #配置相应主刻度标签?
        ax.xaxis.set_major_formatter(formatter)
        #标签旋转50度
        ax.xaxis.set_tick_params(rotation = 50)
        #L^*是科学上亮度通道的标准符号 *表示star  L star-科学名称
        ax.set_ylabel('Lightness $L^*$',fontsize = 12)

    #cmap_category本就是字符串,无需处理
    ax.set_xlabel(cmap_category + 'colormaps' ,fontsize = 14)

    #h_pad表示垂直之间的而外距离补偿,pad表示图外侧的总边距
    fig.tight_layout(h_pad = 0.0,pad = 1.5)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.random.randn(20,20)
```

```python
from matplotlib.colors import LinearSegmentedColormap

#默认的话就是按照颜色数量和索引位置均匀分布
continuous_cmap_even = LinearSegmentedColormap.from_list('',['darkblue','skyblue','white','pink','magenta'])
continuous_cmap_even
```

```python
fig,ax = plt.subplots()
HM = sns.heatmap(data,ax = ax,
                 vmin = -3,vmax = 3,#配置颜色对应值的显示范围,会自动归一化并映射到cmap上拿到颜色
                 cmap = continuous_cmap_even,
                 xticklabels = False,#控制刻度文字隐藏不显示
                 yticklabels = False,
                 linecolor = 'grey',
                 linewidths = 0.1)

#控制刻度线隐藏不显示,数据还是在的修改参数就能恢复.  像是ax.set_yticks([])是直接清空,需要重新加载数据
HM.tick_params(left = False,bottom = False)

fig.savefig('Figures/热图_连续色谱.svg',format = 'svg')
```

```python
list_nodes = [0.0,0.1,0.5,0.9,1.0]
list_colors = ['darkblue','skyblue','white','pink','magenta']

continuous_cmap_uneven = LinearSegmentedColormap.from_list('',list(zip(list_nodes,list_colors)))
continuous_cmap_uneven
```

```python
fig,ax = plt.subplots()
HM = sns.heatmap(data,ax = ax,
                 vmin = -3,vmax = 3,
                 cmap = continuous_cmap_uneven,
                 xticklabels = False,
                 yticklabels = False,
                 linecolor = 'grey',
                 linewidths = 0.1)

HM.tick_params(left = False,bottom = False)

fig.savefig('Figures/热图_连续色谱_不均匀.svg',format = 'svg')
```
