```python
#生成圆弧的函数
#半径,起始角,弧长角度
def arc(r,angle_start,angle_arc):
    delta_radian = np.pi/720#小步长保证平滑度
    #配置角度范围,每次走delta_radian
    angle_array = np.arange(angle_start,angle_start+angle_arc,delta_radian)

    #生成圆弧坐标点
    x_array = r*np.cos(angle_array)
    y_array = r*np.sin(angle_array)

    return x_array,y_array
```

```python
cmap = mpl.colormaps.get_cmap('RdYlBu')#新的api,
num = 200

#可以生成指定范围的均匀随机分布
#np.random.rand只能生成0-1范围内的均匀随机分布,左开右闭
r_array = np.random.uniform(0,1,num)
angle_1_array = np.random.uniform(0,2*np.pi,num)
#旋转了多少角度,不是终点
angle_2_array = np.random.uniform(0,2*np.pi,num)

fig,ax = plt.subplots()

for r_idx,a_1_idx,a_2_idx in zip(r_array,angle_1_array,angle_2_array):
    x_array,y_array = arc(r_idx,a_1_idx,a_2_idx)
    #color形状如果是和点数一样,归一化映射
    #一个值,一个字符串直接配置颜色
        #或者是一个颜色的rgb格式
    plt.plot(x_array,y_array,color = cmap(r_idx),lw = 0.2)

ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.set_xticks([])
ax.set_yticks([])

ax.axis('off')
ax.set_aspect('equal')
fig.savefig('Figures/随机弧.svg')
```
