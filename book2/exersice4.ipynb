```python
x_array = np.linspace(-4,4,200)
y_array = x_array*np.exp(-x_array**2)
```

```python
fig = plt.figure(figsize = (6,6),tight_layout = True)

ax = fig.add_subplot(3,2,1)
ax.plot(x_array,y_array)
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)

ax = fig.add_subplot(3,2,2)
ax.plot(x_array,y_array)
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
#图脊仍然存在,但是透明
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

ax = fig.add_subplot(3,2,3)
ax.plot(x_array,y_array)
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
#图脊完全关闭,不绘制
#使用[[]]是因为虽然允许操作多个脊,但是字典只能接受一个对象
ax.spines[['bottom','left']].set_visible(False)
ax.yaxis.set_ticks_position('right')
ax.xaxis.set_ticks_position('top')

ax = fig.add_subplot(3,2,4)
ax.plot(x_array,y_array)
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
#将bottom脊的位置挪到y = 0的位置,data说明用的是数据坐标系
#用两个括号表示是因为传入必须为元组()  第一个表示类型,第二个表示值,data表示数据坐标系
ax.spines['bottom'].set_position(('data',0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax = fig.add_subplot(3,2,5)
ax.plot(x_array,y_array)
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
ax.spines['bottom'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax = fig.add_subplot(3,2,6)
ax.plot(x_array,y_array)
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
ax.spines[['bottom','left']].set_position(('data',0))
ax.spines[['right','top']].set_color('none')

fig.savefig('Figures/修改图脊,第一组.svg',format = 'svg')
```

```python
import numpy as np
import matplotlib.pyplot as plt

import os
if not os.path.isdir('Figures'):
    os.makedirs('Figures')
```

```python
p = plt.rcParams
#p['font.sans-serif'] = ['Roboto']
p['font.weight'] = 'light'
p['axes.grid'] = True
p['grid.color'] = '0.5'
p['grid.linewidth'] = 0.5
```

```python
x_array = np.linspace(-4,4,200)
y_array = x_array*np.exp(-x_array**2)
```

```python
fig = plt.figure(figsize = (6,6),tight_layout = True)

ax = fig.add_subplot(3,2,1)
ax.plot(x_array,y_array)
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
ax.grid(True)

ax = fig.add_subplot(3,2,2)
ax.plot(x_array,y_array)
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
ax.grid(linestyle = '-',
        linewidth = '0.5',color = 'r')

ax = fig.add_subplot(3,2,3)
ax.plot(x_array,y_array)
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
#网格画在数据线下方
    #因为网格是axis的元素,所以用axis
    #原意是网格和刻度线都画在下面
ax.set_axisbelow(True)
#打开次刻度,直接就是打开函数,无需参数
ax.minorticks_on()
ax.grid(which = 'major',linestyle = '-',
        linewidth = '0.5',color = 'red')#如果使用了which参数,那么默认这个grid显示,无需grid(True)
ax.grid(which = 'minor',linestyle = ':',
        linewidth = 0.5,color = 'k')#不加which默认是major

ax = fig.add_subplot(3,2,4)
ax.plot(x_array,y_array)
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
ax.set_axisbelow(True)
#次刻度需要手动开启,因为默认是关闭的
ax.minorticks_on()
ax.grid(which = 'major',linestyle = '-',
        linewidth = '0.5',color = 'red')
ax.grid(which = 'minor',linestyle = ':',
        linewidth = '0.5',color = 'k')
#只关闭刻度线,不关闭刻度标签和次刻度
ax.tick_params(which = 'both',
               top = False,
               left = False,
               right = False,
               bottom = False)

ax = fig.add_subplot(3,2,5)
ax.plot(x_array,y_array)
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
ax.grid(True)
#如果执行ax.minorticks_on()会影响后续的子图的配置,也会默认添加次刻度(老版本不会,只影响当前ax)
ax.minorticks_off()
for tick in ax.xaxis.get_major_ticks():
    #关闭内侧刻度tick1和tick2都只是关闭主刻度,是内外侧主刻度
        #只有使用tick_params才能用which配置次刻度,当前语法只能主刻度
    tick.tick1line.set_visible(False)#关闭外侧主刻度
    tick.tick2line.set_visible(False)#关闭内侧主刻度
    tick.label1.set_visible(False)#隐藏主标签,标签不分内外
    tick.label2.set_visible(False)#隐藏次标签

ax = fig.add_subplot(3,2,6)
ax.plot(x_array,y_array)
ax.set_xlim(-4,4)
ax.set_ylim(-0.5,0.5)
ax.grid(True)
ax.minorticks_off()
for tick in ax.yaxis.get_major_ticks():
    tick.tick1line.set_visible(False)
    tick.tick2line.set_visible(False)
    tick.label1.set_visible(False)
    tick.label2.set_visible(False)
ax.set_facecolor('#DBEEF8')
plt.show()

fig.savefig('Figures/背景网格.svg',format = 'svg')
```

```python
import numpy as np
import matplotlib.pyplot as plt
#导入ticker(针对刻度)的模块  依次是自动生成次刻度的定位器,等间距刻度定位器,自定义函数格式化刻度文本,格式字符串控制刻度的显示方式
from matplotlib.ticker import AutoMinorLocator,MultipleLocator,FuncFormatter,FormatStrFormatter

import os

if not os.path.isdir('Figures'):
    os.makedirs('Figures')
```

```python
p = plt.rcParams
#p['font.sans-serif'] = ['Roboto']
#全局字体粗细
p['font.weight'] = 'light'
p['axes.grid'] = True
p['grid.color'] = '0.5'
#设置网格粗细默认值
p['grid.linewidth'] = 0.5
```

```python
x_array = np.linspace(-15,15,200)
y_array = np.sin(x_array)
```

```python
fig = plt.figure(figsize = (6,6),tight_layout = True)

ax = fig.add_subplot(3,2,1)
ax.plot(x_array,y_array)
ax.set_xlim(-15,15)#范围左闭右闭
ax.set_ylim(-1.5,1.5)

ax = fig.add_subplot(3,2,2)
ax.plot(x_array,y_array)
ax.set_xlim(-15,15)#范围左闭右闭
ax.set_ylim(-1.5,1.5)
ax.set_xticks([-15,0,5,10,15])
ax.set_yticks([-0.8,-0.4,0,0.8])

ax = fig.add_subplot(3,2,3)
ax.plot(x_array,y_array)
ax.set_xlim(-15,15)#范围左闭右闭
ax.set_ylim(-1.5,1.5)
ax.tick_params(axis = 'x',rotation = 45,
               labelcolor = 'r',labelsize = 10)
ax.tick_params(axis = 'y',rotation = -45,
               labelcolor = 'r',labelsize = 10)

ax = fig.add_subplot(3,2,4)
ax.plot(x_array,y_array)
ax.set_xlim(-15,15)#范围左闭右闭
ax.set_ylim(-1.5,1.5)
#轴和脊的概念是分开的,不显示轴但是脊还存在
#如果改变脊的位置轴跟着动是因为轴的初始化位置是根据脊来的,但是脊不在轴也还是在的
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)


ax = fig.add_subplot(3,2,5)
ax.plot(x_array,y_array)
ax.set_xlim(-15,15)#范围左闭右闭
ax.set_ylim(-1.5,1.5)
#刻度标签显示格式控制器,里面传入的位置参数表示不显示任何刻度(输出空字符串)
    #只能接受formatter对象
    #这里刻度标签还是存在的但是是空字符串,所有刻度都是空字符串
    #这个生成对应位置为''的函数会检索主标签数量生成对应数量然后再传递给外侧函数
    #区别于set_ticks([]),这个是连刻度线都删除了
ax.yaxis.set_major_formatter(plt.NullFormatter())
ax.xaxis.set_major_formatter(plt.NullFormatter())

ax = fig.add_subplot(3,2,6)
ax.plot(x_array,y_array)
ax.set_xlim(-15,15)#范围左闭右闭
ax.set_ylim(-1.5,1.5)
ax.set_xticks(np.linspace(-4*np.pi,4*np.pi,5))
ax.set_xticklabels([r'$-4\pi$',r'$-2\pi$',
                    r'$0\pi$',r'$2\pi$',r'$4\pi$'])
```

```python
fig = plt.figure(figsize = (6,6),tight_layout = True)

ax = fig.add_subplot(3,2,1)
ax.plot(x_array,y_array)
ax.set_xlim(-15,15);ax.set_ylim(-1.5,1.5)
ax.set_xticks(np.linspace(-9/2*np.pi,9/2*np.pi,7))
ax.tick_params(axis = 'x',
               direction = 'in',
               color = 'b',
               width = 1,
               length = 10)
ax.tick_params(axis = 'y',
               direction = 'in',
               color = 'b',
               width = 1,
               length = 10)
ax.set_xticklabels([r'$-\frac{9\pi}{2}$',
                    r'$-3\pi$',
                    r'$-\frac{3\pi}{2}$',
                    r'$0\pi$',
                    r'$\frac{3\pi}{2}$',
                    r'$3\pi$',
                    r'$\frac{9\pi}{2}$'])

ax = fig.add_subplot(3,2,2)
ax.plot(x_array,y_array)
ax.set_xlim(-15,15);ax.set_ylim(-1.5,1.5)
ax.tick_params(axis = 'x',
               direction = 'inout',
               color = 'b',
               width = 1,
               length = 10)
ax.tick_params(axis = 'y',
               direction = 'inout',
               color = 'r',
               width = 1,
               length = 10)

ax = fig.add_subplot(3,2,3)
ax.plot(x_array,y_array)
ax.set_xlim(-15,15);ax.set_ylim(-1.5,1.5)
ax.minorticks_off()
ax.tick_params(bottom = False,
               top = True,
               left = False,
               right = True)

ax = fig.add_subplot(3,2,4)
ax.plot(x_array,y_array)
ax.set_xlim(-15,15);ax.set_ylim(-1.5,1.5)
ax.minorticks_off()
ax.tick_params(bottom = False,
               top = True,
               left = False,
               right = True)
ax.tick_params(labelbottom = False,
               labeltop = True,
               labelleft = False,
               labelright = True)

ax = fig.add_subplot(3,2,5)
ax.minorticks_off()
ax.plot(x_array,y_array)
ax.set_xlim(-15,15);ax.set_ylim(-1.5,1.5)
#创建次刻度定位器,对象为刻度定位器
minor_locator = AutoMinorLocator(5)
#这个函数只接受刻度定位器对象,默认带这个函数,但是传入的刻度定位器的生成需要导入函数才能生成
ax.xaxis.set_minor_locator(minor_locator)
#次刻度是后生成的,所以需要先配置后才能显示
plt.grid(which = 'minor')

ax = fig.add_subplot(3,2,6)
ax.minorticks_off()
ax.plot(x_array,y_array)
ax.set_xlim(-15,15);ax.set_ylim(-1.5,1.5)
minor_locator = AutoMinorLocator(5)
ax.yaxis.set_minor_locator(minor_locator)
plt.grid(which = 'minor')

fig.savefig('Figures/修改图轴,第2组.svg',format = 'svg')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter,MultipleLocator
```

```python
X = np.linspace(0.001,90,5000)
```

```python
#matplotlib扩展工具包中的3d子模块导入了3d坐标功能对象
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
```

```python
#生成一组内置3d实例数据
#范围带盖-30,30(X,Y)网格密度大约为0.05
#Z = 0.5*sin(X)*cos(Y)
X,Y,Z = axes3d.get_test_data(0.05)
```

```python
#这个layout比tight_layout更智能, tight这个仅仅是简单挤压,有时候可能出错
fig = plt.figure(figsize = (12,12),constrained_layout = True)

angle_array = np.linspace(0,180,13)
num_grid = len(angle_array)
gspec = fig.add_gridspec(num_grid,num_grid)

nrows,ncols = gspec.get_geometry()#返回网格布局形状

#不再用2,2,1这种定位方式,直接用网格形状定位,有了位置和形状,完全由gridspec决定(更灵活)适合大量构建子图
axs = np.array([[fig.add_subplot(gspec[i,j],projection = '3d') for j in range(ncols)] for i in range(nrows)])

for i in range(nrows):
    elev = angle_array[i]
    for j in range(ncols):
        azim = angle_array[j]

        #后两个参数是行步长和列步长(多少单位画一条线)行步长是y方向取点间隔,列步长是x方向取点间隔
        axs[i,j].plot_wireframe(X,Y,Z,rstride = 10,cstride = 10)
        axs[i,j].set_proj_type('ortho')
        axs[i,j].set_xticks([])
        axs[i,j].set_yticks([])
        axs[i,j].set_zticks([])
        axs[i,j].set_xlim(X.min(),X.max())
        axs[i,j].set_ylim(Y.min(),Y.max())
        axs[i,j].set_zlim(Z.min(),Z.max())

        axs[i,j].view_init(elev = elev,azim = azim)

fig.savefig('Figures/子图,三维曲面视角.svg',format = 'svg')
plt.show()
```
