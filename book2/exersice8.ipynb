```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_array = np.linspace(0,4*np.pi,9)
y_array = np.sin(x_array)

fig,ax = plt.subplots(figsize = (5,3))

plt.plot(x_array,y_array,'.--',ms = 10,color = '#0088FF')
ax.set_xlim((0,4*np.pi))
ax.set_ylim(-1.2,1.2)
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

fig.savefig('Figures/平面线图,低颗粒度_1.svg')
```

```python
x_array_coarse = np.linspace(0,4*np.pi,25)
y_array_coarse = np.sin(x_array_coarse)

x_array_fine = np.linspace(0,4*np.pi,101)
y_array_fine = np.sin(x_array_fine)

fig,ax = plt.subplots(figsize = (5,3))

plt.plot(x_array_fine,y_array_fine,'--',color = '#888888')
#参数where是默认的'pre'
#plt.step(x_array_coarse,y_array_coarse,'.-',ms = 10,color = '#0088FF')
plt.plot(x_array_coarse,y_array_coarse,'.-',drawstyle = 'steps',ms = 10,color = '#0088FF')

ax.set_xlim((0,4*np.pi))
ax.set_ylim((-1.2,1.2))
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

fig.savefig('Figures/阶跃,向前填充.svg')
```

```python
x_array_coarse = np.linspace(0,4*np.pi,25)
y_array_coarse = np.sin(x_array_coarse)

x_array_fine = np.linspace(0,4*np.pi,101)
y_array_fine = np.sin(x_array_fine)

fig,ax = plt.subplots(figsize = (5,3))

plt.plot(x_array_fine,y_array_fine,'--',color = '#888888')
plt.step(x_array_coarse,y_array_coarse,'.-',ms = 10,color = '#0088FF',where = 'mid')

ax.set_xlim((0,4*np.pi))
ax.set_ylim((-1.2,1.2))
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

fig.savefig('Figures/阶跃,中间填充.svg')
```

```python
x_array_coarse = np.linspace(0,4*np.pi,25)
y_array_coarse = np.sin(x_array_coarse)

x_array_fine = np.linspace(0,4*np.pi,101)
y_array_fine = np.sin(x_array_fine)

fig,ax = plt.subplots(figsize = (5,3))

plt.plot(x_array_fine,y_array_fine,'--',color = '#888888')
plt.step(x_array_coarse,y_array_coarse,'.-',ms = 10,color = '#0088FF',where = 'post')

ax.set_xlim((0,4*np.pi))
ax.set_ylim((-1.2,1.2))
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

fig.savefig('Figures/阶跃,向后填充.svg')
```

```python
num = 20
n_array = np.arange(1,num+1)
a_0 = 1
d = 2

a_n_array = a_0 + (n_array - 1)*d

fig,ax = plt.subplots(figsize = (5,3))

plt.stem(n_array,a_n_array)

ax.set_xlim((n_array.min(),n_array.max()))
ax.set_xlabel('$n$')
ax.set_ylabel('n_th term, $a_n$')

fig.savefig('Figures/火柴梗图,数列.svg')
```
