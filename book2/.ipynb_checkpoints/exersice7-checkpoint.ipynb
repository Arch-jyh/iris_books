```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
num_points = 21
XX,YY = np.meshgrid(np.linspace(-2,2,num_points),np.linspace(-2,2,num_points))
color_function = XX * np.exp(-(XX**2 + YY**2))

fig,ax = plt.subplots(figsize = (5,5))
plt.scatter(XX,YY,c = color_function,#c的形状必须和XX,YY相同
            s = 20,cmap = 'RdYlBu_r')#自动从mpl中的cm读取对应的cm,如果是cm对象了就直接用

ax.set_xlim(-2.05,2.05)
ax.set_ylim(-2.05,2.05)

ax.set_xticks(np.linspace(-2,2,6))
ax.set_yticks(np.linspace(-2,2,6))

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.yaxis.set_major_formatter(plt.NullFormatter())
ax.xaxis.set_major_formatter(plt.NullFormatter())

ax.set_xlabel(r'$x_1$')
ax.set_ylabel(r'$x_2$',rotation = 0)

plt.grid(color = (0.8,0.8,0.8))
ax.set_axisbelow(True)
#自动识别第一个字符串参数是否有.svg这种格式后缀(扩展名)
#如果字符串中没有格式后缀,format才是指定添加到字符串后面的格式后缀(扩展名)
fig.savefig('Figures/网络节点,低颗粒度.svg',format = 'svg')
```

```python
from sklearn.neighbors import KernelDensity#核密度估计
#.fit要求输入的是二维矩阵,所以用[[]]索引   取单列[''] 取多列输出2d的话pd索引不一样的[['','']]直接输入列名列表就行
#传入字符串默认提取列,只有在使用.loc和.iloc才会分行列
kde = KernelDensity(kernel = 'gaussian',bandwidth = 0.2).fit(iris_sns[['sepal_length','sepal_width']])
log_PDF = kde.score_samples(iris_sns[['sepal_length','sepal_width']])


fig,ax = plt.subplots()

ax = sns.scatterplot(data = iris_sns,
                     #设计规范就是按照字符串>>列名
                     x = 'sepal_length',
                     y = 'sepal_width',
                     c = log_PDF,
                     cmap = 'RdYlBu_r')

ax.set_xlabel('Sepal length (cm)')
ax.set_ylabel('Sepal width (cm)')
ax.set_xticks(np.arange(4,8+1,step = 1))
ax.set_yticks(np.arange(1,5+1,step = 1))
ax.axis('scaled')
ax.grid(linestyle = '--',linewidth = 0.25,color = [0.7,0.7,0.7])
ax.set_xbound(lower = 4,upper = 8)
ax.set_ybound(lower = 1,upper = 5)

norm = plt.Normalize(log_PDF.min(),log_PDF.max())
#生成颜色映射器,具有数值到颜色的功能,不用绘图自带的映射功能
sm = plt.cm.ScalarMappable(cmap="RdYlBu_r", norm=norm)
#必须有一个范围读取数据,这个属性必须存在.如果制定了其他读取方式会按照其他读取方式走,但是还是要求这里存在,所以配置一个空的
sm.set_array([])

#这是ax反向找到自己所在figure调出来,然后figure配置的colorbar不是在ax上,所以需要ax = ax指定
ax.figure.colorbar(sm,ax = ax)

fig.savefig('Figures/绘制样本数据散点图，PDF紧密程度.svg', format='svg')
```

```python
import matplotlib.pyplot as plt
import numpy as np
#scipy是科学计算库,导入空间计算 计算凸包
from scipy.spatial import ConvexHull
import seaborn as sns
```

```python
#**kw在定义和调用有两种用途
    #在定义部分表示将其余参数打包
    #在函数调用部分表示将打包的参数解包
def encircle(x,y,ax = None,**kw):
    if not ax: ax = plt.gca()
    p = np.c_[x,y]
    #计算凸包,返回hull的一个类不是点集
    hull = ConvexHull(p)
    #这里先用.vertices提取出hull类中的最外层点的数据,然后绘制矩形
    poly = plt.Polygon(p[hull.vertices,:],**kw)
    ax.add_patch(poly)
```

```python
iris_sns = sns.load_dataset('iris')

iris_sns.head()
```

```python
iris_sns.species.unique()
```

```python
fig,ax = plt.subplots()

ax = sns.scatterplot(data = iris_sns,
                     x = 'sepal_length',
                     y = 'sepal_width',
                     hue = 'species')

#.species也是一种等价的按照列提取数据的功能,返回列数据,和iris_sns['species']等价
#== 'setosa'得到布尔数组,用这个布尔数组做行筛选   pd数组很灵活和一用布尔数组和列名同时出现
encircle(iris_sns.loc[iris_sns.species == 'setosa','sepal_length'].to_numpy(),
         iris_sns.loc[iris_sns.species == 'setosa','sepal_width'].to_numpy(),
         #edgecolor facecolor
         ec = 'blue',fc = 'blue',alpha = 0.2)#
encircle(iris_sns.loc[iris_sns.species == 'versicolor','sepal_length'].to_numpy(),
         iris_sns.loc[iris_sns.species == 'versicolor','sepal_width'].to_numpy(),
         ec = 'orange',fc = 'orange',alpha = 0.2)
encircle(iris_sns.loc[iris_sns.species == 'virginica','sepal_length'].to_numpy(),
         iris_sns.loc[iris_sns.species == 'virginica','sepal_width'].to_numpy(),
         ec = 'green',fc = 'green',alpha = 0.2)

ax.set_xlabel('Seapal length (cm)')
ax.set_ylabel('Seapal width (cm)')
ax.set_xticks(np.arange(4,8+1,step = 1))
ax.set_yticks(np.arange(1,5+1,step = 1))

ax.axis('scaled')
ax.grid(linestyle = '--',linewidth = 0.25,color = [0.7,0.7,0.7])
ax.set_xbound(4,8)
ax.set_ybound(1,5)

fig.savefig('Figures/散点包络先.svg')
```

```python
#pd和numpy中&可以作用于两个布尔数组,是数组逻辑,不能用and
#做与运算1和1才能为1 得到数据中中间的值
mask_outside = ((X>-2)&(X<2))

fig,ax = plt.subplots()

#内侧点.不满足配置Nan
X_inside = X.copy()
#~是布尔数组反算,得到只保留的内侧点数据,外侧点对应true的地方为nan
X_inside[~mask_outside] = np.nan

#外侧点.满足配置nan
X_outside = X.copy()
#内侧点赋予nan,保留了外侧点
X_outside[mask_outside] = np.nan

colors = np.array(['#377eb8','#ff7f00'])

plt.scatter(np.arange(num),X_inside,color = colors[0],marker = '.')
plt.scatter(np.arange(num),X_outside,color = colors[1],marker = 'x')

plt.axhline(y = 0,color = 'r')
plt.axhline(y = -2,color = 'r')
plt.axhline(y = 2,color = 'r')
plt.xlim(0,num)
plt.ylim(-4,4)

fig.savefig('Figures/正负两倍标准差,区分内外.svg')
print('Number of points outside = ' + str(mask_outside.sum()))
print('Percentage of points outside = ' + str(mask_outside.sum()/num*100) + '%')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num = 2000
#可以指定范围,均匀分布
X = np.random.uniform(low = -1,high = 1,size = (num,2))
```

```python
fig,ax = plt.subplots()
plt.scatter(X[:,0],X[:,1],marker = '.')
circ = plt.Circle((0,0),radius = 1,edgecolor = 'k',facecolor = 'None')
ax.add_patch(circ)

ax.set_aspect('equal',adjustable = 'box')
plt.xlim(-1,1)
plt.ylim(-1,1)
ax.set_xticks((-1,0,1))
ax.set_yticks((-1,0,1))
fig.savefig('Figures/估算圆周率,不区分内外.svg')
```

```python
mask_inside = (X[:,0]**2 + X[:,1]**2 <= 1)
mask_inside
```

```python
fig,ax = plt.subplots()

X_inside = X[mask_inside,:]
X_outside = X[~mask_inside,:]

colors = np.array(['#377eb8','#ff7f00'])

circ = plt.Circle((0,0),radius = 1,edgecolor = 'k',facecolor = 'None')
ax.add_patch(circ)
plt.scatter(X_inside[:,0],X_inside[:,1],color = colors[0],marker = '.')
plt.scatter(X_outside[:,0],X_outside[:,1],color = colors[1],marker = 'x')

ax.set_aspect('equal',adjustable = 'box')
plt.xlim(-1,1)
plt.ylim(-1,1)
ax.set_xticks((-1,0,1))
ax.set_yticks((-1,0,1))
fig.savefig('Figures/估算圆周率,区分内外.svg')
```
