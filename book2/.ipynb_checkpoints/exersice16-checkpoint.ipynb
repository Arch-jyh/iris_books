```python
import math
import numpy as np
import matplotlib.pyplot as plt
from sympy import lambdify ,diff,exp,latex
from sympy.abc import x,y

from matplotlib import cm
```

```python
f_xy =  3*(1-x)**2*exp(-(x**2) - (y+1)**2)\
    - 10*(x/5 - x**3 - y**5)*exp(-x**2-y**2)\
    - 1/3*exp(-(x+1)**2 - y**2) 

f_xy_fcn = lambdify([x,y],f_xy)

def mesh(num = 101):
    
    # number of mesh grids
    x_array = np.linspace(-3,3,num)
    y_array = np.linspace(-3,3,num)
    xx,yy = np.meshgrid(x_array,y_array)
    
    return xx, yy

xx, yy = mesh(num = 101)

ff = f_xy_fcn(xx,yy)
```

```python
fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

ax.plot_wireframe(xx,yy,ff,
                  color = '0.8',
                  rstride = 5,cstride = 5,
                  linewidth = 0.25)

#默认就是z方向绘制等高线
ax.contour(xx,yy,ff,levels = 20,cmap = 'RdYlBu_r')

ax.set_proj_type('ortho')

ax.set_xlabel(r'$\it{x}$')
ax.set_ylabel(r'$\it{y}$')
ax.set_zlabel(r'$\it{f}$($\it{x}$,$\it{y}$)')

ax.set_xlim(xx.min(),xx.max())
ax.set_ylim(yy.min(),yy.max())
ax.set_zlim(-8,8)
ax.view_init(azim = -120,elev = 30)
ax.grid(False)
```

```python
fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

ax.plot_wireframe(xx,yy,ff,
                  color = '0.8',
                  rstride = 5,cstride = 5,
                  linewidth = 0.25)

ax.contour(xx,yy,ff,
           #指定投影方向,offset指定了偏移值,沿着z轴正方向移动,将所有等高线平移到z = 8上
               #区别于zs zs是配置xyz的一种快捷配置值的方法,可以配置xy后用zs = 5快捷指定所有z的值为5,是广播的作用(和现在的广播等价,是早期接口)
           zdir = 'z',offset = 8,
           levels = 20,cmap = 'RdYlBu_r')

ax.set_proj_type('ortho')

ax.set_xlabel(r'$\it{x}$')
ax.set_ylabel(r'$\it{y}$')
ax.set_zlabel(r'$\it{f}$($\it{x}$,$\it{y}$)')

ax.set_xlim(xx.min(),xx.max())
ax.set_ylim(yy.min(),yy.max())
ax.set_zlim(-8,8)
ax.view_init(azim = -120,elev = 30)
ax.grid(False)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
#第一个和最后一个分别是将符号转换为可计算函数和简化符号表达式(简化只针对符号表达式,化简符号)
#转换为可计算函数的过程不进行化简
from sympy import lambdify,diff,exp,latex,simplify
#导入模块用于定义符号变量
from sympy import symbols

from matplotlib import cm
```

```python
def mesh(num = 101):
    x_array = np.linspace(-1.2,1.2,num)
    y_array = np.linspace(-1.2,1.2,num)
    xx_yy = np.meshgrid(x_array,y_array)

    return xx,yy
```

```python
x1,x2 = symbols('x1 x2')
def quadratic(Q,xx1,xx2):

    x = np.array([[x1],
                  [x2]])

    f_x1x2 = x.T @ Q @ x

    #加[0][0]是为了取出来标量
    f_x1x2_fcn = lambdify([x1,x2],f_x1x2[0][0])

    ff = ff_x1x2_fcn(xx1,xx2)

    #返回数值结果和符号公式
    return ff,simplify(f_x1x2[0][0])
```

```python
def visualize(Q,title):

    xx1,xx2 = mesh(num = 201)
    ff,f_x1x2 = quadratic(Q,xx1,xx2)

    theta_array = np.linspace(0,2*np.pi,100)
    x1_circle = np.cos(theta_array)
    x2_circle = np.sin(theta_array)

    fig = plt.figure(figsize = (8,4))

    ax = fig.add_subplot(1,2,1)
```
