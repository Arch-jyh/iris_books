```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.array([[-2],
              [5]])
b = np.array([[5],
              [-1]])
```

```python
c = a+b

fig,ax = plt.subplots(figsize = (5,5))

#字典的键必须是字符串形式
quiver_kw = {'angles':'xy','scale_units':'xy','scale' : 1}

ax.quiver(0,0,a[0],a[1],
          #angles表示向量在xy轴的角度进行显示,scale_units表示按照x,y的缩放单位进行表示
          #**专用于字典,使用时候将字符串键映射为参数名,值不变
          **quiver_kw,color = '#92D050')
ax.text(-3,2.5,'$a (-2,5)$',fontsize = 10)

ax.quiver(0,0,b[0],b[1],
          **quiver_kw,color = '#FFC000')
ax.text(1.5,-1.5,'$b (5,-1)$',fontsize = 10)

ax.quiver(0,0,c[0],c[1],
          **quiver_kw,color = '#0099FF')
ax.text(2,2,'$c (3,4)$',fontsize = 10)

#绘制连线
ax.plot([a[0],c[0]],
          [a[1],c[1]],
        c = 'k',ls = '--')

ax.plot([b[0],c[0]],
          [b[1],c[1]],
        c = 'k',ls = '--')

fill_color = np.array([219,238,243])/255

X = np.array([[0,0],
              [-2,5],
              [3,4],
              [5,-1],
              [0,0]])#实际上少了最后一个坐标值fill也能自动连接第一个点完成填充

plt.fill(X[:,0],X[:,1],
         color = fill_color,
         edgecolor = None,
         alpha = 0.5)

ax.axhline(y = 0,c = 'k',lw = 0.25)
ax.axvline(x = 0,c = 'k',lw = 0.25)

ax.set_xlim([-6,6])
ax.set_ylim([-6,6])
ax.set_xticks(np.arange(-5,6))
ax.set_yticks(np.arange(-5,6))
ax.set_aspect('equal',adjustable = 'box')
ax.grid(c = '0.8',zorder = 1e3)

ax.spines[['top','right','bottom','left']].set_visible(False)


ax.set_xticklabels([])
ax.set_yticklabels([])
plt.show()
```

```python
fig,ax = plt.subplots(figsize = (5,5))

quiver_kw = {'angles':'xy','scale_units':'xy','scale' : 1}

ax.quiver(0,0,c[0],c[1],
          **quiver_kw,color = '#0099FF')
ax.text(2,2,'$c (3,4)$',fontsize = 10)
#rotation的输入值单位是度,这里先用弧度然后转换为度
ax.text(0.5,2,r'$\sqrt{3^2 + 4^2} = 5$',fontsize = 10,rotation = np.arctan(4/3) * 180/np.pi)

ax.axhline(y = 0,c = 'k',lw = 0.25)
ax.axvline(x = 0,c = 'k',lw = 0.25)

ax.set_xlim([-6,6])
ax.set_ylim([-6,6])
ax.set_xticks(np.arange(-5,6))
ax.set_yticks(np.arange(-5,6))
ax.set_aspect('equal',adjustable = 'box')
ax.grid(c = '0.8',zorder = 1e3)

ax.spines[['top','right','bottom','left']].set_visible(False)


ax.set_xticklabels([])
ax.set_yticklabels([])
plt.show()
```

```python
d = a - b

fig,ax = plt.subplots(figsize = (5,5))

quiver_kw = {'angles':'xy','scale_units':'xy','scale' : 1}

ax.quiver(0,0,a[0],a[1],
          **quiver_kw,color = '#92D050')
ax.text(-3,2.5,'$a (-2,5)$',fontsize = 10)

ax.quiver(0,0,b[0],b[1],
          **quiver_kw,color = '#FFC000')
ax.text(1.5,-1.5,'$b (5,-1)$',fontsize = 10)

ax.quiver(b[0],b[1],d[0],d[1],
          **quiver_kw,color = '#000000')
ax.text(2,2,'$d = a - b, (-7,6)$',fontsize = 10)

fill_color = np.array([219,238,243])/255

X = np.array([[0,0],
              [-2,5],
              [5,-1],
              [0,0]])

plt.fill(X[:,0],X[:,1],
         color = fill_color,
         edgecolor = None,
         alpha = 0.5)

ax.axhline(y = 0,c = 'k',lw = 0.25)
ax.axvline(x = 0,c = 'k',lw = 0.25)

ax.set_xlim([-6,6])
ax.set_ylim([-6,6])
ax.set_xticks(np.arange(-5,6))
ax.set_yticks(np.arange(-5,6))
ax.set_aspect('equal',adjustable = 'box')
ax.grid(c = '0.8',zorder = 1e3)

ax.spines[['top','right','bottom','left']].set_visible(False)


ax.set_xticklabels([])
ax.set_yticklabels([])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1,2,4])
b = np.array([4,3,1])
```

```python
c = a+b

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

ax.quiver(0,0,0,
          a[0],a[1],a[2],
          color = '#92D050',
          normalize = False,#不进行归一化.归一化是表示为单位向量,只表示方向
          arrow_length_ratio = .07,#箭头头部与总长图的比例  另一个函数arrow是绘制单个箭头图形(针对符号,2d中),这里quiver可以绘制一组箭头
          linestyles = 'solid',#样式为实现
          linewidths = 1)

ax.quiver(0,0,0,
          b[0],b[1],b[2],
          color = '#FFC000',
          normalize = False,
          arrow_length_ratio = .07,
          linestyles = 'solid',
          linewidths = 1)

ax.quiver(0,0,0,
          c[0],c[1],c[2],
          color = '#0099FF',
          normalize = False,
          arrow_length_ratio = .07,
          linestyles = '-',
          linewidths = 1)

ax.plot([a[0],c[0]],
        [a[1],c[1]],
        [a[2],c[2]],
        ls = '--',
        lw = 1,
        c = 'k')

ax.plot([b[0],c[0]],
        [b[1],c[1]],
        [b[2],c[2]],
        ls = '--',
        lw = 1,
        c = 'k')

ax.set_proj_type('ortho')
ax.view_init(azim = 30,elev = 30)
ax.set_box_aspect([1,1,1])

ax.xaxis.set_pane_color((1.0,1.0,1.0,1.0))
ax.yaxis.set_pane_color((1,1,1,1))
ax.zaxis.set_pane_color((1.0,1.0,1.0,1.0))

ax.set_xlim(0,5)
ax.set_ylim(0,5)
ax.set_zlim(0,5)

ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
```
