```python
import math
import numpy as np
import matplotlib.pyplot as plt
from sympy import lambdify,diff,exp,latex
from sympy.abc import x,y
from matplotlib import cm
```

```python
f_xy =  3*(1-x)**2*exp(-(x**2) - (y+1)**2)\
    - 10*(x/5 - x**3 - y**5)*exp(-x**2-y**2)\
    - 1/3*exp(-(x+1)**2 - y**2)

f_xy_fcn = lambdify([x,y],f_xy)
```

```python
def mesh(num = 101):
    x_array = np.linspace(-3,3,num)
    y_array = np.linspace(-3,3,num)
    xx,yy = np.meshgrid(x_array,y_array)

    return xx,yy
```

```python
xx,yy = mesh(num = 11)
zz = xx*0

fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

ax.plot_wireframe(xx,yy,zz,
                  color = '0.5',
                  lw = 0.25)

ax.set_proj_type('ortho')

ax.set_xlabel(r'$\it{x}$')
ax.set_ylabel(r'$\it{y}$')
ax.set_zlabel('')
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

ax.set_xlim(xx.min(),xx.max())
ax.set_ylim(yy.min(),yy.max())

ax.view_init(azim = -135,elev = 30)
ax.grid(False)
```

```python
ff = f_xy_fcn(xx,yy)

fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

ax.plot_wireframe(xx,yy,ff,
                  color = '0.5',
                  lw = 0.25)

ax.set_proj_type('ortho')

ax.set_xlabel(r'$\it{x}$')
ax.set_ylabel(r'$\it{y}$')
ax.set_zlabel('')
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

ax.set_xlim(xx.min(),xx.max())
ax.set_ylim(yy.min(),yy.max())

ax.view_init(azim = -135,elev = 30)
ax.grid(False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from sympy.abc import x,y
from sympy import lambdify,diff,exp,latex
```

```python
def mesh(num = 101):
    x_array = np.linspace(-3,3,num)
    y_array = np.linspace(-3,3,num)
    xx,yy = np.meshgrid(x_array,y_array)

    return xx,yy
```

```python
xx,yy = mesh(num = 201)
f_xy =  3*(1-x)**2*exp(-(x**2) - (y+1)**2)\
    - 10*(x/5 - x**3 - y**5)*exp(-x**2-y**2)\
    - 1/3*exp(-(x+1)**2 - y**2) 
f_xy_fcn = lambdify([x,y],f_xy)
f_xy_zz = f_xy_fcn(xx,yy)
```

```python
fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

surf = ax.plot_surface(xx,yy,f_xy_zz,
                       cmap = 'turbo',
                       lw = 1,
                       #停用阴影
                       shade = False)

ax.set_proj_type('ortho')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$f(x,y)$')

ax.set_xlim(xx.min(),xx.max())
ax.set_ylim(yy.min(),yy.max())
ax.set_zticks([])
#shrink缩放长度,aspect控制长宽比,分别对应了长和宽
fig.colorbar(surf,shrink = .8,aspect = 20)#
ax.view_init(azim = -120,elev = 30)
plt.tight_layout()
ax.grid(False)
```

```python
fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

norm_plt = plt.Normalize(f_xy_zz.min(),f_xy_zz.max())
colors = cm.turbo(norm_plt(f_xy_zz))

surf = ax.plot_surface(xx,yy,f_xy_zz,
                       #这里是传入面片指定的颜色,cmap是颜色映射
                       facecolors = colors,
                       lw = 1,
                       shade = False)
surf.set_facecolor((0,0,0,0))

ax.set_proj_type('ortho')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$f(x,y)$')

ax.set_xlim(xx.min(),xx.max())
ax.set_ylim(yy.min(),yy.max())
ax.set_zticks([])

ax.view_init(azim = -120,elev = 30)
plt.tight_layout()
ax.grid(False)
```

```python
import numpy as np
from scipy.stats import dirichlet
import matplotlib.pyplot as plt
from matplotlib import cm
import os
```

```python
def visualize(alpha_array,mesh_only = True):

    theta_1_array = theta_2_array = np.linspace(0,1,401)

    tt1,tt2 = np.meshgrid(theta_1_array,theta_2_array)
    tt3 = 1 - tt1 - tt2

    PDF_array = []

    for t_1_idx,t_2_idx in zip(tt1.ravel(),tt2.ravel()):

        t_3_idx = 1 - t_1_idx - t_2_idx
        if t_3_idx < 0.005:
            PDF_idx = np.nan
        else:
            PDF_idx = dirichlet.pdf([t_1_idx,t_2_idx,t_3_idx],alpha_array)

        PDF_array.append(PDF_idx)

    PDF_FF = np.reshape(PDF_array,tt1.shape)

    fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

    #归一化器初始化不能接受nan,必须有值
    #默认的.min和.max只要有nan输出就是nan,这里用nanmin过滤了nan
    norm_plt = plt.Normalize(np.nanmin(PDF_FF),np.nanmax(PDF_FF))
    #归一化器本身支持输入nan,输出点还是nan
    colors = cm.RdYlBu_r(norm_plt(PDF_FF))

    surf = ax.plot_surface(tt1,tt2,tt3,
                           facecolors = colors,
                           linewidth = 0.25,shade = False,
                           cstride = 10,rstride = 10)

    if mesh_only:

        surf.set_facecolor((0,0,0,0))

    ax.view_init(azim = 30,elev = 30)
    ax.set_proj_type('ortho')
    ax.set_box_aspect([1,1,1])
    ax.set_xlim((0,1))
    ax.set_ylim(0,1)
    ax.set_zlim(0,1)
    ax.set_xlabel(r'$\theta_1$')
    ax.set_ylabel(r'$\theta_2$')
    ax.set_zlabel(r'$\theta_3$')
    ax.set_xticks([0,1])
    ax.set_yticks([0,1])
    ax.set_zticks([0,1])
    ax.grid(False)
    #join接收可迭代对象
    #(str(v) for v in alpha_array)
        #如果for外面套上了()表示这里是生成器语法,是迭代器
        #如果join((str(v) for v in alpha_array))中的参数只有生成器表达式,一层()省略变成现在的样子
    title = '_'.join(str(v) for v in alpha_array)
    title = 'alphas_' + title

    if mesh_only:
        title = title + '_mesh_only'

    fig.savefig('Figures/' + title + '.svg')
```

```python
alpha_array = [1,2,2]
visualize(alpha_array,True)
```

```python
alpha_array = [1,2,2]
visualize(alpha_array,False)
```

```python
alpha_array = [2,1,2]
visualize(alpha_array,True)
```

```python
alpha_array = [2,1,2]
visualize(alpha_array,False)
```

```python
alpha_array = [2,2,1]
visualize(alpha_array,True)
```

```python
alpha_array = [2,2,1]
visualize(alpha_array,False)
```

```python
alpha_array = [2,2,2]
visualize(alpha_array,True)
```

```python
alpha_array = [2,2,2]
visualize(alpha_array,False)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from sympy import lambdify,diff,exp,latex #导入的模块分别是:将符号转换为可计算函数,针对符号计算偏导数,符号形式的指数函数,将符号表达式转换为latex字符串
from sympy.abc import x,y
import os

from matplotlib import cm
```

```python
f_xy =  3*(1-x)**2*exp(-(x**2) - (y+1)**2)\
    - 10*(x/5 - x**3 - y**5)*exp(-x**2-y**2)\
    - 1/3*exp(-(x+1)**2 - y**2) 

f_xy_fcn = lambdify([x,y],f_xy)
```

```python
def mesh(num = 101):
    x_array = np.linspace(-3,3,num)
    y_array = np.linspace(-3,3,num)
    xx,yy = np.meshgrid(x_array,y_array)

    return xx,yy
```

```python
xx,yy = mesh(num = 121)
ff = f_xy_fcn(xx,yy)
```

```python
z_level = 2 #z的高度

xx_,yy_ = np.meshgrid(np.linspace(-3,3,2),np.linspace(-3,3,2))

fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

zz_ = np.zeros_like(xx_) + (z_level)

ax.plot_surface(xx_,yy_,zz_,color = 'b',alpha = 0.1)
ax.plot_wireframe(xx_,yy_,zz_,color = 'b',
                  lw = 0.2)

ax.plot_wireframe(xx,yy,ff,
                  color = [0.6,0.6,0.6],
                  rstride = 5,cstride = 5,
                  lw = 0.25)

ax.contour(xx,yy,ff,
           levels = [z_level],
           colors = 'r',
           linewidths = 1)

ax.set_proj_type('ortho')

ax.set_xlabel(r'$\it{x}$') #\it表示斜体
ax.set_ylabel(r'$\it{y}$')
ax.set_zlabel(r'$\it{f}$($\it{x}$,$\it{y}$)')

ax.set_xlim(xx.min(),xx.max())
ax.set_ylim(yy.min(),yy.max())
ax.set_zlim(-8,8)
ax.view_init(azim = -120,elev = 30)
ax.grid(False)
fig.savefig('Figures/剖面线,平行于xy.svg',format = 'svg')
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import multivariate_normal
from matplotlib import cm

from matplotlib.collections import PolyCollection
```

```python
def mesh(num = 101):
    x_array = np.linspace(-3,3,num)
    y_array = np.linspace(-3,3,num)
    xx,yy = np.meshgrid(x_array,y_array)

    return xx,yy
```

```python
xx1,xx2 = mesh(num = 101)
#h表示水平方向(列方向,axis = 1),v表示垂直方向(行方向,axis=0),d表示深度方向axis = 2
#从0-2依次向内,这里堆叠完成后从(m,n) >> (m,n,2)
points = np.dstack((xx1,xx2))

bivariate_normal = multivariate_normal([0,0],
                                       [[1,-0.6],
                                        [-0.6,1]])
#二元高斯分布概率密度函数值
PDF_ff = bivariate_normal.pdf(points)
```

```python
points
```

```python
fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

ax.plot_wireframe(xx1,xx2,PDF_ff,
                  color = '0.5',
                  rstride = 0,cstride = 2,
                  lw = 0.25)

x1 = np.linspace(-3,3,101)
x2 = np.linspace(-3,3,101)
#arange是间隔为10
x1_loc_array = np.arange(0,len(x1),10)
facecolors = cm.rainbow(np.linspace(0,1,len(x1_loc_array)))

for idx in range(len(x1_loc_array)):
    x_loc = x1_loc_array[idx]
    x_idx = x1[x_loc]
    x_i_array = x2*0 + x_idx
    z_array = PDF_ff[:,x_loc]

    ax.plot(x_i_array,x2,z_array,color = facecolors[idx,:],
            lw = 1.5)

    verts = [
        list(zip(x2,z_array)) + 
        list(zip(x2[::-1],np.zeros_like(z_array)))]#反向为了正确闭合多边形
    #调用创建2d多边形面片
    poly = PolyCollection(
        verts,
        facecolors = [facecolors[idx,:]],
        alpha = 0.2)

    ax.add_collection3d(poly,
                        #指定了x轴的坐标
                        zs = x_idx,
                        #放置到x桌上的位置
                        zdir = 'x')

ax.set_proj_type('ortho')

ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('Joint PDF, $f_{X_1,X_2}(x_1,x_2)$')
ax.set_xlim(xx1.min(), xx1.max())
ax.set_ylim(xx2.min(), xx2.max())
ax.set_zticks([0, 0.05, 0.1, 0.15, 0.2])
ax.set_zlim3d([0,0.2])
ax.view_init(azim=-120, elev=30)
ax.grid(False)
fig.savefig('Figures/指定 x1 具体值.svg', format='svg')
```

```python
fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

ax.plot_wireframe(xx1,xx2,PDF_ff,
                  color = '0.5',
                  #r表示行步长,c表示列步长
                  rstride = 2,cstride = 0,
                  lw = 0.25)

x2_loc_array = np.arange(0,len(x1),10)
facecolors = cm.rainbow(np.linspace(0,1,len(x2_loc_array)))

for idx in range(len(x2_loc_array)):
    x_loc = x2_loc_array[idx]
    x_idx = x2[x_loc]
    x_i_array = x1*0 + x_idx
    z_array = PDF_ff[x_loc,:]

    ax.plot(x1,x_i_array,z_array,color = facecolors[idx,:],
            lw = 1.5)

    verts = [
        list(zip(x1,z_array)) + 
        list(zip(x1[::-1],np.zeros_like(z_array)))]
    poly = PolyCollection(
        verts,
        facecolors = [facecolors[idx,:]],
        alpha = 0.2)

    ax.add_collection3d(poly,
                        zs = x_idx,
                        zdir = 'y')

ax.set_proj_type('ortho')

ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('Joint PDF, $f_{X_1,X_2}(x_1,x_2)$')
ax.set_xlim(xx1.min(), xx1.max())
ax.set_ylim(xx2.min(), xx2.max())
ax.set_zticks([0, 0.05, 0.1, 0.15, 0.2])
ax.set_zlim3d([0,0.2])
ax.view_init(azim=-120, elev=30)
ax.grid(False)
fig.savefig('Figures/指定 x2 具体值.svg', format='svg')
```

```python
import numpy as np
#从scipy(numpy扩展的数值计算库)的special模块导入两个函数
from scipy.special import jn,jn_zeros
import matplotlib.pyplot as plt

mmax = 5
```

```python
def displacement(n,m,r,theta,mmax = 5):
    k = jn_zeros(n,mmax + 1)[m]
    return np.sin(n*theta)*jn(n,r*k)

r = np.linspace(0,1,1001)
theta = np.linspace(0,2*np.pi,1001)
xx = np.array([rr*np.cos(theta) for rr in r])
yy = np.array([rr*np.sin(theta) for rr in r])
```

```python
def visualize(n,m,title):
    zz = np.array([displacement(n,m,rr,theta) for rr in r])

    fig = plt.figure(figsize = (8,4))
    ax = fig.add_subplot(121,projection = '3d')

    surf = ax.plot_wireframe(xx,yy,zz,
                            cstride = 50,
                            rstride = 50,
                            colors = '0.8',
                            lw = 0.25)

    ax.contour(xx,yy,zz,cmap = 'RdYlBu_r',
               #传入累表表示按照列表中的数值绘制,传入15表示分成15条等高线绘制
              levels = 15,lw = 1)

    ax.set_proj_type('ortho')

    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$f(x,y)$')

    ax.set_xlim(xx.min(),xx.max())
    ax.set_ylim(yy.min(),yy.max())
    ax.set_zlim(zz.min()*5,zz.max()*5)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    ax.view_init(azim = 15,elev = 45)
    plt.tight_layout()
    ax.grid(False)
    ax.axis('off')

    ax = fig.add_subplot(122)

    ax.contourf(xx,yy,zz,cmap = 'RdYlBu_r',levels = 15)
    ax.contour(xx,yy,zz,colors = 'w',levels = 15,lw = 0.25)

    ax.plot(np.cos(theta),np.sin(theta),'k')
```
