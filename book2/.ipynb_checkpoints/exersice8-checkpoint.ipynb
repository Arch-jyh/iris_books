```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_array = np.linspace(0,4*np.pi,9)
y_array = np.sin(x_array)

fig,ax = plt.subplots(figsize = (5,3))

plt.plot(x_array,y_array,'.--',ms = 10,color = '#0088FF')
ax.set_xlim((0,4*np.pi))
ax.set_ylim(-1.2,1.2)
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

fig.savefig('Figures/平面线图,低颗粒度_1.svg')
```

```python
x_array = np.linspace(0,4*np.pi,13)
y_array = np.sin(x_array)

fig,ax = plt.subplots(figsize = (5,3))

#.--表示每个数据点都标记'.',默认是每个点都画,这是快捷方式,将这个参数传递给fmt,会自动解析成marker和line的两个参数下的数据
plt.plot(x_array,y_array,'.--',ms = 10,color = '#0088FF')
ax.set_xlim((0,4*np.pi))
ax.set_ylim(-1.2,1.2)
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

fig.savefig('Figures/平面线图,低颗粒度_2.svg')
```

```python
x_array = np.linspace(-0.1,0.1,100001)
y_array = np.sin(1/x_array)

fig,ax = plt.subplots(figsize = (5,3))

plt.plot(x_array,y_array,color = '#0088FF')
ax.set_xlim((x_array.min(),x_array.max()))
ax.set_ylim(-1.2,1.2)
ax.set_xticks((x_array.min(),0,x_array.max()))
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

fig.savefig('Figures/平面线图,根据函数需要配置更高颗粒度.svg')
```
