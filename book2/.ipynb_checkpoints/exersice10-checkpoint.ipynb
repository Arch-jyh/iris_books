```python
#符号函数
from sympy import symbols
x,y = symbols('x,y')
from sympy import lambdify,exp

f = x*exp(-x**2 - y**2)
#创建函数不检查变量是否与函数匹配所以要检查好
#def也是将函数复制给变量,所以只要变量指向的是函数,使用()就是表示调用
f_fcn = lambdify([x,y],f)
```

```python
fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

ax.plot_wireframe(xx1_square,xx2_square,f_fcn(xx1_square,xx2_square),
                  color = '0.5',lw = 0.25)

#传入给c的形状和点数一致才会做归一化
ax.scatter(xx1_square,xx2_square,f_fcn(xx1_square,xx2_square),s = 5,
           c = f_fcn(xx1_square,xx2_square),cmap = 'RdYlBu_r')

ax.set_proj_type('ortho')

ax.set_xlabel(r'$\it{x_1}$')
ax.set_ylabel(r'$\it{x_2}$')
ax.set_zlabel(r'$\it{x_3}$')
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

ax.set_xlim(xx1_square.min(),xx1_square.max())
ax.set_ylim(xx2_square.min(),xx2_square.max())
ax.set_box_aspect([1,1,1])
ax.view_init(azim = -120,elev = 30)
ax.grid(False)

fig.savefig('Figures/网格状散图,函数.svg')
```

```python
def circle_points(num_r,num_n):
    r = np.linspace(0,3,num_r)

    n = r*num_n+1
    n = n.astype(int)

    #0行2列空数组
    #0行表示0这个维度可以不断增加,2列表示这个维度是固定的
    circles = np.empty((0,2))

    for r_i,n_i in zip(r,n):

        t_i = np.linspace(0,2*np.pi,n_i,endpoint = False)
        r_i = np.ones_like(t_i)*r_i

        x_i = r_i*np.cos(t_i)
        y_i = r_i*np.sin(t_i)

        #第一个参数就是需要放在一起的拼接的多个数据组成的集合
        circle_i = np.column_stack([x_i,y_i])
        #
        circles = np.append(circles,circle_i,axis = 0)

    return circles
```

```python
points_circles = circle_points(10,20)

triang_circles_auto = mtri.Triangulation(points_circles[:,0],points_circles[:,1])

fig,ax = plt.subplots()

ax.triplot(triang_circles_auto,'o-',color = '#0070C0')
ax.set_aspect('equal',adjustable = 'box')

ax.set_xlabel(r'$\it{x_1}$')
ax.set_ylabel(r'$\it{x_2}$')

ax.axis('off')

ax.set_xlim(xx1_polar.min(),xx1_polar.max())
ax.set_ylim(xx2_polar.min(),xx2_polar.max())

fig.savefig('Figures/三角网格,均匀圆盘散点.svg')
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
#这是导入的具体函数 第一个是转换为python函数,后面的都是针对符号的操作函数(操作符号表达式的函数)
#引文python本身没有这些diff,exp,latex的函数,如果针对符号计算需要导入
from sympy import lambdify,diff,exp,latex
from sympy.abc import x,y
from matplotlib import cm
```

```python
def mesh(num = 101):
    x_array = np.linspace(-3,3,num)
    y_array = np.linspace(-3,3,num)
    xx,yy = np.meshgrid(x_array,y_array)

    return xx,yy
```

```python
f_xy = 3*(1-x)**2*exp(-(x**2) - (y+1)**2)\
     - 10*(x/5 - x**3 - y**5)*exp(-x**2 - y**2)\
     - 1/3*exp(-(x+1)**2 - y**2)

f_xy_fcn = lambdify([x,y],f_xy)

xx,yy = mesh(num = 121)
x_array = np.linspace(-3,3,101)
ff = f_xy_fcn(xx,yy)
```

```python
z_level = 2
xx_,yy_ = np.meshgrid(np.linspace(-3,3,2),np.linspace(-3,3,2))

fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

zz_ = np.zeros_like(xx_) + z_level
#按照网格点形成四边形网格骨架填充成实体面,和下面的结构是一样的
ax.plot_surface(xx_,yy_,zz_,color = 'b',alpha = 0.1)
#按照网格点形成四边形网格骨架
ax.plot_wireframe(xx_,yy_,zz_,color = 'b',lw = 0.2)

ax.plot_wireframe(xx,yy,ff,color = '0.6',rstride = 5,cstride = 5,lw = 0.25)

ax.contour(xx,yy,ff,levels = [z_level],colors = 'b',linewidths = 1)

ax.set_proj_type('ortho')

ax.set_xlabel(r'$\it{x_1}$')
ax.set_xlabel(r'$\it{x_2}$')
ax.set_xlabel(r'$\it{f}$($\it{x_1}$,$\it{x_2}$)')

ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

ax.set_xlim(xx.min(),xx.max())
ax.set_ylim(yy.min(),yy.max())
ax.set_zlim(-8,8)
ax.view_init(azim = -120,elev = 30)
ax.grid(False)
fig.savefig('Figures/等高线原理,空间一条等高线.svg')
```

```python
import numpy as np
from matplotlib import pyplot as plt
#导入knn临近算法和内置数据集
#sklearn是一个python机器学习库,用于传统机器学习
from sklearn import neighbors,datasets
#可以创建自定义颜色映射,可以生成色谱
from matplotlib.colors import ListedColormap

#朴素贝叶斯模块导入高斯朴素贝叶斯分类器
from sklearn.naive_bayes import GaussianNB
#从决策树模块导入决策树分类器
from sklearn.tree import DecisionTreeClassifier
```

```python
p = plt.rcParams
#p["font.sans-serif"] = ["Roboto"]
p["font.weight"] = "light"
p["ytick.minor.visible"] = True
p["xtick.minor.visible"] = True
p["axes.grid"] = True
p["grid.color"] = "0.5"
p["grid.linewidth"] = 0.5
```

```python
#这个色谱是一个离散色谱,按照索引给颜色,不是连续色谱映射
cmap_light = ListedColormap(['#FFAAAA','#AAFFAA','#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000','#00FF00','#0000FF'])

iris = datasets.load_iris()
X = iris.data[:,:2]
y = iris.target

x_min,x_max = 4,8
y_min,y_max = 2,5
xx,yy = np.meshgrid(np.linspace(x_min,x_max,1000),np.linspace(y_min,y_max,100))
```

```python
classifiers = [
    DecisionTreeClassifier(max_depth = 5),
    GaussianNB()
]

fig = plt.figure(figsize = (8,12))

for idx,classifier_idx in enumerate(classifiers):
    ax = fig.add_subplot(2,1,idx+1)

    classifier_idx.fit(X,y)

    Z = classifier_idx.predict(np.c_[xx.ravel(),yy.ravel()])

    Z = Z.reshape(xx.shape)

    plt.contourf(xx,yy,Z,cmap = cmap_light)

    #如果是离散色谱,色彩按照传入的c进行离散取值
    plt.scatter(X[:,0],X[:,1],c = y,cmap = cmap_bold)
    plt.xlabel('Sepal length (cm)')
    plt.ylabel('Sepal width (cm)')
    plt.axis('tight')

fig.savefig('Figures/决策树_朴素贝叶斯分类_contourf.svg')
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from sympy import lambdify, diff, exp, latex
from sympy.abc import x, y
from matplotlib import cm
```

```python
f_xy =  3*(1-x)**2*exp(-(x**2) - (y+1)**2)\
    - 10*(x/5 - x**3 - y**5)*exp(-x**2-y**2)\
    - 1/3*exp(-(x+1)**2 - y**2) 

f_xy_fcn = lambdify([x,y],f_xy)

f_xy
```

```python
def mesh(num = 101):
    x_array = np.linspace(-3,3,num)
    y_array = np.linspace(-3,3,num)
    xx,yy = np.meshgrid(x_array,y_array)

    return xx,yy
```

```python
xx,yy = mesh(num = 7)
ff = f_xy_fcn(xx,yy)
```

```python
fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

ax.plot_wireframe(xx,yy,ff,
                  color = '0.6',rstride = 1,cstride = 1,linewidth = 0.25)

ax.scatter(xx,yy,ff,marker = '.',color = 'k')

CS = ax.contour(xx,yy,ff,levels = np.linspace(-8,8,17),
                cmap = 'RdYlBu_r',linewidths = 1)

ax.set_proj_type('ortho')

ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

ax.set_xlabel(r'$\it{x_1}$')
ax.set_ylabel(r'$\it{x_2}$')
ax.set_zlabel(r'$\it{f}$($\it{x_1}$,$\it{x_2}$)')

ax.set_xlim(xx.min(), xx.max())
ax.set_ylim(yy.min(), yy.max())
ax.set_zlim(-8, 8)
ax.view_init(azim=-120, elev=30)
ax.grid(False)
fig.savefig('Figures/颗粒度过低，三维.svg', format='svg')
```

```python
fig = plt.figure(figsize = (8,12))

X_ = X+np.sin(Y*2)*0.2
Y_ = Y+np.cos(Y*2)*0.3

ax = fig.add_subplot(3,2,1)
ax.contour(np.abs(X_ - np.round(X_)),1,colors = 'k',linewidths = 0.25)
ax.contour(np.abs(Y_ - np.round(Y_)),1,colors = 'k',linewidths = 0.25)

ax.set_xticks([])
ax.set_yticks([])


def Z_sq(X,Y):
    #j是虚数单位
    Z = X + 1j * Y
    c = Z**2
    #real表示实数部分
    X_ = c.real
    #imag表示虚部部分
    Y_ = c.imag
    return X_,Y_
X_,Y_ = Z_sq(X,Y)

ax = fig.add_subplot(3,2,2)
ax.contour(np.abs(X_ - np.round(X_)),1,colors = 'k',linewidths = 0.25)
ax.contour(np.abs(Y_ - np.round(Y_)),1,colors = 'k',linewidths = 0.25)

ax.set_xticks([])
ax.set_yticks([])


def Z_sq(X,Y):
    Z = X + 1j * Y
    c = np.sqrt(Z**2 + 1)
    X_ = c.real
    Y_ = c.imag
    return X_,Y_
X_,Y_ = Z_sq(X,Y)

ax = fig.add_subplot(3,2,3)
ax.contour(np.abs(X_ - np.round(X_)),1,colors = 'k',linewidths = 0.25)
ax.contour(np.abs(Y_ - np.round(Y_)),1,colors = 'k',linewidths = 0.25)

ax.set_xticks([])
ax.set_yticks([])


def Z_cubic(X,Y):
    Z = X + 1j * Y
    c = Z**3
    X_ = c.real
    Y_ = c.imag
    return X_,Y_
X_,Y_ = Z_cubic(X,Y)

ax = fig.add_subplot(3,2,4)
ax.contour(np.abs(X_ - np.round(X_)),1,colors = 'k',linewidths = 0.25)
ax.contour(np.abs(Y_ - np.round(Y_)),1,colors = 'k',linewidths = 0.25)

ax.set_xticks([])
ax.set_yticks([])


def Z_fourth(X,Y):
    Z = X + 1j * Y
    c = Z**4
    X_ = c.real
    Y_ = c.imag
    return X_,Y_
X_,Y_ = Z_fourth(X,Y)

ax = fig.add_subplot(3,2,5)
ax.contour(np.abs(X_ - np.round(X_)),1,colors = 'k',linewidths = 0.25)
ax.contour(np.abs(Y_ - np.round(Y_)),1,colors = 'k',linewidths = 0.25)

ax.set_xticks([])
ax.set_yticks([])


def Z_exp(X,Y):
    Z = X + 1j * Y
    c = np.exp(Z)
    X_ = c.real
    Y_ = c.imag
    return X_,Y_
X_,Y_ = Z_exp(X,Y)

ax = fig.add_subplot(3,2,6)
ax.contour(np.abs(X_ - np.round(X_)),1,colors = 'k',linewidths = 0.25)
ax.contour(np.abs(Y_ - np.round(Y_)),1,colors = 'k',linewidths = 0.25)

ax.set_xticks([])
ax.set_yticks([])

fig.savefig('Figures/线性非线性变换,第2组.svg')
```

```python
import matplotlib.pyplot as plt
import matplotlib.tri as mtri
```

```python
from sympy import symbols
#这个是通过函数创建符号
#sympy.abc是从sympy.abc.x和sympy.abc.y导入的这是预先配置好的快速使用
x,y = symbols('x,y')
from sympy import lambdify,exp

f = x*exp(-x**2 - y**2)
f_fcn = lambdify([x,y],f)
```

```python
#asarray表示引用的意思,不一定每次都要创建
#如果里面是一个变量a且a是array格式那么将直接将a指向的内存赋值给外部变量,导致外部变量和a变量相等都指向一个值,不复制
#如果不是array将会创建新的
x_tri = np.asarray([-3.,-1.,1.,3.,-2.,0.,2.,-1.,1.,0.])
y_tri = np.asarray([-3.,-3.,-3.,-3,-1.,-1.,-1.,1.,1.,3.])

#生成10个点对应了坐标
```

```python
#c_[]是特殊语法糖 等用于np.column_stack([])
points = np.c_[x_tri,y_tri]
points
```

```python
from scipy.spatial import Delaunay
#直接剖析完成存在对象中,存储相关所有数据
tri_form_scipy = Delaunay(points)
```

```python
fig,ax = plt.subplots(figsize = (5,5))
#simplices是三角形剖分后三角形顶点索引,每一个小三角形都有对应点值,n_tri,3的大小
    #这个索引是关于前面输入参数的点的,所以需要一起配合
#画出来的是根据索引取出的点的连线,最后自动去除重复连线
ax.triplot(points[:,0],points[:,1],tri_form_scipy.simplices)

ax.plot(points[:,0],points[:,1],'.r',markersize = 10,lw = 0.25)
ax.set_aspect('equal')
ax.set_xlim(-3,3);ax.set_ylim(-3,3)
```

```python
#定义哪些点口称一个小三角形,逆时针顺序
triangles = [[0,1,4],[1,2,5],[2,3,6],[1,5,4],[2,6,5],[4,5,7],
             [5,6,8],[5,8,7],[7,8,9]]
#数据中也是三角形点数据和索引但是没有之前的生成的其他集合信息
#生成不规则三角形网格
triang = mtri.Triangulation(x_tri,y_tri,triangles)
```

```python
fig,ax = plt.subplots(figsize = (5,5))

#函数针对三角形绘图,可以自动读取triang对象解析
#自动按照z进行映射
#levels如果是整数,自动分成整数等份,如果是数组才按照数组划分等高线高度
ax.tricontourf(triang,f_fcn(x_tri,y_tri),cmap = 'RdYlBu_r')
#ko-中的k表示黑色
ax.triplot(triang,'ko-')
ax.set_aspect('equal')
ax.set_xlim(-3,3);ax.set_ylim(-3,3)
```

```python
def circle_points(num_r,num_n):
    r = np.linspace(0,3,num_r)
    n = r*num_n+1
    n = n.astype(int)

    circles = np.empty((0,2))

    for r_i,n_i in zip(r,n):
        t_i = np.linspace(0,2*np.pi,n_i,endpoint = False)
        r_i = np.ones_like(t_i)*r_i

        x_i = r_i*np.cos(t_i)
        y_i = r_i*np.sin(t_i)

        circle_i = np.column_stack([x_i,y_i])
        circles = np.append(circles,circle_i,axis = 0)

    return circles
```

```python
points_circles = circle_points(10,20)
triang_circles_auto = mtri.Triangulation(points_circles[:,0],points_circles[:,1])

fig,ax = plt.subplots(figsize = (5,5))

ax.tricontourf(triang_circles_auto,f_fcn(points_circles[:,0],points_circles[:,1]),
               cmap = 'RdYlBu_r',levels = 20)
ax.triplot(triang_circles_auto,'ko-')
ax.set_xlim(-3,3)
ax.set_ylim(-3,3)
ax.set_aspect('equal',adjustable = 'box')
```

```python
fig = plt.figure(figsize = (5,5))
ax = fig.add_subplot(projection = '3d')

surf = ax.plot_trisurf(points_circles[:,0],points_circles[:,1],
                       f_fcn(points_circles[:,0],points_circles[:,1]),
                       cmap = 'RdYlBu_r',linewidth = 0.2,shade = False)

surf.set_facecolor((0,0,0,0))
ax.set_proj_type('ortho')
ax.view_init(azim = -120,elev = 30)
ax.set_xlim(-3,3)
ax.set_ylim(-3,3)
ax.grid(False)
ax.set_box_aspect(aspect = (1,1,1))
```

```python
import matplotlib.tri as tri
```

```python
corners = np.array([[0,0],[1,0],[0.5,0.75**0.5]])

triangle = tri.Triangulation(corners[:,0],corners[:,1])
```

```python
from scipy.spatial import Delaunay

tri_from_scipy = Delaunay(corners)

fig,ax = plt.subplots(figsize = (5,5))
#simplices是单形,表示最简单的多面体,二维下就是三角形
ax.triplot(corners[:,0],corners[:,1],tri_from_scipy.simplices)
ax.plot(corners[:,0],corners[:,1],'.r',markersize = 10)

ax.set_aspect('equal')
ax.set_xlim(0,1);ax.set_ylim(0,1)
```

```python
refiner = tri.UniformTriRefiner(triangle)#
subdiv_array = [1,2,3,4]

fig = plt.figure(figsize = (12,3))

for idx,subdiv_idx in enumerate(subdiv_array,1):#
    trimesh_idx = refiner.refine_triangulation(subdiv = subdiv_idx)

    plt.subplot(1,4,idx)

    plt.triplot(trimesh_idx)
    plt.axis('off');plt.axis('equal')
    plt.title('Small triangles = ' + str(4**subdiv_idx))
```
