```python
a = np.array([[-2],
              [5]])
b = np.array([[5],
              [-1]])
```

```python
A = np.array([[1.25,-0.75],
              [-0.75,1.25]])

xx1,xx2 = np.meshgrid(np.linspace(-8,8,9),np.linspace(-8,8,9))
num_vecs = np.prod(xx1.shape)#prod计算各个维度元素数量乘积,得到元素总数

thetas = np.linspace(0,2*np.pi,num_vecs)

thetas = np.reshape(thetas,(-1,9))#将1d变为2d,限制了列数,行数自动计算,和thetas.reshape(xx1.shape)等价
thetas = np.flipud(thetas)#上下反转数据,行顺序反转,列不变

uu = np.cos(thetas)
vv = np.sin(thetas)
```

```python
V = np.array([uu.flatten(),vv.flatten()]).T
W = V@A

uu_new = np.reshape(W[:,0],(-1,9))
vv_new = np.reshape(W[:,1],(-1,9))
```

```python
fig,ax = plt.subplots(figsize = (6,6))

ax.quiver(xx1,xx2,
          uu,vv,
          angles = 'xy',scale_units = 'xy',
          scale = 0.8,#小于1为放大  向量模长/scale.这里是放大
          width = 0.0025,
          edgecolor = 'none',facecolor = 'b')#color是总颜色(填充+边缘),facecolor是作用在填充部分颜色

ax.quiver(xx1,xx2,
          uu_new,vv_new,
          angles = 'xy',scale_units = 'xy',
          scale = 0.8,
          width = 0.0025,
          edgecolor = 'none',facecolor = 'r')

plt.axis('scaled')#强制xy方向单位长度相同 等价于ax.set_aspect('equal', adjustable='box')
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
ax.grid(ls = '-',lw = 0.25,color = '0.5')
ax.set_xticks(np.linspace(-10,10,11))
ax.set_yticks(np.linspace(-10,10,11))

ax.set_xticklabels([])
ax.set_yticklabels([])
```

```python
fig,ax = plt.subplots(figsize = (6,6))

import matplotlib
cm = matplotlib.cm.rainbow
#这里直接创建完成了一个归一化器,后续调用的最大最小映射值按照默认值配置.默认没有限制最大值和最小值
norm = matplotlib.colors.Normalize()

ax.quiver(xx1*0,xx2*0,uu,vv,
          angles = 'xy',scale_units = 'xy',scale = 1,
          width = 0.0025,edgecolor = 'none',
          facecolor = cm(norm(range(len(xx1.ravel())))))

ax.quiver(xx1*0,xx2*0,uu_new,vv_new,
          angles = 'xy',scale_units = 'xy',scale = 1,
          width = 0.0025,edgecolor = 'none',
          #这个类的实例可以直接使用()是因为类配置了__call__可以直接用()调用
          #实例创建时候没最大最小值(默认的)这种情况会按照给实例传入的最大最小值进行归一化
          facecolor = cm(norm(range(len(xx1.ravel())))))

plt.axis('scaled')
ax.set_xlim([-2,2])
ax.set_ylim([-2,2])
ax.grid(ls = '-',lw = 0.25,color = '0.5')
ax.set_xticks(np.linspace(-2,2,5))
ax.set_yticks(np.linspace(-2,2,5))

ax.set_xticklabels([])
ax.set_yticklabels([])
```

```python
import numpy as np
import matplotlib.pyplot as plt
#模块分别是创建符号变量,符号表达式简化,展开符号表达式,符号表达式变为函数,符号求导
from sympy import symbols,simplify,expand,lambdify,diff
```

```python
def fcn_n_grdnt(A,xxx1,xxx2,xxx3):
    x1,x2,x3 = symbols('x1 x2 x3')
    x = np.array([[x1,x2,x3]]).T

    f_x = x.T@A@x
    f_x = f_x[0][0]
    print(simplify(expand(f_x)))

    grad_f = [diff(f_x,var) for var in (x1,x2,x3)]

    f_x_fcn = lambdify([x1,x2,x3],grad_f)
    ff_x = f_x_fcn(xxx1,xxx2,xxx3)

    grad_fcn = lambdify([x1,x2,x3],grad_f)

    V = grad_fcn(xxx1,xxx2,xxx3)

    if isinstance(V[0],int):
        V[0] = np.zeros_like(xxx1)

    if isinstance(V[1],int):
        V[0] = np.zeros_like(xxx1)

    if isinstance(V[2],int):
        V[0] = np.zeros_like(xxx1)

    return ff_x,V
```

```python
x = np.arange(-3,3,0.3)
y = np.arange(-3,3,0.3)
xx,yy = np.meshgrid(x,y)

Fx = np.cos(xx + 2 * yy)
Fy = np.sin(xx - 2 * yy)

color_array = np.sqrt(Fx**2 + Fy**2)
```

```python
fig = plt.figure(figsize = (6,6))

ax = fig.add_subplot(111)

ax.streamplot(xx,yy,Fx,Fy,density = 2,
              arrowstyle = 'fancy')
ax.set_aspect('equal')
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```
