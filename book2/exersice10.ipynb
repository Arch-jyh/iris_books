```python
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as mtri
```

```python
x_array = np.linspace(-3,3,20)
y_array = np.ones_like(x_array)

fig,ax = plt.subplots(figsize = (4,1))
ax.scatter(x_array,y_array,s = 5)
ax.plot(x_array,y_array,color = [0.5,0.5,0.5],linewidth = 0.25)
ax.axis('off')

fig.savefig('Figures/一维,沿着横轴.svg')
```

```python
fig,ax = plt.subplots(figsize = (1,4))
ax.scatter(y_array,x_array,s = 5)
ax.plot(y_array,x_array,
        color = '0.5',lw = 0.25)
ax.axis('off')
fig.savefig('Figures/一维纵轴.svg')
```

```python
xx1_square,xx2_square = np.meshgrid(np.linspace(-3,3,20),np.linspace(-3,3,20))

fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

ax.plot_wireframe(xx1_square,xx2_square,xx1_square*0,
                 color = '0.5',lw = 0.25)
ax.scatter(xx1_square,xx2_square,xx1_square*0,s =5)
ax.set_proj_type('ortho')

#\it是是斜体
ax.set_xlabel(r'$\it{x_1}$')
ax.set_ylabel(r'$\it{x_2}$')
ax.set_zlabel('')
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

ax.set_xlim(xx1_square.min(),xx1_square.max())
ax.set_ylim(xx2_square.min(),xx2_square.max())
ax.set_box_aspect([1,1,1])
ax.view_init(azim = 90,elev = 90)
ax.grid(False)

fig.savefig('Figures/二维,平面.svg')
```

```python
xxx1_square,xxx2_square,xxx_square = np.meshgrid(np.linspace(-3,3,13),
                                                 np.linspace(-3,3,13),
                                                 np.linspace(-3,3,13))

fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

ax.scatter(xxx1_square,xxx2_square,xxx_square,s = 5)
ax.set_proj_type('ortho')

#\it是是斜体
ax.set_xlabel(r'$\it{x_1}$')
ax.set_ylabel(r'$\it{x_2}$')
ax.set_zlabel(r'$\it{x_3}$')
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

ax.set_xlim(xx1_square.min(),xx1_square.max())
ax.set_ylim(xx2_square.min(),xx2_square.max())
ax.set_box_aspect([1,1,1])
ax.view_init(azim = -120,elev = 30)
ax.grid(False)

fig.savefig('Figures/三维.svg')
```

```python
#符号函数
from sympy import symbols
x,y = symbols('x,y')
from sympy import lambdify,exp

f = x*exp(-x**2 - y**2)
#创建函数不检查变量是否与函数匹配所以要检查好
#def也是将函数复制给变量,所以只要变量指向的是函数,使用()就是表示调用
f_fcn = lambdify([x,y],f)
```

```python
fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

ax.plot_wireframe(xx1_square,xx2_square,f_fcn(xx1_square,xx2_square),
                  color = '0.5',lw = 0.25)

#传入给c的形状和点数一致才会做归一化
ax.scatter(xx1_square,xx2_square,f_fcn(xx1_square,xx2_square),s = 5,
           c = f_fcn(xx1_square,xx2_square),cmap = 'RdYlBu_r')

ax.set_proj_type('ortho')

ax.set_xlabel(r'$\it{x_1}$')
ax.set_ylabel(r'$\it{x_2}$')
ax.set_zlabel(r'$\it{x_3}$')
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

ax.set_xlim(xx1_square.min(),xx1_square.max())
ax.set_ylim(xx2_square.min(),xx2_square.max())
ax.set_box_aspect([1,1,1])
ax.view_init(azim = -120,elev = 30)
ax.grid(False)

fig.savefig('Figures/网格状散图,函数.svg')
```

```python
theta = np.linspace(0,2*np.pi,20)
r = np.linspace(0,3,10)
tt,rr = np.meshgrid(theta,r)
xx1_polar = np.cos(tt)*rr
xx2_polar = np.sin(tt)*rr
```

```python
fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

ax.plot_wireframe(xx1_polar,xx2_polar,xx2_polar*0,
                 color = '0.5',lw = 0.25)
ax.scatter(xx1_polar,xx2_polar,xx2_polar*0,s = 5)

ax.set_proj_type('ortho')

ax.set_xlabel(r'$\it{x_1}$')
ax.set_ylabel(r'$\it{x_2}$')
ax.set_zlabel(r'$\it{x_3}$')
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

ax.set_xlim(xx1_polar.min(),xx1_polar.max())
ax.set_ylim(xx2_polar.min(),xx2_polar.max())
ax.set_box_aspect([1,1,1])
ax.view_init(azim = 90,elev = 90)
ax.grid(False)

fig.savefig('Figures/极坐标网格,平面.svg')
```
