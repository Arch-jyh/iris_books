```python
arr_10_1=np.random.rand(10)
arr_10_2=np.random.rand(10)
```

```python
arr_8_by_5=np.random.rand(40).reshape((8,5))
arr_8_by_5
```

```python
arr_3_by_5 = np.random.rand(15).reshape((3,5))
arr_3_by_5
```

```python
np.stack((arr_8_by_5,arr_3_by_5),axis=0) #两个数组必须形状相同
```

```python
np.block([[arr_8_by_5],[arr_3_by_5]])
```

```python
np.concatenate((arr_8_by_5,arr_3_by_5),axis=0)
```
