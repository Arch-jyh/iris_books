```python
from sympy import symbols,exp,lambdify
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1,x2 = symbols('x1 x2')
f_gaussian_x1x2=exp(-x1**2-x2**2)
f_gaussian_x1x2_fcn = lambdify([x1,x2],f_gaussian_x1x2) #将符号表达式转换为可计算的python函数
xx1,xx2 = np.meshgrid(np.linspace(-3,3,201),np.linspace(-3,3,201))
```

```python
ff=f_gaussian_x1x2_fcn(xx1,xx2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

ax.plot_wireframe(xx1,xx2,ff,
                  rstride=10,cstride=10)

ax.set_proj_type('ortho')
ax.view_init(azim = -120,elev=30)
ax.grid(False)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x1,x2)')
ax.set_xlim(-3,3)
ax.set_ylim(-3,3)
ax.set_zlim(0,1)
ax.set_box_aspect(aspect=(1,1,1)) #配置比例
fig.savefig('二元高斯函数.svg',format='svg')
```

```python
import numpy as np
import matplotlib.pyplot as plt
#继续导入符号变量, 表达式转换为可计算函数 , 表达式展开  ,  自动简化表达式
from sympy import symbols,lambdify,expand,simplify

#三个输入,两个表示坐标,然后有个矩阵数值
def visualize(xx1,xx2,f2_array):
    fig = plt.figure(figsize=(6,3))#创建对象,配置显示尺寸
    ax_3D = fig.add_subplot(1,2,1,projection='3d')#fig添加3d子图,划分为1行2列子图,并添加到第一个中为3d图
    #绘制为3d网格图
    ax_3D.plot_wireframe(xx1,xx2,f2_array,
                        rstride = 10,cstride=10,#行和列的步幅,控制网格间隔
                        color = [0.8,0.8,0.8],#线框为浅灰色
                        linewidth = 0.25)#绘制的线框的线的宽度
    ax_3D.contour(xx1,xx2,f2_array,levels=12,cmap = 'RdYlBu_r') #增加色彩变化的等高线

    ax_3D.set_xlabel('$x_1$');ax_3D.set_ylabel('$x_2$')
    ax_3D.set_zlabel('$f(x_1,x_2)$')
    ax_3D.set_proj_type('ortho')
    #配置坐标可度标签为不显示
    ax_3D.set_xticks([]);ax_3D.set_yticks([])
    ax_3D.set_zticks([])
    #配置角度,azim为方位角绕z轴旋转,逆时针为正,后者为仰角向上为正
    ax_3D.view_init(azim=-120,elev=30)
    ax_3D.grid(False)
    ax_3D.set_xlim(xx1.min(),xx1.max());
    ax_3D.set_ylim(xx2.min(),xx2.max())

    #右子图平面等高线
    ax_2D = fig.add_subplot(1,2,2)
    ax_2D.contour(xx1,xx2,f2_array,levels=12,cmap='RdYlBu_r')
    ax_2D.set_xlabel('$x_1$');ax_2D.set_ylabel('$x_2$')
    ax_2D.set_xticks([]);ax_2D.set_yticks([])
    ax_2D.set_aspect('equal');ax_2D.grid(False)
    ax_2D.set_xlim(xx1.min(),xx1.max());
    ax_2D.set_ylim(xx2.min(),xx2.max())
    plt.tight_layout()
    
```

```python
x1_array = np.linspace(-2,2,201)
x2_array = np.linspace(-2,2,201)

xx1,xx2 = np.meshgrid(x1_array,x2_array)

#定义二元函数
def fcn(A,xx1,xx2):
    #定义符号变量
    x1,x2 = symbols('x1 x2')
    #构造了列向量,这里还与外部x1x2无关,仅仅是符号
    x = np.array([[x1,x2]]).T
    #运算结果为二维数组但是大小为1x1 因为@是矩阵乘法,保留矩阵形式
    f_x = x.T@A@x
    #提取符号表达式,将其中的数据提取出来,就是表达式
    f_x = f_x[0][0]
    print(simplify(expand(f_x)))
    #转化为python函数 列表中表示符号变量,表示输入变量,后者f_x表示符号本身
    f_x_fcn = lambdify([x1,x2],f_x)
    ff_x = f_x_fcn(xx1,xx2)
    return ff_x
```

```python
A = np.array([[0,1],
              [1,0]])
```

```python
f2_array = fcn(A,xx1,xx2)
```

```python
visualize(xx1,xx2,f2_array)
```
