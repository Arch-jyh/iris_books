```python
x1,x2 = symbols('x1 x2')
```

```python
f4 = x1+x2
f4
```

```python
f5 = (x1-x2)**5
f5
```

```python
f5_expanded = expand(f5)
f5_expanded
```

```python
from sympy import Poly
f5_expanded_x1 = Poly(f5_expanded,x1) #将表达式转为多项式,将整个表达式当做关于x1的多项式
f5_expanded_x1.coeffs() #提取多项式的系数列表,仅仅只对x1这个多项式展开
```

```python
f5_expanded_x2 = Poly(f5_expanded,x2)
f5_expanded_x2.coeffs()
```

```python
f5_expanded_x1x2 = Poly(f5_expanded,[x1,x2])
coefficients=f5_expanded_x1x2.coeffs()
coefficients
```

```python
type(coefficients)
```

```python
type(coefficients[2]) #如果结果为整数1直接写One不是Integer
```

```python
#转化为浮点
coefficients_float = eval(str(coefficients))
#先转化为字符串然后为浮点 都使用python自带内置函数  str能够正常处理sympy对象是因为sympy对象自带str的方法,告诉str如何转换为str
#eval执行合法表达式(多简单都行)返回列表,如果为√2则s输出为sqrt(2) 整数,浮点数输出浮点数
```

```python
import matplotlib.pyplot as plt
plt.stem(coefficients_float)
```

```python
f_gaussian_x1x2 = exp(-x1**2 - x2**2)
f_gaussian_x1x2
```

```python
f_gaussian_x1x2_fcn = lambdify([x1,x2],f_gaussian_x1x2)
f_gaussian_x1x2_fcn
```

```python
xx1,xx2 = np.meshgrid(np.linspace(-3,3,100),np.linspace(-3,3,100))
```

```python
ff = f_gaussian_x1x2_fcn(xx1,xx2)
plt.contourf(xx1,xx2,ff,levels = 20 , cmap = 'RdYlBu_r')
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot_wireframe(xx1,xx2,ff)
```
