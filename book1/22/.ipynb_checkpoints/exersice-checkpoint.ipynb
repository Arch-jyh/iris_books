```python
import pandas as pd

df1=pd.DataFrame({'X1':[1,2,3],
                  'X2':['X','Y','Z']},
                 index=[0,1,2])

df2=pd.DataFrame({'X3':['A','B','C'],
                  'X4':['4','5','6']},
                 index=[1,2,3])

df_outer = pd.concat([df1,df2],join='outer',axis=1)
df_inner = pd.concat([df1,df2],join='inner',axis=1)
```

```python
import pandas as pd
left_data = {
    'M':[1,2,3],
    'X':['a','b','c']}
left_df = pd.DataFrame(left_data)
left_df
```

```python
right_data = {
    'X':['b','c','d'],
    'N':[22,33,44]}
right_df=pd.DataFrame(right_data)
right_df
```

```python
left_exl = left_df.merge(right_df,on = 'X',how = 'left',
                         indicator = True)
left_exl
```

```python
left_exl = left_exl[left_exl['_merge'] == 'left_only'].drop(columns=['_merge'])
left_exl
```

```python
right_exl = left_df.merge(right_df,on='X',how='right',indicator=True)
right_exl
```

```python
right_exl = right_exl[right_exl['_merge'] == 'right_only'].drop(columns=['_merge'])
right_exl
```

```python
outer_exl = left_df.merge(right_df,on='X',how='outer',indicator=True)
outer_exl
```

```python
outer_exl = outer_exl[outer_exl['_merge'] != 'both'].drop(columns = ['_merge'])
outer_exl
```

```python
import pandas as pd
data = {
    'Student ID':['1','1','2','2','3','3','4','4'],
    'Subject':['Math','Art','Science','Art',
               'Math','Science','Art','Math'],
    'Midterm':[4,5,3,5,4,5,3,5],
    'Final':[3,4,5,3,4,4,4,5]}

df = pd.DataFrame(data)
df
```

```python
df.pivot(index='Student ID',
         columns='Subject',
         values=['Midterm','Final'])
```

```python
import pandas as pd
import numpy as np

student_ids = [1,2,3,4]
subjects = ['Art','Math','Science']
np.random.seed(0)
scores =  np.random.randint(3,6,size=(len(student_ids),len(subjects)))

df = pd.DataFrame(scores,index=student_ids,columns=subjects)
df
```

```python
df.columns.names = ['Subjects']
df.index.names = ['Students ID']

df
```

```python
df.stack().reset_index().rename(columns={0:'Final'})
#stack,将宽>>>长
#reset_index将索引转换为列
#rename重新命名数值列 0代表旧列名,值'final'为新列列名,因为前面的变换,仅剩0了
```
