```python
import pandas as pd
left_data = {
    'M':[1,2,3],
    'X':['a','b','c']}
left_df = pd.DataFrame(left_data)
left_df
```

```python
right_data = {
    'X':['b','c','d'],
    'N':[22,33,44]}
right_df=pd.DataFrame(right_data)
right_df
```

```python
left_exl = left_df.merge(right_df,on = 'X',how = 'left',
                         indicator = True)
left_exl
```

```python
left_exl = left_exl[left_exl['_merge'] == 'left_only'].drop(columns=['_merge'])
left_exl
```

```python
right_exl = left_df.merge(right_df,on='X',how='right',indicator=True)
right_exl
```

```python
right_exl = right_exl[right_exl['_merge'] == 'right_only'].drop(columns=['_merge'])
right_exl
```

```python
outer_exl = left_df.merge(right_df,on='X',how='outer',indicator=True)
outer_exl
```

```python
outer_exl = outer_exl[outer_exl['_merge'] != 'both'].drop(columns = ['_merge'])
outer_exl
```

```python
def map_fnc(sepal_length):
    if sepal_length<5:
        return 'D'
    elif 5<= sepal_length <6:
        return 'C'
    elif 6<= sepal_length <7:
        return 'B'
    else:
        'A'
```

```python
iris_df['ctg'] = iris_df['sepal_length'].apply(map_fnc)
iris_df
```

```python
import seaborn as sns
import pandas as pd
```

```python
iris_df = sns.load_dataset("iris")
iris_df
```

```python
mean_X2_by_species = iris_df.groupby(
    'species')['petal_width'].mean()
mean_X2_by_species
```

```python
def map_petal_width(petal_width,species):
    if petal_width > mean_X2_by_species[species]:
        return "YES"
    else:
        return "NO"
```

```python
iris_df['greater_than_mean'] = iris_df.apply(lambda
        row: map_petal_width(row['petal_width'],
                             row['species']),axis=1)
iris_df
```
