```python
import seaborn as sns
import pandas as pd

iris_df = sns.load_dataset("iris")

X_df = iris_df.copy()
X_df.rename(columns = {'sepal_length':'X1',
                         'sepal_width':'X2'},
              inplace = True)
X_df_ = X_df[['X1','X2','species']]

#数据运算并添加
X_df_['X1-E(X1)']=X_df_['X1'] - X_df_['X1'].mean()
X_df_['X2-E(X2)']=X_df_['X2'] - X_df_['X2'].mean()
X_df_['X1+X2']=X_df_['X1'] + X_df_['X2']
X_df_['X1-X2']=X_df_['X1'] - X_df_['X2']
X_df_['X1*X2']=X_df_['X1'] * X_df_['X2']
X_df_['X1/X2']=X_df_['X1'] / X_df_['X2']

X_df_.drop(['X1','X2'],axis=1,inplace=True)

sns.pairplot(X_df_,corner=True,hue='species')
```

```python
import pandas as pd
import requests
import io

url = 'https://fred.stlouisfed.org/data/SP500.txt'

response = requests.get(url)

if response.status_code == 200:  
    # 读取数据，跳过前44行，自动适配空格分隔符，并跳过格式错误的行
    df = pd.read_csv(io.StringIO(response.text), skiprows=44, sep=r'\s+', engine='python', on_bad_lines='skip')
    
    # 处理日期列并设置索引
    df['DATE'] = pd.to_datetime(df['DATE'])
    df.set_index('DATE', inplace=True)

    print(df.head())  # 打印前5行数据
else:
    print("Failed to fetch data from the source")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#第三方库获取数据
import pandas_datareader as pdr
import seaborn as sns

#下载数据
#默认将时间格式作为索引
df = pdr.data.DataReader(['DGS6MO','DGS1',
                          'DGS2','DGS5',
                          'DGS7','DGS10',
                          'DGS20','DGS30',],
                         data_source='fred',
                         start='01-01-2022',
                         end='12-31-2022')
#删除数据中缺失的值 NaN
df=df.dropna()
#修改列标签
df = df.rename(columns={'DGS6MO':'0.5 yr',
                        'DGS1': '1yr',
                        'DGS2': '2yr',
                        'DGS3': '3yr',
                        'DGS5': '5yr',
                        'DGS7': '7yr',
                        'DGS10':'10yr',
                        'DGS20':'20yr',
                        'DGS30':'30yr'})
```

```python
df.plot(#配置标签
        xlabel='Time',ylabel='IR level',
        #显示图例 这个单词具有传说/解释的意思,这里代表显示的图例
        legend=True,
        xlim=(df.index.min(),df.index.max()))

plt.savefig("利率走势线图.svg")

#绘制利率走势线图,子图布置
df.plot(subplots=True,layout=(2,4),
        sharex=True,sharey=True,
        #不显示标签
        xticks = [],yticks= [],
        xlim = (df.index.min(),df.index.max()))
plt.savefig("利率走势线子图.svg")
```
