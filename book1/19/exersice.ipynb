```python
#list_to_DataFrame
import pandas as pd

list_fruits = [['appel',11],
               ['banana',22],
               ['cherry',33],
               ['durian',44]]

df_list1 = pd.DataFrame(list_fruits)

#设置行索引
df_list1.set_axis(['a','b','c','d'],axis='index')
#配置列标签
df_list1.set_axis(['fruit','number'],axis='columns')
df_list2=pd.DataFrame(list_fruits,
                      columns=['Fruit','number'],
                      index=['a','b','c','d'])
```

```python
import pandas as pd
import numpy as np
#独立的按照正太分布抽取的随机数,并配置成10,4的形状的矩阵
np_array=np.random.normal(size=(10,4))

df_np = pd.DataFrame(np_array,
                      columns=['X1','X2','X3','X4'])

data=[]
#for循环生成列表
for idx in range(10):
    data_idx = np.random.normal(size=(1,4)).tolist()  #tolist()从numpy.array转换为python的列表
    data.append(data_idx[0]) #list的更快因为预先留了内存,直接添加会更快.

df_loop = pd.DataFrame(data,
                       columns = ['X1','X2','X3','X4'])
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris_df=sns.load_dataset("iris")

fig,ax=plt.subplots(figsize=(5,9))
sns.heatmap(iris_df.iloc[:,0:4],
            cmap='RdYlBu_r',
            ax=ax,
            vmax=0,vmin=8,
            cbar_kws={'orientation':'vertical'},
            annot=False)
fig.savefig('iris_DataFrame.svg',format='svg')
```
