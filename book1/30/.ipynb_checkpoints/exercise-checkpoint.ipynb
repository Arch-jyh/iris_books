```python
import numpy as np
import matplotlib.pyplot as plt
#创建训练使用回归模型
from sklearn.linear_model import LinearRegression
```

```python
num = 50
np.random.seed(0)
x_data = np.random.uniform(0,10,num)
y_data = 0.5 * x_data + 1 + np.random.normal(0,1,num)
```

```python
x_data = x_data.reshape((-1,1))
#调整为列向量 -1表示自动推断行数量 1表示只有一列
```

```python
data = np.column_stack([x_data,y_data])
#因为堆叠必须为二维,所以x_data将转换为二维方便y_data的这种一维数据堆叠
```

```python
#创建回归对象并拟合
#面相对象设计语言,先创建对象,再调用方法,拟合后存取数据可以进行预测查看等
#与其他对象不冲突
#逻辑为先配置,再训练
LR = LinearRegression()
LR.fit(x_data,y_data)
```

```python
slope = LR.coef_ #返回全部的斜率,一组X_data输出一个斜率
intercept = LR.intercept_
```

```python
x_array = np.linspace(0,10,101).reshape((-1,1))
predicted = LR.predict(x_array)
```

```python
data_ = np.column_stack([x_data,LR.predict(x_data)])
```

```python
fig,ax = plt.subplots()
ax.scatter(x_data,y_data)
ax.scatter(x_data,LR.predict(x_data),color = 'k',marker = 'x')
#用直线连接相近的点
ax.plot(x_array,predicted,color='r')
ax.plot(([i for (i,j) in data_] , [i for (i,j) in data]),
        ([j for (i,j) in data_] , [j for (i,j) in data]),
        #这里plot知道给的x和y下面分别有两个数组,知道绘制线段,每一个元组的两列,第一列为起点,第二列为终点,一次x一次y就变成了对应点
        c = [0.6,0.6,0.6],alpha = 0.5)
ax.set_xlabel('x');ax.set_ylabel('y')
ax.set_aspect('equal',adjustable = 'box')
#equal表示xy单位相等
ax.set_xlim(0,10);ax.set_ylim(-2,8)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
n_samples = 100
X = np.random.randn(n_samples,2)
#标准正太分布 生成两组X  x1,x2  
y = -3 * X[:,0] + 2*X[:,1] +1 +0.5*np.random.randn(n_samples)
```

```python
LR = LinearRegression()
y_predicted = LR.fit(X,y)
```

```python
slope = LR.coef_
intercept = LR.intercept_
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.scatter(X[:,0],X[:,1],y)

#生成回归平面的数据点 两组,分别为10*10
x1_grid,x2_grid = np.meshgrid(np.linspace(-3,3,10),
                              np.linspace(-3,3,10))
#flatten将10*10的数组拉成一维,然后堆叠成对应点的坐标
#参数要求为可迭代对象,所以将两个数组堆叠 相当于将两个数组变成(100,2)的大小
X_grid = np.column_stack((x1_grid.flatten(),x2_grid.flatten()))

#将坐标传递给已训练的回归模型LR 这里输出的形状为100,1
y_pred = LR.predict(X_grid)
#将调整形状与数组形状相同 变为10,10
y_pred = y_pred.reshape(x1_grid.shape)

#绘制回归平面,线框格式
ax.plot_wireframe(x1_grid,x2_grid,y_pred)

ax.set_xlabel('$x_1$');ax.set_ylabel('$x_2$')
ax.set_zlabel('y')

ax.set_xlim([-3,3]);ax.set_ylim([-3,3])
#平行投影 z轴逆时针旋转120度,垂直抬高30°
ax.set_proj_type('ortho');ax.view_init(azim = -120,elev=30)
```

```python
import numpy as np
import matplotlib.pyplot as plt
#因为原始特征x不能表达数据关系,引入多项式特征,是原始特征的次幂组合 这里的类可以配置特定阶数
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
num = 30
#uniform生成均匀分布随机数,不包括右端点 数量为30不是从0-30
X = np.random.uniform(0,4,num)
#randn生成均值0 标准差1的随机数
y = np.sin(0.4*np.pi*X) + 0.4*np.random.randn(num)
data = np.column_stack([X,y])
```

```python
#线性均匀生成0-4的数组闭区间,然后转换为n行1列
x_array = np.linspace(0,4,101).reshape(-1,1)
#指定的多项式回归阶数 故意跳过了5 6 (绘制6副图)
degree_array = [1,2,3,4,7,8]
```

```python
fig,axes = plt.subplots(3,2,figsize = (10,20))
axes = axes.flatten()
for ax,degree_idx in zip(axes,degree_array):

    #配置多项式回归的最高阶数
    poly = PolynomialFeatures(degree = degree_idx)
    #将X转换为2维数组,因为前面方法接受二维数组
    #根据之前配置的poly的特征,将X转换为多项式 行为样本,列为特征
    #算是处理输入数据列成表格方便拟合
    X_poly = poly.fit_transform(X.reshape(-1,1))

    #训练回归模型
    poly_reg = LinearRegression()
    #输入二维特征矩阵,计算每个特征的系数
    poly_reg.fit(X_poly,y)
    #数据预测
    y_poly_pred = poly_reg.predict(X_poly)
    #在拟合曲线上的点
    data_ = np.column_stack([X,y_poly_pred])

    #特征转换后进行预测
    #特征转换是将原模型数据转换为适合模型的形式,这里特征转换从原来的二维数组变成了符合模型特征的二维数组
    y_array_pred = poly_reg.predict(poly.fit_transform(x_array))

    #绘制散点图
    ax.scatter(X,y,s = 20)
    ax.scatter(X,y_poly_pred,marker = 'x',color='k')

    ax.plot(x_array,y_array_pred,color = 'r')
    ax.set_title('Degree = %d' % degree_idx)

    #提取参数
    #提取系数
    coef = poly_reg.coef_
    #提取截距
    intercept = poly_reg.intercept_
    
    #回归解析式
    #将截距插入字符串 左边配置了保留一位小数 使用方法format将数值插入到对应的位置
    equation = '$y = {:.1f}'.format(intercept)
    for j in range(1,len(coef)):
        #这里有两个插入值,前一个为保留一位小数的浮点数,后者为整数
        equation += ' + {:.1f}x^{}'.format(coef[j],j)
    equation += '$'
    #将+ - 替换为-
    equation = equation.replace("+ -","-")
    ax.text(0.05,-1.8,equation)
    ax.set_aspect('equal',adjustable = 'box')
    ax.set_xlim(0,4)
    ax.grid(False)
    ax.set_ylim(-2,2)
        
```
