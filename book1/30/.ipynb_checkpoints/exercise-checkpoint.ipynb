```python
import numpy as np
import matplotlib.pyplot as plt
#创建训练使用回归模型
from sklearn.linear_model import LinearRegression
```

```python
# 生成随机数据
num = 50
np.random.seed(0)
x_data = np.random.uniform(0,10,num)
y_data = 0.5 * x_data + 1 + np.random.normal(0, 1, num)
```

```python
x_data = x_data.reshape((-1, 1))
# 将x调整为列向量
```

```python
data = np.column_stack([x_data,y_data])
```

```python
#创建回归对象并拟合
#面相对象设计语言,先创建对象,再调用方法,拟合后存取数据可以进行预测查看等
#与其他对象不冲突
#逻辑为先配置,再训练
LR = LinearRegression()
LR.fit(x_data, y_data)
```

```python
slope = LR.coef_ #返回全部的斜率,一组X_data输出一个斜率
intercept = LR.intercept_
```

```python
x_array = np.linspace(0,10,101).reshape((-1,1))
predicted = LR.predict(x_array)
```

```python
x_array = np.linspace(0,10,101).reshape((-1, 1))
# 预测
predicted = LR.predict(x_array)
```

```python
data_ = np.column_stack([x_data,LR.predict(x_data)])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_data, y_data)
ax.scatter(x_data, LR.predict(x_data), 
            color = 'k', marker = 'x')
ax.plot(x_array, predicted, 
        color = 'r')
ax.plot(([i for (i,j) in data_], [i for (i,j) in data]),
         ([j for (i,j) in data_], [j for (i,j) in data]),
        #这里plot知道给的x和y下面分别有两个数组,知道绘制线段,每一个元组的两列,第一列为起点,第二列为终点,一次x一次y就变成了对应点
        c = [0.6,0.6,0.6],alpha = 0.5)
ax.set_xlabel('x');ax.set_ylabel('y')
ax.set_aspect('equal',adjustable = 'box')
#equal表示xy单位相等
ax.set_xlim(0,10); ax.set_ylim(-2,8)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
n_samples = 100
X = np.random.randn(n_samples,2)
#标准正太分布 生成两组X  x1,x2  
y = -3 * X[:,0] + 2*X[:,1] +1 +0.5*np.random.randn(n_samples)
```

```python
LR = LinearRegression()
y_predicted = LR.fit(X,y)
```

```python
slope = LR.coef_
intercept = LR.intercept_
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.scatter(X[:,0],X[:,1],y)

#生成回归平面的数据点 两组,分别为10*10
x1_grid,x2_grid = np.meshgrid(np.linspace(-3,3,10),
                              np.linspace(-3,3,10))
#flatten将10*10的数组拉成一维,然后堆叠成对应点的坐标
#参数要求为可迭代对象,所以将两个数组堆叠 相当于将两个数组变成(100,2)的大小
X_grid = np.column_stack((x1_grid.flatten(),x2_grid.flatten()))

#将坐标传递给已训练的回归模型LR 这里输出的形状为100,1
y_pred = LR.predict(X_grid)
#将调整形状与数组形状相同 变为10,10
y_pred = y_pred.reshape(x1_grid.shape)

#绘制回归平面,线框格式
ax.plot_wireframe(x1_grid,x2_grid,y_pred)

ax.set_xlabel('$x_1$');ax.set_ylabel('$x_2$')
ax.set_zlabel('y')

ax.set_xlim([-3,3]);ax.set_ylim([-3,3])
#平行投影 z轴逆时针旋转120度,垂直抬高30°
ax.set_proj_type('ortho');ax.view_init(azim = -120,elev=30)
```

```python
import numpy as np
import matplotlib.pyplot as plt
#因为原始特征x不能表达数据关系,引入多项式特征,是原始特征的次幂组合 这里的类可以配置特定阶数
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
num = 30
#uniform生成均匀分布随机数,不包括右端点 数量为30不是从0-30
X = np.random.uniform(0,4,num)
#randn生成均值0 标准差1的随机数
y = np.sin(0.4*np.pi*X) + 0.4*np.random.randn(num)
data = np.column_stack([X,y])
```

```python
#线性均匀生成0-4的数组闭区间,然后转换为n行1列
x_array = np.linspace(0,4,101).reshape(-1,1)
#指定的多项式回归阶数 故意跳过了5 6 (绘制6副图)
degree_array = [1,2,3,4,7,8]
```

```python
fig,axes = plt.subplots(3,2,figsize = (10,20))
axes = axes.flatten()
for ax,degree_idx in zip(axes,degree_array):

    #配置多项式回归的最高阶数
    poly = PolynomialFeatures(degree = degree_idx)
    #将X转换为2维数组,因为前面方法接受二维数组
    #根据之前配置的poly的特征,将X转换为多项式 行为样本,列为特征
    #算是处理输入数据列成表格方便拟合
    X_poly = poly.fit_transform(X.reshape(-1,1))

    #训练回归模型
    poly_reg = LinearRegression()
    #输入二维特征矩阵,计算每个特征的系数
    poly_reg.fit(X_poly,y)
    #数据预测
    y_poly_pred = poly_reg.predict(X_poly)
    #在拟合曲线上的点
    data_ = np.column_stack([X,y_poly_pred])

    #特征转换后进行预测
    #特征转换是将原模型数据转换为适合模型的形式,这里特征转换从原来的二维数组变成了符合模型特征的二维数组
    y_array_pred = poly_reg.predict(poly.fit_transform(x_array))

    #绘制散点图
    ax.scatter(X,y,s = 20)
    ax.scatter(X,y_poly_pred,marker = 'x',color='k')

    ax.plot(x_array,y_array_pred,color = 'r')
    ax.set_title('Degree = %d' % degree_idx)

    #提取参数
    #提取系数
    coef = poly_reg.coef_
    #提取截距
    intercept = poly_reg.intercept_
    
    #回归解析式
    #将截距插入字符串 左边配置了保留一位小数 使用方法format将数值插入到对应的位置
    equation = '$y = {:.1f}'.format(intercept)
    for j in range(1,len(coef)):
        #这里有两个插入值,前一个为保留一位小数的浮点数,后者为整数
        equation += ' + {:.1f}x^{}'.format(coef[j],j)
    equation += '$'
    #将+ - 替换为-
    equation = equation.replace("+ -","-")
    ax.text(0.05,-1.8,equation)
    ax.set_aspect('equal',adjustable = 'box')
    ax.set_xlim(0,4)
    ax.grid(False)
    ax.set_ylim(-2,2)
        
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
#导入Ridge完成岭回归任务,本身带有线性回归的全部功能
from sklearn.linear_model import Ridge
```

```python
np.random.seed(0)
num = 30
```

```python
X = np.random.uniform(0,4,num)
y = np.sin(0.4*np.pi*X)+0.4*np.random.randn(num)
#按照列进行堆叠生成[num,2]的二维矩阵
data = np.column_stack([X,y])
```

```python
x_array = np.linspace(0,4,101).reshape(-1,1)
degree = 8 #多项式回归最高次数
#算上全1列向量,有9列
poly = PolynomialFeatures(degree = degree)
X_poly = poly.fit_transform(X.reshape(-1,1))
```

```python
fig,axes = plt.subplots(3,2,figsize = (10,20))
axes = axes.flatten()
#惩罚因子
alpha_array = [0.00001,0.0001,0.01,1,10,100]
for ax,alpha_idx in zip(axes,alpha_array):

    #训练岭回归模型
    #创建岭回归模型,alpha输入值就是惩罚因子
    ridge = Ridge(alpha = alpha_idx)
    #使用数据进行训练
    ridge.fit(X_poly,y.reshape(-1,1))
    #预测
    y_array_pred = ridge.predict(poly.fit_transform(x_array))
    y_poly_pred = ridge.predict(X_poly)
    data_ = np.column_stack([X,y_poly_pred])
    #绘制散点图
    ax.scatter(X , y,s=20)
    ax.scatter(X,y_poly_pred,marker = 'x',color='k')
    #绘制残差
    ax.plot(([i for (i,j) in data_],[i for (i,j) in data]),
            ([j for (i,j) in data_],[j for (i,j) in data]),
            c = [0.6,0.6,0.6],alpha = 0.5)
    #绘制回归曲线
    ax.plot(x_array,y_array_pred,color = 'r')
    ax.set_title('Alpha = %f' %alpha_idx)

    #提取参数
    #提取系数 第一个系数都是0,因为这里是截距,避免重复
    coef = ridge.coef_[0]
    #提取截距
    intercept = ridge.intercept_[0]
    #回归解析式
    #列出y并添加截距
    equation = '$y = {:.3f}'.format(intercept)
    #这里coef长度为9  第一项截距已经导入,直接从1开始
    #虽然len输出为9,但是这里使用索引进行取值,索引为0-8一共9个,这里的1位索引值的1
    #这里表示生成[1-9)的一个整数数列左闭右开
    for j in range(1,len(coef)):
        equation += '+ {:.3f}x^{}'.format(coef[j],j)
    equation += '$'
    equation = equation.replace("+ -","-")
    print(equation)
    ax.set_aspect('equal',adjustable = 'box')
    ax.set_xlim(0,4);ax.set_ylim(-2,2);ax.grid(False)
```

```python
#多项式回归模型参数随惩罚因子的变化
#根据10的幂作为单位进行均匀分布,作为惩罚因子
alphas = np.logspace(4,-2,100)
#range(10)生成0-9的整数序列,索引从0开始
degrees = ['Degree = ' + str(d_i) for d_i in range(10)]
#配置颜色数组 有10个元素,配置10种颜色
colors = plt.cm.jet(np.linspace(0,1,len(degrees)))
```

```python
coefs = []
for alpha_idx in alphas:
    ridge = Ridge(alpha=alpha_idx)
    #只进行一次使用X_poly的输入,degree最高为8
    ridge.fit(X_poly,y.reshape(-1,1))
    #将每次岭回归生成的稀疏导入其中生成数列
    #每一次是一个惩戒印子生成的系数,9个值,第一个为0,后8个为x^{}的系数
    #因为生成的coef_形状为二维数组(1,9),取[0]表示提取第一行的数列  后面数学可能有多个y,所以一直为二维数组
    coefs.append(ridge.coef_[0])
#转化为np数组
#每一次添加一组0-8 9个值的系数,有100 (100,9)的形状
coefs = np.array(coefs)
```

```python
fig,ax = plt.subplots(figsize = (5,3))

for idx in range(9):
    #idx用于选择第几列 对应degree :表示所有行
    ax.plot(alphas,coefs[:,idx],color = colors[idx])
#使用参数字符串配置横坐标刻度单位
ax.set_xscale("log")
#a:b:c ab为索引 c为步长 不填写ab就是从头到尾 -1为反向
ax.set_xlim(ax.get_xlim()[::-1])
ax.set_xlabel(r"Regularization strength,penalty $\alpha$")
ax.set_ylabel("Coefficients")
#添加图例 标签列表的字符 配置了图例的锚点位置为左侧居中(自身)
#1已经到了右边界,水平方向中间0.5 因为根据的是主图坐标区域的'比例位置'
ax.legend(degrees,loc = 'center left',bbox_to_anchor=(1,0.5))
```
