```python
import numpy as np
import matplotlib.pyplot as plt
#创建训练使用回归模型
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
#0带[]返回二维数组 不带返回一维数组
sepal_length = X[:,[0]]
sepal_width = X[:,[1]]
petal_length = X[:,2]
petal_width = X[:,3]
```

```python
data = np.column_stack([sepal_length,sepal_width])
#堆叠一维数据将一维数据看成一列进行堆叠
```

```python
#创建回归对象并拟合
#面相对象设计语言,先创建对象,再调用方法,拟合后存取数据可以进行预测查看等
#与其他对象不冲突
#逻辑为先配置,再训练
LR = LinearRegression()
#fit这一步自变量输入特征要求为二维,因为可能有很多项
LR.fit(sepal_length,sepal_width)
```

```python
slope = LR.coef_ #返回全部的斜率,一组X_data输出一个斜率
intercept = LR.intercept_
```

```python
x_array = np.linspace(4,9,101).reshape((-1,1))
predicted = LR.predict(x_array)
```

```python
data_ = np.column_stack([sepal_length,LR.predict(sepal_length)])
```

```python
fig,ax = plt.subplots()
ax.scatter(sepal_length,sepal_width)
ax.scatter(sepal_length,LR.predict(sepal_length),color = 'k',marker = 'x')
#用直线连接相近的点
ax.plot(x_array,predicted,color='r')
ax.plot(([i for (i,j) in data_] , [i for (i,j) in data]),
        ([j for (i,j) in data_] , [j for (i,j) in data]),
        #这里plot知道给的x和y下面分别有两个数组,知道绘制线段,每一个元组的两列,第一列为起点,第二列为终点,一次x一次y就变成了对应点
        c = [0.6,0.6,0.6],alpha = 0.5)
ax.set_xlabel('x');ax.set_ylabel('y')
ax.set_aspect('equal',adjustable = 'box')
#equal表示xy单位相等
ax.set_xlim(4,9);ax.set_ylim(1,5)
```
