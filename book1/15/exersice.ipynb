```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
from matplotlib import cm
import os

#如果文件不存在创建文件夹
if not os.path.isdir("Figures"):
    os.makedirs("Figures")
```

```python
def visualize_2D(array,title,vmax,vmin): #vmax vmin表示绘制的圆圈的颜色的最大值最小值

    fig_width  = math.ceil(array.shape[1]*0.5)
    fig_length = math.ceil(array.shape[0]*0.5)

    fig,ax=plt.subplots(figsize=(fig_width,fig_length))

    sns.heatmap(array,vmax=vmax,vmin=vmin,annot=True,
                fmt=".2f",square=True,cmap='RdYlBu_r',
                linewidths=.5,cbar=False,
                yticklabels=False,xticklabels=False,
                ax=ax)
    fig.savefig('Figures/' + title + '.svg', format='svg')

def visualize_1D(array,title,vmax,vmin):
    fig, ax = plt.subplots()
    cmap = plt.colormaps.get_cmap("RdYlBu")
    array_norm=(array-vmin)/(vmax-vmin)
    colors = cmap(array_norm)

    for idx in range(len(array)):
        circle_idx=plt.Circle((idx,0),0.5,facecolor=colors[idx],edgecolor='w')
        ax.add_patch(circle_idx)
        ax.text(idx,0,s="{:.1f}".format(array[idx]),#s表示字符串,其中{}中的含义表示保留一位小数点
                horizontalalignment='center',
                verticalalignment='center')

    ax.set_xlim(-0.6,0.6+len(array))
    ax.set_ylim(-0.6,0.6)
    ax.set_aspect('equal',adjustable='box')
    ax.axis('off')
    fig.savefig('Figures/' +title +'.svg',format='svg')
```

```python
a_1D=np.arange(-2,2+1)
visualize_1D(a_1D,'一维数组',-4,4)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

#导入鸢尾花数据
iris=load_iris()
iris_data_array=iris.data  #因为data属性是numpy数组,所以不需要添加()

print(iris_data_array.max())
print(iris_data_array.max(axis=0))
print(np.argmax(iris_data_array,axis=0))  #每一列最大值位置
print(iris_data_array.max(axis=1))
print(np.average(iris_data_array,axis=0))  #每一列均值

print(np.var(iris_data_array,axis=0))  #每一列方差,默认分母为n

print(np.var(iris_data_array,axis=0,ddof=1))  #将分母设为n-1

print(np.std(iris_data_array,axis=0))  #每一列标准差
```

```python
SIGMA=np.cov(iris_data_array.T,ddof=1)  #转至后计算协方差矩阵
print(SIGMA)
```

```python
#可视化协方差矩阵
fig,ax=plt.subplots(figsize=(5,5))
sns.heatmap(SIGMA, cmap = 'RdYlBu_r', annot = True,
            ax = ax,fmt = ".2f", square=True,
            xticklabels=[],yticklabels=[],cbar=True)
```

```python
CORR=np.corrcoef(iris_data_array.T)  #计算相关系数矩阵
print(CORR)
```

```python
fig,ax=plt.subplots(figsize=(5,5))
sns.heatmap(CORR,cmap='RdYlBu_r',annot=True,
            ax=ax,fmt=".2f",square=True,
            xticklabels=[],yticklabels=[],cbar=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#定义可视化函数,fx表示函数类似f(x)的意思,f_array的f表示f(x)的值,也可以叫做y值
def visualize_fx(x_array,f_array,title,step=False):#step表示阶跃表示,90度向上或者向下,正常的plot表示的是斜线的矢量图

    fig,ax=plt.subplots(figsize=(5,5))
    #绘制从坐标(-5,-5)到(5,5)的对角线,符合x的范围和y的范围
    ax.plot([-5,5],[-5,5],c='r',ls='--',lw=0.5)

    if step:
        ax.step(x_array,f_array)
    else:
        #绘制自定义函数的折线图
        ax.plot(x_array,f_array)


    ax.set_xlim(-5,5)
    ax.set_ylim(-5,5)
    #k为黑色,绘制x和y轴的加粗线
    ax.axvline(0,c='k')
    ax.axhline(0,c='k')
    ax.set_xticks(np.arange(-5,5+1))
    ax.set_yticks(np.arange(-5,5+1))
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    plt.grid(True)
    ax.set_aspect('equal',adjustable='box')
    fig.savefig(title+'.svg',format='svg')
```

```python
x_array=np.linspace(-5,5,1001)
f_array=np.power(x_array,2)

visualize_fx(x_array,f_array,'幂函数_p=2')
```

```python
x_array_=np.copy(x_array)
#_仅表示命名习惯,代表备份的数组名称. []表示索引,|表示'或'这里表示返回小于1和大于一的位置 知道这个位置之后,使用nan处理
#nan表示not a number  ,将超出的值配置成NaN的形式 使得运算节约资源
x_array_[(x_array_<-1)|(x_array_>1)]=np.nan
f_array=np.arcsin(x_array_)
visualize_fx(x_array_,f_array,'反正弦函数')
```

```python
f_array=np.tan(x_array)
f_array[:-1][np.diff(f_array) < 0]=np.nan
visualize_fx(x_array,f_array,'正切函数')
```

```python
f_array=np.floor(x_array)
visualize_fx(x_array,f_array,'向下取整',True)
```

```python
x_array_=np.copy(x_array)
x_array_[x_array_<=0] = np.nan
f_array=np.log(x_array_)
visualize_fx(x_array_,f_array,'对数函数')
```
