```python
sigma_X=1
sigma_Y=1
rho_XY=0.6
mu_X=0
mu_Y=0
```

```python
def f_XY_PDF(xx1,xx2,sigma_X,sigma_Y,rho_XY, mu_X, mu_Y):
    coeff=2 * np.pi * sigma_Y *np.sqrt(1 - rho_XY**2)
    coeff=1 / coeff
    xx1_=(xx1-mu_X)/sigma_X
    xx2_=(xx2-mu_Y)/sigma_Y

    ellipse = 1/(1-rho_XY**2) * (xx1**2 - 2*rho_XY*xx1_*xx2_ + xx2**2)
    PDF = coeff * np.exp(-1/2 * ellipse )

    return PDF
```

```python
PDF_ff = f_XY_PDF(xx1,xx2,sigma_X,sigma_Y,rho_XY,mu_X,mu_Y)
```

```python
fig,ax=plt.subplots(figsize=(4,4))
ax.contourf(xx1,xx2,PDF_ff)

ax.set_xlim(-3,3)
ax.set_ylim(-3,3)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.grid()
ax.set_aspect('equal',adjustable='box')
```
