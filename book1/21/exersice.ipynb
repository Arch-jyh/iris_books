```python
import pandas as pd
import seaborn as sns

iris_df = sns.load_dataset("iris")
#使用drop删除一列 inplace直接修改df本身不创建新的副本
iris_df.drop(columns='species',inplace=True)
#创建了boll_df,仅仅应用于这一特定的列,生成的条件也只有一列,只和这列的数值有关系
condition=iris_df['sepal_length']>=7
#上述生成的值仅对行有关系,因为也是一列的df数据,所以针对这个数据,无法筛选某一列的值(无法指定)
#所以仅配置于行即可,不管bool表的生成原因为那一列,用于行后保留所有列信息
#根据新的布尔df进行判断,结果保留整行,只要对应列的bolldf不符合
iris_df_filtered = iris_df[condition]
#继承了原索引值,所以范围还是0-150,但是数据点有跳跃
iris_df_filtered.plot()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#创建日期范围
date_range = pd.date_range(start='2023-01-01',
                           periods=365*2,freq='D')  #frep='D'表示每天,是频率参数
#空数据帧,指定索引为日期索引
df=pd.DataFrame(index=date_range)
#指定有多少条线
num_path=50
#指定种子
np.random.seed(0)

for i in range(num_path):
    #每天生成的一个步长
    step_index = np.random.normal(loc=0.0,scale=1.0,
                                  #后面定义了初始值为0作为第一个起点,不需要添加第一个值
                                  size=len(date_range)-1)
    step_idx = np.append(0,step_index)
    #计算逐步累加值,生成新的array
    walk_idx = step_idx.cumsum()
    #添加一个列标签然后按照列顺序添加至df的指定列中,配置了列标签
    #这里指的是df数据的某一列,赋值将会赋值到这一列
    df[f'Walk_{i+1}'] = walk_idx

#全部随机行走轨迹
df.plot(legend = False)
plt.grid(True)
plt.ylim(-80,80)
#前两条随机行走轨迹
df.iloc[:,[1,0]].plot(legend=True)
plt.grid(True)
plt.ylim(-80,80)
#时间索引显示前两条轨迹
df.loc['2023-01-01':'2023-08-08',['Walk_1','Walk_2']].plot(legend=True)
plt.grid(True)
plt.ylim(-80,80)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#创建日期范围
date_range = pd.date_range(start='2023-01-01',
                           periods=365*2,freq='D')  #frep='D'表示每天,是频率参数

#指定有多少条线
num_path=50
#指定种子
np.random.seed(0)
#源代码方式不同,源代码在for循环中生成数据所以是50次个749,这里直接生成749*50个数据
step_index = np.random.normal(loc=0.0,scale=1.0,
                              #后面定义了初始值为0作为第一个起点,不需要添加第一个值
                              size=((len(date_range)-1),50))
#行方向(向下axis=0方向)合并为vstack 列方向(向右)为hstack
step_idx = np.vstack((np.zeros((1,50)),step_index))
#计算逐步累加逐步记住每一次的累加值,生成新的array
#指定按照行方向进行累加
walk_idx = step_idx.cumsum(axis=0)
#添加一个列标签然后按照列顺序添加至df的指定列中,配置了列标签
#不使用空数据帧,指定索引为日期索引.正常方式添加数据即可
#在这里同时配置列索引名称
df=pd.DataFrame(walk_idx,index=date_range,columns=[f'Walk_{i+1}' for i in range(50)])

#全部随机行走轨迹
df.plot(legend = False)
plt.grid(True)
plt.ylim(-80,80)
#前两条随机行走轨迹
df.iloc[:,[1,0]].plot(legend=True)
plt.grid(True)
plt.ylim(-80,80)
#时间索引显示前两条轨迹
df.loc['2023-01-01':'2023-08-08',['Walk_1','Walk_2']].plot(legend=True)
plt.grid(True)
plt.ylim(-80,80)
```
