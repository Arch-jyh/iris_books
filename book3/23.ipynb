```python
import numpy as np
```

```python
A = np.array([[1,1],
              [2,4]])
b = np.array([[35],
              [94]])
```

```python
A_inv = np.linalg.inv(A)
```

```python
x = A_inv@b
print(x)
```

```python
x_ = np.linalg.solve(A,b)
print(x_)
```

```python
from sympy import *
x1,x2 = symbols(['x1','x2'])
```

```python
#sympy中的solve,通用求解器,只能是一次的线性的
sol = solve([x1 + x2 - 35,2*x1 + 4*x2 - 94],[x1,x2]) 
print(sol)
```

```python
#更简单的写法:from sympy import linsolve
from sympy.solvers.solveset import linsolve
```

```python
sol_ = linsolve([x1 + x2 - 35,2*x1 + 4*x2 - 94],[x1,x2])
print(sol_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_vector(vector,RBG):
    array = np.array([[0,0,vector[0],vector[1]]],dtype = object)
    X,Y,U,V = zip(*array)
    plt.quiver(X,Y,U,V,angles = 'xy',scale_units = 'xy',
               scale = 1,color = RBG,zorder = 1000)
```

```python
x1 = np.arange(-25,25 + 1,step = 1)
x2 = np.arange(-25,25 + 1,step = 1)
```

```python
XX1,XX2 = np.meshgrid(x1,x2)
XX1
```

```python
X = np.column_stack((XX1.ravel(),XX2.ravel()))#默认按行,从第一行开始展开
```

```python
A = np.matrix([[1,1],
               [2,4]])
```

```python
Z = X @ A.T
```

```python
ZZ1 = Z[:,0].reshape((len(x1),len(x2)))#默认按行,从第一行开始合并
ZZ2 = Z[:,1].reshape((len(x1),len(x2)))
```

```python
fig,ax = plt.subplots()

plt.plot(XX1,XX2,color = [0.8,0.8,0.8]) #matplotlib默认按数值的列绘制
plt.plot(XX1.T,XX2.T,color = [0.8,0.8,0.8])

a1 = A[:,0].tolist()
a2 = A[:,1].tolist()

b = [3,8]

draw_vector(a1,np.array([0,112,192])/255)
draw_vector(a2,np.array([255,0,0])/255)
draw_vector(b,np.array([255,125,255])/255)

plt.xlabel('$e_1$')  # 设置 x 轴标签
plt.ylabel('$e_2$')  # 设置 y 轴标签

plt.axis('scaled')  # 设置坐标轴比例
ax.set_xlim([0, 8])  # 设置 x 轴范围
ax.set_ylim([0, 8])  # 设置 y 轴范围

plt.xticks(np.arange(0, 8 + 1, step=2))  # 设置 x 轴刻度
plt.yticks(np.arange(0, 8 + 1, step=2))  # 设置 y 轴刻度

ax.spines['top'].set_visible(False)  # 隐藏顶部边框
ax.spines['right'].set_visible(False)  # 隐藏右边框
```

```python
fig,ax = plt.subplots()

plt.plot(ZZ1,ZZ2,color = '0.8')
plt.plot(ZZ1.T,ZZ2.T,color = '0.8')

draw_vector(a1,np.array([0,112,192])/255)
draw_vector(a2,np.array([255,0,0])/255)
draw_vector(b,np.array([255,125,255])/255)


plt.axis('scaled')  # 设置坐标轴比例
ax.set_xlim([0, 8])  # 设置 x 轴范围
ax.set_ylim([0, 8])  # 设置 y 轴范围

plt.xticks(np.arange(0, 8 + 1, step=2))  # 设置 x 轴刻度
plt.yticks(np.arange(0, 8 + 1, step=2))  # 设置 y 轴刻度

ax.spines['top'].set_visible(False)  # 隐藏顶部边框
ax.spines['right'].set_visible(False)  # 隐藏右边框
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_vector(vector,RGB):
    array = np.array([[0,0,vector[0],vector[1]]])
    X,Y,U,V = zip(*array)
    plt.quiver(X,Y,U,V,angles = 'xy',scale_units = 'xy',
               scale = 1,color = RGB,zorder = 1000)
```

```python
x1 = np.arange(-25,25 + 1,step = 1)
x2 = np.arange(-25,25 + 1,step = 1)
XX1,XX2 = np.meshgrid(x1,x2)
X = np.column_stack((XX1.ravel(),XX2.ravel()))
```

```python
R = np.matrix('1,3;2,1')#旧方法,可以使用字符串
Z = X@R.T
ZZ1 = Z[:,0].reshape((len(x1),len(x2)))
ZZ2 = Z[:,1].reshape((len(x1),len(x2)))
```

```python
fig,ax = plt.subplots()

plt.plot(XX1,XX2,color = [0.8,0.8,0.8]) #matplotlib默认按数值的列绘制
plt.plot(XX1.T,XX2.T,color = [0.8,0.8,0.8])

draw_vector([1,2],np.array([0,112,192])/255)
draw_vector([3,1],np.array([255,0,0])/255)
draw_vector([10,10],'#FF99FF')

plt.xlabel('$x_1$ (number of chickens)')
plt.ylabel('$x_2$ (number of rabbits)')

plt.axis('scaled')  # 设置坐标轴比例
ax.set_xlim([0, 10])  # 设置 x 轴范围
ax.set_ylim([0, 10])  # 设置 y 轴范围

plt.xticks(np.arange(0, 10 + 1, step=2))  # 设置 x 轴刻度
plt.yticks(np.arange(0, 10 + 1, step=2))  # 设置 y 轴刻度

ax.spines['top'].set_visible(False)  # 隐藏顶部边框
ax.spines['right'].set_visible(False)  # 隐藏右边框
```

```python
fig,ax = plt.subplots()

plt.plot(ZZ1,ZZ2,color = [0.8,0.8,0.8]) #matplotlib默认按数值的列绘制
plt.plot(ZZ1.T,ZZ2.T,color = [0.8,0.8,0.8])

draw_vector([1,2],np.array([0,112,192])/255)
draw_vector([3,1],np.array([255,0,0])/255)
draw_vector([10,10],'#FF99FF')

plt.xlabel('$z_1$ (combo A)')
plt.ylabel('$z_2$ (combo B)')

plt.axis('scaled')  # 设置坐标轴比例
ax.set_xlim([0, 10])  # 设置 x 轴范围
ax.set_ylim([0, 10])  # 设置 y 轴范围

plt.xticks(np.arange(0, 10 + 1, step=2))  # 设置 x 轴刻度
plt.yticks(np.arange(0, 10 + 1, step=2))  # 设置 y 轴刻度

ax.spines['top'].set_visible(False)  # 隐藏顶部边框
ax.spines['right'].set_visible(False)  # 隐藏右边框
```
