```python
from sympy import latex, lambdify, diff, sin, log, exp, series #series用来做泰勒展开
from sympy.abc import x  # 定义符号变量 x
import numpy as np
from matplotlib import pyplot as plt  # 导入绘图库 Matplotlib
from matplotlib import cm  # 导入颜色映射模块
```

```python
f_x = exp(x)
f_x
```

```python
x_array = np.linspace(-2,2,100)
x_0 = 0#中心点
```

```python
y_0 = f_x.evalf(subs = {x:x_0})
y_0 #符号形式的值
```

```python
f_x_fcn = lambdify(x,f_x)
f_x_array = f_x_fcn(x_array)
```

```python
fig,ax = plt.subplots()

ax.plot(x_array,f_x_array,'k',linewidth = 1.5)
ax.plot(x_0,y_0,'xr',markersize = 12)
ax.set_xlabel(r'$\it{x}$')
ax.set_ylabel(r'$\it{f}(\it{x})$')

highest_order = 5
order_array = np.arange(0,highest_order + 1) #不包含右端点 加一处理
colors = plt.cm.rainbow(np.linspace(0,1,len(order_array)))

i = 0

for order in order_array:
    f_series = f_x.series(x,x_0,order + 1).removeO() #默认最后有高阶误差,所以也加一.最后删除高阶误差用removeO
    f_series_fcn = lambdify(x,f_series)
    f_series_array = f_series_fcn(x_array)

    ax.plot(x_array,x_array*0 + f_series_array,linewidth = 0.5,color = colors[i,:],label = 'Order = %0.0f' % order)
    i+=1

ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])  # 添加网格
ax.set_xlim(x_array.min(), x_array.max())  # 设置 x 轴范围
ax.spines['right'].set_visible(False)  # 隐藏右侧边框
ax.spines['top'].set_visible(False)  # 隐藏顶部边框

plt.legend()  # 添加图例
```

```python
fig = plt.figure(figsize = plt.figaspect(0.5)) #得到的是固定的英寸尺寸,不会因0.5给不同的
ax = fig.add_subplot(1,2,1)

ax.plot(x_array,f_x_array,'k',linewidth = 1.5)
ax.plot(x_0,y_0,'xr',markersize = 12)
ax.set_xlabel(r'$\it{x}$')
ax.set_ylabel(r'$\it{f}(\it{x})$')

highest_order = 2

f_series = f_x.series(x,x_0,highest_order + 1).removeO()
f_series_fcn = lambdify(x,f_series)
f_series_array = f_series_fcn(x_array)
f_series_array = x_array * 0 + f_series_array#防止order = 0和拟合出常数直线

ax.plot(x_array,f_series_array,linewidth = 1.5,color = 'b')

ax.fill_between(x_array,f_x_array,x_array*0 + f_series_array,color = '#DEEAF6')

ax = fig.add_subplot(1,2,2)

error = f_x_array - f_series_array
ax.plot(x_array,error,'r',linewidth = 1.5)
ax.fill_between(x_array,error,color = '#DEEAF6')#默认和y = 0之间填充
plt.axhline(y = 0,color = 'k',linestyle = '--',linewidth = 0.25)
ax.set_xlabel(r"$\it{x}$")  # 设置 x 轴标签
ax.set_ylabel("Error")  # 设置 y 轴标签

ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])  # 添加网格
ax.set_xlim(x_array.min(), x_array.max())  # 设置 x 轴范围
ax.set_ylim(np.floor(error.min()), np.ceil(error.max()))  # floor向下取整, ceil向上取整
ax.spines['right'].set_visible(False)  # 隐藏右侧边框
ax.spines['top'].set_visible(False)  # 隐藏顶部边框
```

```python
import numpy as np
from sympy import lambdify, diff, exp, latex, simplify  # 导入符号计算和数学库
from sympy.abc import x, y  # 定义符号变量 x 和 y
from matplotlib import pyplot as plt  # 导入 Matplotlib 进行绘图
from matplotlib import cm  # 导入颜色映射模块
```

```python
num = 301
x_array = np.linspace(-1.5,1.5,num)
y_array = np.linspace(-1.5,1.5,num)
```

```python
xx,yy = np.meshgrid(x_array,y_array)
num_stride = 10 #绘制步长
```

```python
f_xy = exp(-x**2 - y**2)
f_xy_fcn = lambdify([x,y],f_xy)
f_xy_zz = f_xy_fcn(xx,yy)
```

```python
x_a = -0.1
y_b = -0.2
```

```python
x_a_array = np.linspace(x_a - 0.5,x_a + 0.5,101)
y_b_array = np.linspace(y_b - 0.5,y_b + 0.5,101)

xx_local,yy_local = np.meshgrid(x_a_array,y_b_array) #局部网格

f_xy_zz_local = f_xy_fcn(xx_local,yy_local)
```

```python
f_ab = f_xy_fcn(x_a,y_b)
```

```python
fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})

ax.plot_wireframe(xx,yy,f_xy_zz,
                  color = '0.5',
                  rstride = num_stride,cstride = num_stride,
                  linewidth = 0.25)

approx_zero_order = f_ab + xx_local * 0

ax.plot_wireframe(xx_local,yy_local,approx_zero_order,
                  color = [1,0,0],
                  rstride = num_stride,cstride = num_stride,
                  linewidth = 0.25)
ax.plot(x_a,y_b,f_ab,marker = 'x',color = 'r',
        markersize = 12)

ax.set_proj_type('ortho')  # 设置正交投影

ax.set_xlabel('$x_1$')  # 设置 x 轴标签
ax.set_ylabel('$x_2$')  # 设置 y 轴标签
ax.set_zlabel('$f(x_1,x_2)$')  # 设置 z 轴标签

ax.set_xlim(xx.min(), xx.max())  # 设置 x 轴范围
ax.set_ylim(yy.min(), yy.max())  # 设置 y 轴范围
ax.set_zlim(f_xy_zz.min(), 1.5)  # 设置 z 轴范围

ax.view_init(azim=-145, elev=30)  # 设置视角
plt.tight_layout()
ax.grid(False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy.abc import x
from sympy import latex, lambdify, diff, sin, log, exp  # 导入必要的符号计算函数
```

```python
def num_diff(f,a,method,dx):
    if method == 'central':
        return (f(a + dx) - f(a - dx)) / (2*dx)
    elif method == 'forward':
        return (f(a + dx) - f(a)) / dx
    elif method == 'backward':
        return (f(a) - f(a - dx)) / dx
    else:
        raise ValueError('Method must be "central" , "forward" , "backward" . ')
```

```python
f_x = exp(-x**2)
f_x
```

```python
x_array = np.linspace(-3,3,100)
a_array = np.linspace(-2.5,2.5,11)
```

```python
f_x_fcn = lambdify(x,f_x)
f_x_array = f_x_fcn(x_array)
```

```python
f_x_1_diff = diff(f_x,x)
f_x_1_diff
```

```python
f_x_1_diff_fcn = lambdify(x,f_x_1_diff)
f_x_1_diff_array = f_x_1_diff_fcn(x_array)
```

```python
fig = plt.figure(figsize = plt.figaspect(0.5))
ax = fig.add_subplot(2,1,1)

ax.plot(x_array,f_x_array,'#0070C0',linewidth = 1.5)
ax.set_ylim(np.floor(f_x_array.min()),
            np.ceil(f_x_array.max()))
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_xlim(x_array.min(),x_array.max())
ax.spines['right'].set_visible(False)  # 隐藏右边框
ax.spines['top'].set_visible(False)  # 隐藏上边框


ax = fig.add_subplot(2,1,2)

ax.plot(x_array,f_x_1_diff_array,'#0070C0',linewidth = 1.5)
ax.set_ylim(np.floor(f_x_1_diff_array.min()),
            np.ceil(f_x_1_diff_array.max()))
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_xlim(x_array.min(),x_array.max())
ax.spines['right'].set_visible(False)  # 隐藏右边框
ax.spines['top'].set_visible(False)  # 隐藏上边框
```

```python
dx = 0.2

diff_central = num_diff(f_x_fcn,a_array,'central',dx)
diff_forward = num_diff(f_x_fcn,a_array,'forward',dx)
diff_backward = num_diff(f_x_fcn,a_array,'backward',dx)
```

```python
fig,ax = plt.subplots()

ax.plot(x_array,f_x_1_diff_array,'#0070C0',linewidth = 1.5)

ax.plot(a_array,diff_central,marker = '.',
        markersize = 12,linestyle = 'none',
        label = 'central')

ax.plot(a_array,diff_forward,marker = '>',
        markersize = 12,linestyle = 'none',
        label = 'forward')

ax.plot(a_array,diff_backward,marker = '<',
        markersize = 12,linestyle = 'none',
        label = 'backward')

ax.set_xlabel('x')  # 设置 x 轴标签
ax.set_ylabel('f\'(x)')  # 设置 y 轴标签
ax.spines['right'].set_visible(False)  # 隐藏右边框
ax.spines['top'].set_visible(False)  # 隐藏上边框
ax.set_xlim((x_array.min(), x_array.max()))  # 设置 x 轴范围
plt.axhline(y=0, color='k', linestyle='--', linewidth=0.25)  # 添加 y=0 的参考线
plt.legend()  # 添加图例
```
