```python
from sympy import lambdify,diff,evalf,sin,exp  #evalf进行数值评估,将符号表达式转化为估算的数字
from sympy.abc import x
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
```

```python
f_x = exp(-x**2)
f_x
```

```python
x_array = np.linspace(-3,3,100)
x_0_array = np.linspace(-2.5,2.5,21)#展开点的位置
```

```python
f_x_fcn = lambdify(x,f_x) #两个变量输入lambdify((x,y),f_x)
f_x_array = f_x_fcn(x_array)
```

```python
colors = plt.cm.rainbow(np.linspace(0,1,len(x_0_array)))
```

```python
f_x_1_diff = diff(f_x,x)#顺序反的,先公式后字母
f_x_1_diff
```

```python
f_x_1_diff_fcn = lambdify(x,f_x_1_diff)
```

```python
f_x_2_diff = diff(f_x,x,2)
f_x_2_diff
```

```python
f_x_2_diff_fcn = lambdify(x,f_x_2_diff)
```

```python
fig,ax = plt.subplots()

ax.plot(x_array,f_x_array,linewidth = 1.5)
ax.set_xlabel(r'$\it{x}$')
ax.set_ylabel(r'$\it{f}(\it{x})$')

for i in np.arange(len(x_0_array)):
    color = colors[i,:]#
```
