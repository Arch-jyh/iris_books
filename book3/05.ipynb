```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [4,0,-5,-4,6,0]
y = [2,-2,7,-6,-5,0]

labels = ['A','B','C','D','E','O']
```

```python
fig,ax = plt.subplots()

plt.plot(x,y,'x')

for label,i,j in zip(labels,x,y):
    #文本左下角为锚点,锚点放置在坐标位置.
    #不写{}中的数字将按照顺序填充
    plt.text(i,j+0.5,label + '({},{})'.format(i,j))

plt.xlabel('x');plt.ylabel('y')
plt.axhline(y = 0,color = 'k',ls = '-')
plt.axvline(x = 0,color = 'k',ls = '-')
plt.xticks(np.arange(-8,8+1,step = 1))
plt.yticks(np.arange(-8,8+1,step = 1))
plt.axis('scaled')#等比例,比set_aspect更温和,不覆盖xlim   equal更强硬,覆盖xlim

ax.set_xlim(-8,8)
ax.set_ylim(-8,8)
ax.spines[['top','bottom','left','right']].set_visible(False)

ax.grid(ls = '--',lw = 0.25,color = '0.5')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fig_decor(ax):
    ax.set_xlabel('x')
    ax.set_ylabel('y')

    ax.hlines(y = 0,xmin = -8,xmax = 8,color = 'k')
    ax.vlines(x = 0,ymin = -8,ymax = 8,color = 'k')

    ax.set_xticks(np.arange(-8,8+1,step = 1))
    ax.set_yticks(np.arange(-8,8+1,step = 1))

    ax.axis('scaled')
    ax.grid(ls = '--',lw = 0.25,color = '0.5')

    #配置轴显示边界,类似lim,但是更底层更强制 boundary
    ax.set_xbound(lower = -8,upper = 8)
    ax.set_ybound(lower = -8,upper = 8)

    ax.spines[['top','bottom','right','left']].set_visible(False)
```

```python
x_array = np.linspace(-8,8)
y_array = np.linspace(-8,8)
```

```python
fig,ax = plt.subplots()
y1 = 0.5*x_array + 2
y2 = -2*x_array - 1

ax.plot(x_array,y1)
ax.plot(x_array,y2)
fig_decor(ax)
```

```python
fig,ax = plt.subplots()
y1 = 0.5*x_array + 2
y2 = 0.5*x_array - 1

ax.plot(x_array,y1)
ax.plot(x_array,y2)
fig_decor(ax)
```

```python
fig,ax = plt.subplots()
y1 = 0*x_array + 2
y2 = 0*x_array - 4

ax.plot(x_array,y1)
ax.plot(x_array,y2)
fig_decor(ax)
```

```python
fig,ax = plt.subplots()
x1 = 0*y_array + 2
x2 = 0*y_array - 4

ax.plot(x1,y_array)
ax.plot(x2,y_array)
fig_decor(ax)
```

```python
def plot_polar(theta,r):
    fig,ax = plt.subplots(subplot_kw = {'projection':'polar'})
    ax.plot(theta,r)
    ax.set_rmax(20)
    ax.set_rticks([5,10,15,20])
    ax.set_rlabel_position(-45)
    ax.set_thetagrids(np.arange(0.0,360.0,45.0))
    plot.show()

```

```python
theta = np.linspace(0,6*np.pi,1000)

r = 10 + theta*0
plot_polar(theta,r)
```

```python
theta = np.linspace(0,6*np.pi,1000)

r = 1*theta
plot_polar(theta,r)
```

```python
theta = np.linspace(0,6*np.pi,1000)

r = 10*np.cos(6*theta) + 10
plot_polar(theta,r)
```

```python
t = np.linspace(0,2*np.pi,100)

x1 = np.cos(t)
x2 = np.sin(t)
```

```python
fig,ax = plt.subplots()
plt.plot(x1,x2)

ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')

ax.set_xticks(np.arange(-2,2+1,step = 1))
ax.set_yticks(np.arange(-2,2+1,step = 1))

ax.axis('scaled')
ax.grid(ls = '--',lw = 0.25,color = '0.7')
ax.set_xbound(lower = -2,upper = 2)
ax.set_ybound(lower = -2,upper = 2)
plt.axhline(y = 0,color = 'k',ls = '-')
plt.axvline(x = 0,color = 'k',ls = '-')
```
