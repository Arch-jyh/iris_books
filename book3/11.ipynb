```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w_array = np.array([1/5,1/4,1/3,1/2,1,2,3,4,5])
x_array = np.linspace(-2,2,100)

ww,xx = np.meshgrid(w_array,x_array)

b = 0
ff = ww*xx + b
```

```python
fig,ax = plt.subplots()
colors = plt.cm.jet(np.linspace(0,1,len(w_array)))

for i in np.linspace(1,len(w_array),len(w_array)):
    plt.plot(x_array,ff[:,int(i) - 1],
             color = colors[int(i) - 1],
             label = '$w_1 = {lll:.2f}$'.format(lll = w_array[int(i) - 1]))

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.xticks(np.arange(-2,2.5,step = 0.5))
plt.yticks(np.arange(-2,2.5,step = 0.5))
plt.axis('scaled')#遵循范围使得两者比例相等,可能自动拉伸   equal是单位等长,绘图框尽量不变>>但可能调整范围
ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
ax.spines[['top','bottom','left','right']].set_visible(False)
ax.grid(linestyle = '--',linewidth = 0.25,color = '0.5')
```

```python
x = np.linspace(-2,2,100)
```

```python
def plot_curve(x,y,xlim,ylim):
    fig,ax = plt.subplots()

    plt.xlabel(r'$\it{x}$')
    plt.xlabel(r'$\it{f}(\it{x})$')
    plt.plot(x,y,linewidth = 1.5)
    plt.axhline(y = 0,color = 'k',linewidth = 1.5)
    plt.axvline(x = 0,color = 'k',linewidth = 1.5)
    ax.grid(ls = '--',lw = 0.25,color = '0.5')
    plt.axis('equal')
    plt.xticks(np.arange(-2,2.5,step = 0.5))
    plt.xticks(np.arange(y.min(),y.max()+0.5,step = 0.5))
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.spines[['top','bottom','left','right']].set_visible(False)
    ax.set_aspect('equal','box')
```

```python
y = x+1
plot_curve(x,y,(-2,2),(-1,3))
```

```python
y = -x+1
plot_curve(x,y,(-2,2),(-1,3))
```

```python
y = x**2 - 2
plot_curve(x,y,(-2,2),(-2,2))
```

```python
y = -x**2 + 2
plot_curve(x,y,(-2,2),(-2,2))
```

```python
y = x**3 - x
plot_curve(x,y,(-2,2),(-2,2))
```

```python
y = -x**3 + x
plot_curve(x,y,(-2,2),(-2,2))
```
