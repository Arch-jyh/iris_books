```python
from sympy.abc import x,a
from sympy import Poly#Poly将表达式转换为多项式对象
import matplotlib.pyplot as plt
import numpy as np
```

```python
for n in [4,8,12,5,9,13,36]:
    expr = (x+1)**n
    expr_expand = expr.expand()#展开
    expr_expand = Poly(expr_expand)#转化为多项式形式

    poly_coeffs = expr_expand.coeffs()#求系数

    print(poly_coeffs)#底为二项式的指数展开的系数,但是这不是二项式系数了,这是多项式的

    degrees = np.linspace(n,0,n+1)

    fig,ax = plt.subplots()

    #输入从大到小,数轴默认从小到大不变(方向与数据顺序无关)
        #但是对应关系是先确认的,不会改变前两个参数的映射位置
    plt.stem(degrees,np.array(poly_coeffs,dtype = float))
    plt.xlim(0,n)
    plt.xticks(np.arange(0,n+1))
    y_max = max(poly_coeffs)
    y_max = float(y_max)
    plt.ylim(0,y_max)

    ax.spines[['right','top']].set_visible(False)
    ax.invert_xaxis()#
    plt.xlabel('Degree')
    plt.ylabel('Coefficient')
    plt.show()
```

```python
import itertools#迭代+组合的工具箱 iterate
```

```python
letters = 'ABC'

#cmb是缩写 combination
#参数1传入可迭代对象,字符串本身就是可迭代对象,每个字符都当做一个元素参与组合
#2表示抽取两个
#cmb是一个定义好的计算规则,还没有开始算,但是是一个迭代器
cmb = itertools.combinations(letters,2)

#通过for激活迭代器,才依次计算出来
for val in cmb:
    print(val)
```

```python
letters = 'ABC'

per = itertools.permutations(letters,2)

for val in per:
    print(val)
```
