```python
import numpy as np
from scipy.special import factorial #阶乘库
import matplotlib.pyplot as plt
```

```python
n_array = np.linspace(0,10,11)
```

```python
expansion = factorial(2*n_array) / 2**(4*n_array+2) / factorial(n_array)**2 / (2*n_array - 1) / (2*n_array +3)
expansion
```

```python
est_pi = 24 * (np.sqrt(3) / 32 - np.cumsum(expansion)) #累计和得到数列 cum = cumulative
est_pi
```

```python
fig,ax = plt.subplots()

plt.axhline(y = np.pi,color = 'r',linestyle = '-')
plt.plot(n_array,est_pi,color = 'b',marker = 'x')

plt.tight_layout()
plt.xlabel('n')
plt.ylabel(r'Estimate of $\pi$')
```

```python
num_interval = 20
delta_x = (b - a) / num_interval

x_array = np.linspace(a,b,num_interval + 1)
y_array = f_x_fcn(x_array)

x_array_fine = np.linspace(a,b,200)
y_array_fine = f_x_fcn(x_array_fine)
```

```python
fig = plt.figure(figsize = (5,5))
ax = fig.add_subplot(1,1,1)

plt.plot(x_array_fine,y_array_fine,
         color = '#0070C0')

x_left = x_array[:-1]
y_left = y_array[:-1]

plt.plot(x_left,y_left,'rx',markersize = 10)
plt.bar(x_left,y_left,
        width = delta_x,
        facecolor = '#DEEAF6',
        align = 'edge',#delta_x是正数,所以想有延伸,变成了靠左对齐
        edgecolor = '#B2B2B2')

ax.axvline(x = a,color = 'r',linestyle = '-')
ax.axvline(x = b,color = 'r',linestyle = '-')

left_riemann_sum = np.sum(y_left * delta_x) 

plt.title('Left Riemann sum (N = %0.0f) = %0.3f'
          % (num_interval,left_riemann_sum))

plt.xlim(a,b)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
def riemann_sum(num_interval):
    a,b = 0,1
    delta_x = (b - a) / num_interval
    
    x_array = np.linspace(a,b,num_interval+1)
    y_array = f_x_fcn(x_array)
    
    x_left_array = x_array[:-1]
    y_left_array = y_array[:-1]
    
    x_mid_array = (x_array[:-1] + x_array[1:]) / 2
    y_min_array = f_x_fcn(x_mid_array)
    
    x_right_array = x_array[1:]
    y_right_array = y_array[1:]
    
    left_riemann_sum = np.sum(y_left_array*delta_x)
    min_riemann_sum = np.sum(y_min_array*delta_x)
    right_riemann_sum = np.sum(y_right_array*delta_x)

    return left_riemann_sum,min_riemann_sum,right_riemann_sum
```

```python
left_sum_array = np.array([])
mid_sum_array = np.array([])
right_sum_array = np.array([])
for i in np.arange(5,100+1,step = 5):
    left_sum,mid_sum,right_sum = riemann_sum(int(i))
    
    left_sum_array = np.append(left_sum_array,left_sum) #是创建一个新数组,所以需要赋值
    mid_sum_array = np.append(mid_sum_array,mid_sum)
    right_sum_array = np.append(right_sum_array,right_sum)
```

```python
from mpl_toolkits.mplot3d import axes3d #让axes支持3d图形,新版可以直接用了,旧版要先加载才能用3d绘图
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style #配置图像预设风格
from sympy.abc import x,y
from sympy import *
```

```python
f_xy = exp( - x**2 - y**2)
f_xy
```

```python
f_xy_fcn = lambdify((x,y),f_xy)
```

```python
a,b,c,d = -2,1,-1,2

x_array_fine = np.linspace(a,b,300)
y_array_fine = np.linspace(c,d,300)

xx_fine,yy_fine = np.meshgrid(x_array_fine,y_array_fine)
zz_fine = f_xy_fcn(xx_fine,yy_fine)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

ax.plot_wireframe(xx_fine,yy_fine,zz_fine,
                  color = '#0070C0',
                  rstride = 10,cstride = 10,
                  linewidth = 0.25)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z = f(x,y)')

ax.set_xlim((a, b))  # 设置 x 轴范围
ax.set_ylim((c, d))  # 设置 y 轴范围
ax.set_zlim((0, zz_fine.max()))  # 设置 z 轴范围
ax.grid(False)
ax.view_init(azim=-135, elev=30)  # 设置视角
ax.set_proj_type('ortho')
plt.show()
```

```python
num_array = [5,10,15,20]

for num in num_array:

    x_array = np.linspace(a,b - (b - a) / num,num)
    y_array = np.linspace(c,d - (d - c) / num,num)

    xx,yy = np.meshgrid(x_array,y_array)
    xx_array = xx.ravel()
    yy_array = yy.ravel()

    fig = plt.figure()
    ax = fig.add_subplot(111,projection = '3d')
    zz_array = np.zeros_like(yy_array)

    dx = np.ones_like(yy_array) / num*(b - a)
    dy = np.ones_like(yy_array) / num*(d - c)

    dz = f_xy_fcn(xx_array,yy_array)

    ax.bar3d(xx_array,yy_array,zz_array,dx,dy,dz,shade = False,
            color = '#DEEAF6',edgecolor = '#B2B2B2')#

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z = f(x,y)')
    
    ax.set_xlim((a, b))
    ax.set_ylim((c, d))
    ax.set_zlim((0, zz_fine.max()))
    ax.grid(False)
    ax.view_init(azim=-135, elev=30)
    ax.set_proj_type('ortho')
    plt.show()

    estimated_volume = dz.sum() * (b - a) / num*(d - c) / num
    ax.set_title('Estimated volume = %0.3f' %estimated_volume)
```
