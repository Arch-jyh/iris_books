```python
import numpy as np
from sympy import *
from matplotlib import pyplot as plt
```

```python
x = Symbol('x')
```

```python
f_x = x**2 + 1/2
f_x
```

```python
f_x_fcn = lambdify(x,f_x)
```

```python
integral_f_x = integrate(f_x,x) #不定积分
integral_f_x
```

```python
integral_f_x_fcn = lambdify(x,integral_f_x)
```

```python
a = 0
b = 1

num = 201
x_array = np.linspace(-0.2,1.2,num)
x_a_b_array = np.linspace(a,b,num)
```

```python
y_array = f_x_fcn(x_array)
y_a_b_array = f_x_fcn(x_a_b_array)
```

```python
integral_a_b = integrate(f_x,(x,a,b))
integral_a_b
```

```python
integral_a_b_v2 = float(integral_a_b)
integral_a_b_v2
```

```python
fig,ax = plt.subplots()
ax.plot(x_array,y_array,'b')
ax.axvline(x = a,color = 'r',linestyle = '-')
ax.axvline(x = b,color = 'r',linestyle = '-')
ax.axhline(y = 0,color = 'k',linestyle = '-')

ax.fill_between(x_a_b_array,
                y_a_b_array,
                edgecolor = 'none',
                facecolor = '#DBEEF3')

ax.set_xlim(-0.2, 1.2)  # 设置 x 轴范围
ax.set_ylim(-1, 2)  # 设置 y 轴范围
plt.gca().spines['right'].set_visible(False)  # 隐藏右边框
plt.gca().spines['top'].set_visible(False)  # 隐藏上边框
ax.set_xlabel('x')  # 设置 x 轴标签
ax.set_ylabel('f(x)')  # 设置 y 轴标签

#r之影响‘\'不影响‘%’。 %会先调用float,再去小数位,然后字符串
ax.set_title(r'$\int_a^b f(x) = %0.4f$' %integral_a_b)
```

```python
import numpy as np
from scipy.special import factorial #阶乘库
import matplotlib.pyplot as plt
```

```python
n_array = np.linspace(0,10,11)
```

```python
expansion = factorial(2*n_array) / 2**(4*n_array+2) / factorial(n_array)**2 / (2*n_array - 1) / (2*n_array +3)
expansion
```

```python
est_pi = 24 * (np.sqrt(3) / 32 - np.cumsum(expansion)) #累计和得到数列 cum = cumulative
est_pi
```

```python
fig,ax = plt.subplots()

plt.axhline(y = np.pi,color = 'r',linestyle = '-')
plt.plot(n_array,est_pi,color = 'b',marker = 'x')

plt.tight_layout()
plt.xlabel('n')
plt.ylabel(r'Estimate of $\pi$')
```
