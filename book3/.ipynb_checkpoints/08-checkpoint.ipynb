```python
from matplotlib import pyplot as plt
from sympy import plot_implicit,symbols,Eq
```

```python
x1,x2 = symbols('x1 x2')
```

```python
def plot_curve(Eq_sym):
    #绘制隐式曲线
    h_plot = plot_implicit(Eq_sym,(x1,-2.5,2.5),(x2,-2.5,2.5),
                           xlabel = r'$\it{x_1}$',ylabel = r'$\it{x_2}$')
```

```python
plt.close('all')
```

```python
Eq_sym = Eq(x1**2 / 4 + x2**2,1)
plot_curve(Eq_sym)
```

```python
Eq_sym = Eq(x1**2 + x2**2/4,1)
plot_curve(Eq_sym)
```

```python
Eq_sym = Eq((x1 - 0.5)**2 / 4 + (x2-0.5)**2,1)
plot_curve(Eq_sym)
```

```python
Eq_sym = Eq((x1 - 0.5)**2 + (x2-0.5)**2/4,1)
plot_curve(Eq_sym)
```

```python
Eq_sym = Eq(5*x1**2 / 8 + 5*x2**2/8,1)
plot_curve(Eq_sym)
```

```python
Eq_sym = Eq(5*x1**2 / 8 + 3*x1*x2/4 + 5*x2**2/8,1)
plot_curve(Eq_sym)
```

```python
Eq_sym = Eq(4*1*x2,x1**2)
plot_curve(Eq_sym)
```

```python
Eq_sym = Eq(4*1*(x2 + 1),x1**2)
plot_curve(Eq_sym)
```

```python
Eq_sym = Eq(4*(-1)*(x2 - 1),x1**2)
plot_curve(Eq_sym)
```

```python
Eq_sym = Eq(4*(1)*(x1),x2**2)
plot_curve(Eq_sym)
```

```python
Eq_sym = Eq(4*(1)*(x1 + 1),x2**2)
plot_curve(Eq_sym)
```

```python
Eq_sym = Eq(4*(1)*(x1 - 1),x2**2)
plot_curve(Eq_sym)
```
