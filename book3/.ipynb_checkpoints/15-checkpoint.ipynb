```python
from sympy import latex,lambdify,limit,log,oo
from sympy.abc import x
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_x = (1+1/x)**x
f_x
```

```python
x_array = np.linspace(0.1,1000,1000)
```

```python
f_x_fcn = lambdify(x,f_x)
f_x_array = f_x_fcn(x_array)
```

```python
#和limit_seq不是一样的语法
#第三个参数必须写,表示趋近的点的一个行为,不是求值,表示趋近的点的位置
    #函数可能在某个x取到了极限,所以没有默认值(数列默认是正无穷)
#第四个参数表示趋近位置,从那一侧靠近a,'-'表示从左边靠近x<a  '+'表示从右边靠近x>a
    #不写默认为双向
    #数列只有正方向,不需要这个参数
    #极限点oo,-oo方向趋近方向没有意义
f_x_oo_limit = limit(f_x,x,oo)
f_x_oo_limit
```
