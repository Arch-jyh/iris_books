```python
from sympy import latex,lambdify,limit,log,oo
from sympy.abc import x
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_x = (1+1/x)**x
f_x
```

```python
x_array = np.linspace(0.1,1000,1000)
```

```python
f_x_fcn = lambdify(x,f_x)
f_x_array = f_x_fcn(x_array)
```

```python
#和limit_seq不是一样的语法
#第三个参数必须写,表示趋近的点的一个行为,不是求值,表示趋近的点的位置
    #函数可能在某个x取到了极限,所以没有默认值(数列默认是正无穷)
#第四个参数表示趋近位置,从那一侧靠近a,'-'表示从左边靠近x<a  '+'表示从右边靠近x>a
    #不写默认为双向
    #数列只有正方向,不需要这个参数
    #极限点oo,-oo方向趋近方向没有意义
f_x_oo_limit = limit(f_x,x,oo)
f_x_oo_limit
```

```python
fig,ax = plt.subplots()

ax.plot(x_array,f_x_array,linewidth = 1.5)
ax.axhline(y = f_x_oo_limit,color = 'r')

ax.set_xlim(x_array.min(),x_array.max())
plt.xscale('log')
ax.set_xlabel(r'$\it{x}$',fontname = 'Times New Roman')
ax.set_xlabel(r'$%s$'%latex(f_x),fontname = 'Times New Roman')

plt.grid(True,which = 'both',ls = '-')
```

```python
f_x = 1 / (1 + 2**(-1/x))
f_x
```

```python
f_x_fcn = lambdify(x,f_x)
```

```python
x_array_right = np.linspace(0.01,4,500)
f_x_array_right = f_x_fcn(x_array_right)
f_x_0_limit_right = limit(f_x,x,0,'+')
f_x_0_limit_right
```

```python
x_array_left = np.linspace(-4,-0.01,500)
f_x_array_left = f_x_fcn(x_array_left)
f_x_0_limit_left = limit(f_x,x,0,'-')
f_x_0_limit_left
```

```python
fig,ax = plt.subplots()

ax.plot(x_array_right,f_x_array_right,linewidth = 1.5,color = 'b')
ax.axhline(y = f_x_0_limit_right,color = 'r')

ax.plot(x_array_left,f_x_array_left,lw = 1.5,color = 'b')
ax.axhline(y = f_x_0_limit_left,color = 'r')

ax.axvline(x = 0,color = 'k')
ax.axhline(y = 0.5,color = 'k')

ax.grid(linestyle = '--',linewidth = 0.25,color = '0.5')
ax.set_xlim(x_array_left.min(),x_array_right.max())

ax.set_xlabel(r'$\it{x}$',fontname = 'Times New Roman')
ax.set_ylabel(r'$%s$'%latex(f_x),fontname = 'Times New Roman')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from sympy import latex,lambdify,symbols
from sympy.abc import x
```

```python
def plot_secant(x0,y0,x1,y1,color):
    k = (y1 - y0)/(x1 - x0)
    x = np.linspace(-1,4,100)#局部变量x不影响外部的全局变量x
    secant_y_x = k*(x - x0) + y0
    plt.plot(x,secant_y_x,color = color,linewidth = 0.25)
```

```python
delta_Xs = np.linspace(0.1,1,10)
```

```python
f_x = x**2
f_x
```

```python
x_array = np.linspace(-1,4,100)
f_x_fcn = lambdify(x,f_x)
y_array = f_x_fcn(x_array)

x0 = 1
y0 = f_x_fcn(x0)
```

```python
fig,ax = plt.subplots(figsize = (8,8))

plt.plot(x_array,y_array,color = '#00448A',linewidth = 1.25)
plt.plot(x0,y0,color = '#92D050',marker = 'x',markersize = 12)

colors = plt.cm.RdYlBu_r(np.linspace(0,1,len(delta_Xs)))

for i in np.linspace(1,len(delta_Xs),len(delta_Xs)):
    x1 = x0 + delta_Xs[int(i) - 1]
    y1 = f_x_fcn(x1)
    plt.plot(x1,y1,color = '#00448A',marker = 'x',markersize = 12)
    plot_secant(x0,y0,x1,y1,colors[int(i) - 1])

plt.xlabel('X')  # 设置x轴标签
plt.ylabel('$y = f(x)$')  # 设置y轴标签
ax.set_title('$f(x) = %s$' % latex(f_x))  # 设置标题显示函数表达式
ax.set_xlim(0, 2)  # 设置x轴范围
ax.set_ylim(-1, 4)  # 设置y轴范围
```

```python
from sympy import latex,lambdify,diff,sin
from sympy.abc import x
import numpy as np
from matplotlib import pyplot as plt
```

```python
f_x = x**2 - 2
f_x
```

```python
x_array = np.linspace(-2,2,100)
f_x_fcn = lambdify(x,f_x)
f_x_array = f_x_fcn(x_array)
```

```python
f_x_1_diff = diff(f_x,x)
f_x_1_diff
```

```python
f_x_1_diff_fcn = lambdify(x,f_x_1_diff)#先变量才有公式,多个输入用括号括起来,变成python函数必须有输入,所以先变量
f_x_1_diff_array = f_x_1_diff_fcn(x_array)
```

```python
f_x_2_diff = diff(f_x,x,2)#这是符号运算,必须有符号公式
f_x_2_diff
```

```python
f_x_2_diff_fcn = lambdify(x,f_x_2_diff)#可以用常数函数的规则输出
f_x_2_diff_array = f_x_2_diff_fcn(x_array)#结果是常数返回可能是标量,因为二阶导数可能是标量
f_x_2_diff_array = f_x_2_diff_array + x_array*0#用标量广播一次,确保形状一致
```
