```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_vector(vector,RGB,label):
    array = np.array([[0,0,vector[0],vector[1]]])
    X,Y,U,V = zip(*array)#*作用是拆开array的外层再传入 ,zip是取传入元素的对应位置组合
    plt.quiver(X,Y,U,V,angles = 'xy',scale_units = 'xy',scale = 1,color = RGB)

    label = label + f'( {vector[0]},{vector[1]} )'
    plt.annotate(label,
                 (vector[0],vector[1]),
                 textcoords = 'offset points', #表示要做偏移
                 xytext = (0,10),#对目标点位的偏移量
                 ha = 'center')#对齐方式
```

```python
a = np.array([4,3])
i = np.array([1,0])
j = np.array([0,1])
```

```python
fig,ax = plt.subplots()

draw_vector(4*i,np.array([0,112,192])/255,'4i')
draw_vector(3*j,np.array([255,0,0])/255,'3j')

draw_vector(i,np.array([0,112,192])/255,'i')
draw_vector(j,np.array([255,0,0])/255,'j')

draw_vector(a,np.array([146,208,80])/255,'a')

plt.xlabel('$x$')
plt.ylabel('$y$')

#和ax.set_aspect('equal')类似,但是'scaled是plt层的,是对于当前ax的
plt.axis('scaled') 
ax.set_xlim([0,5])
ax.set_ylim([0,5])
ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])  # 设置网格线
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_vector(vector,RGB,label):
    array = np.array([[0,0,vector[0],vector[1]]])
    X,Y,U,V = zip(*array)#*作用是拆开array的外层再传入 ,zip是取传入元素的对应位置组合
    plt.quiver(X,Y,U,V,angles = 'xy',scale_units = 'xy',scale = 1,color = RGB)

    label = label + f'( {vector[0]},{vector[1]} )'
    plt.annotate(label,
                 (vector[0],vector[1]),
                 textcoords = 'offset points', #表示要做偏移
                 xytext = (0,10),#对目标点位的偏移量
                 ha = 'center')#对齐方式
```

```python
a = np.array([4,1])
b = np.array([1,3])
```

```python
fig,ax = plt.subplots()

draw_vector(a,np.array([0,112,192])/255,'a')
draw_vector(b,np.array([255,0,0])/255,'b')
draw_vector(a+b,np.array([146,208,80])/255,'a+b')

plt.xlabel('$x$')
plt.ylabel('$y$')

plt.axis('scaled') 
ax.set_xlim([0,5])
ax.set_ylim([0,5])
ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])  # 设置网格线
plt.show()
```

```python
fig,ax = plt.subplots()

draw_vector(a,np.array([0,112,192])/255,'a')
draw_vector(b,np.array([255,0,0])/255,'b')
draw_vector(a - b,np.array([146,208,80])/255,'a+b')

plt.xlabel('$x$')
plt.ylabel('$y$')

plt.axis('scaled') 
ax.set_xlim([0,5])
ax.set_ylim([-2,3])
ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])  # 设置网格线
plt.show()
```

```python
a = np.array([2,2])
```

```python
fig,ax = plt.subplots()

draw_vector(2*a,np.array([0,112,192])/255,'2a')
draw_vector(a,np.array([255,0,0])/255,'a')
draw_vector(0.5*a,np.array([146,208,80])/255,'0.5a')

plt.xlabel('$x$')
plt.ylabel('$y$')

plt.axis('scaled') 
ax.set_xlim([0,5])
ax.set_ylim([0,5])
ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])  # 设置网格线
plt.show()
```

```python
import numpy as np
```

```python
a = [4,1]
b = [1,3]
```

```python
a_norm = np.linalg.norm(a)
b_norm = np.linalg.norm(b)

print(a_norm)
print(b_norm)
```

```python
a_dot_b = np.dot(a,b)
print(a_dot_b)
```

```python
cos_result = a_dot_b / a_norm / b_norm
print(cos_result)
```

```python
angle = np.degrees(np.arccos(cos_result))
print(angle)
```

```python
def draw_vector(ax,vector,color,label):
    ax.quiver(0,0,0,vector[0],vector[1],vector[2],
              #arrow_length_ratio是剪头头部长度的比例
              color = color,arrow_length_ratio = 0.1,normalize = False)

    label_text = f'{label} ( {vector[0]},{vector[1]},{vector[2]})'
    ax.text(vector[0],vector[1],vector[2],label_text,
            #verticalalignment表示垂直对齐
           color = color ,verticalalignment = 'center')
```

```python
c = np.array([4,3,5])
a = np.array([4,3,0])
i = np.array([1,0,0])
j = np.array([0,1,0])
k = np.array([0,0,1])
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

draw_vector(ax,a,'blue','a')
draw_vector(ax,c,'green','c')
draw_vector(ax,i,'red','i')
draw_vector(ax,j,'purple','j')
draw_vector(ax,k,'orange','k')

ax.set_proj_type('ortho')

# 设置正交投影
ax.set_proj_type('ortho')

# 设置坐标范围
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.set_zlim(0, 5)

# 添加坐标轴标签
ax.set_xlabel(r'$\it{x}$')
ax.set_ylabel(r'$\it{y}$')
ax.set_zlabel(r'$\it{z}$')

# 设置视角
ax.view_init(azim=60, elev=20)

#控制对应轴刻度产生的网格线
ax.xaxis._axinfo['grid'].update({'linewidth' : 0.25,'linestyle':':'})
ax.yaxis._axinfo['grid'].update({'linewidth' : 0.25,'linestyle':':'})
ax.zaxis._axinfo['grid'].update({'linewidth' : 0.25,'linestyle':':'})

plt.tight_layout()
plt.show()
```
