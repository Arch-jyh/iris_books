```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches
```

```python
x = np.linspace(-4,4,num = 201)
y = np.linspace(-4,4,num = 201)

m = 1
n = 1.5

xx,yy = np.meshgrid(x,y)

e_array = np.linspace(0,3,num = 31)
```

```python
fig,ax = plt.subplots(figsize = (8,8))

colors = plt.cm.RdYlBu(np.linspace(0,1,len(e_array)))

for i in range(0,len(e_array)):
    e = e_array[i]

    ellipse = yy**2 - (e**2 - 1)*xx**2 - 2*xx #p = 1

    #这里已经降维成1D了
    color_code = colors[i,:].tolist()#转换为python列表

    #将colors传入的参数再包一层最符合语法(可能有多等高线情况)
    plt.contour(xx,yy,ellipse,levels = [0],colors = [color_code])

plt.axvline(x = 0,color = 'k',ls = '-')
plt.axhline(y = 0,color = 'k',ls = '-')
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim(-4,4)
ax.set_ylim(-4,4)
#spines这本身是一个类似字典类型的plt特殊格式,通过[[]]能够特殊索引key(仅适用于这里,或其他类字典格式)
    #返回的是容器对象(类似array),有广播功能,但是不是list和array
ax.spines[['top','right','bottom','left']].set_visible(False)
```

```python
m = 1
n = 1.5
x = np.linspace(-4,4,num = 201)
y = np.linspace(-4,4,num = 201)

xx,yy = np.meshgrid(x,y)

rho_array = np.linspace(-0.95,0,num = 20)

fig,ax = plt.subplots(figsize = (8,8))

rect = patches.Rectangle((-m,-n),2*m,2*n,
                         linewidth = 0.25,edgecolor = 'k',
                         linestyle = '--',
                         facecolor = 'none')

ax.add_patch(rect)

colors = plt.cm.RdYlBu(np.linspace(0,1,len(rho_array)))

for i in range(0,len(rho_array)):
    rho = rho_array[i]

    ellipse = ((xx / m)**2 - 2*rho*(xx/m)*(yy/n) + (yy/n)**2) / (1 - rho**2)

    color_code = colors[i,:].tolist()

    plt.contour(xx,yy,ellipse,levels = [1],colors = [color_code])

plt.axvline(x = 0,color = 'k',ls = '-')
plt.axhline(y = 0,color = 'k',ls = '-')
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim(-4,4)
ax.set_ylim(-4,4)

ax.spines[['top','right','bottom','left']].set_visible(False)
```

```python
a = 1
b = 1
p = [0.5,1,2,3]
q = p

pp,qq = np.meshgrid(p,q)
#flatten一定返回副本,ravel优先返回视图
pp = pp.flatten()
qq = qq.flatten()

x1 = np.linspace(-2,2,num = 101)
x2 = x1

xx1,xx2 = np.meshgrid(x1,x2)

fig,axes = plt.subplots(ncols = 4,nrows = 4,
                        figsize = (12,12))

#axes.flat本身就是迭代器,直接用
for p,q,ax in zip(pp,qq,axes.flat):

    #判断p是否为无穷
    if np.isinf(p):
        zz = np.maximum(np.abs(xx1/a),np.abs(xx2/b))#逐元素最大值,得到还是矩阵
    else:
        zz = ((np.abs((xx1/a)) ** p) + (np.abs((xx2/b))**q))**(1./q)

    ax.contourf(xx1,xx2,zz,20,cmap = 'RdYlBu_r')
    ax.contour(xx1,xx2,zz,[1],colors = 'k',linewidths = 2)

    plt.axvline(x = 0,color = 'k',ls = '-')
    plt.axhline(y = 0,color = 'k',ls = '-')
    ax.set_xlim(-2,2)
    ax.set_ylim(-2,2)
    ax.spines[['top','right','bottom','left']].set_visible(False)   
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    #两个纯字符串之间去掉+可以自动拼接,编译层直接支持
    #str(p)不行,编译不支持这种语法,还没到运行层,所以还是函数表达式 使用+可以正常在编译层运行
    ax.set_title('p = ' + str(p) +'     ' + 'q = ' + str(q))
    ax.set_aspect('equal',adjustable = 'box')#为box后续调整将保持数据范围不变,只改变外观盒子

plt.show()
```

```python
a = 2
b = 1
p = [0.02,0.2,0.5,1,1.5,2,4,8,np.inf]
q = p

pp,qq = np.meshgrid(p,q)
pp = pp.flatten()
qq = qq.flatten()

x1 = np.linspace(-2,2,num = 101)
x2 = x1

xx1,xx2 = np.meshgrid(x1,x2)

fig,axes = plt.subplots(ncols = 9,nrows = 9,
                        figsize = (12,12))

for p,q,ax in zip(pp,qq,axes.flat):

    if np.isinf(p):
        zz = np.maximum(np.abs(xx1/a),np.abs(xx2/b))
    else:
        zz = ((np.abs((xx1/a)) ** p) + (np.abs((xx2/b))**q))**(1./q)

    ax.contourf(xx1,xx2,zz,20,cmap = 'RdYlBu_r')
    ax.contour(xx1,xx2,zz,[1],colors = 'k',linewidths = 2)

    plt.axvline(x = 0,color = 'k',ls = '-')
    plt.axhline(y = 0,color = 'k',ls = '-')
    ax.set_xlim(-2,2)
    ax.set_ylim(-2,2)
    ax.spines[['top','right','bottom','left']].set_visible(False)   
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    ax.set_title('p = ' + str(p) +'     ' + 'q = ' + str(q))
    ax.set_aspect('equal',adjustable = 'box')

plt.show()
```
