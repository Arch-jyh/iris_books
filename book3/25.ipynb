```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def draw_vector(vector,RGB,ax):
    ax.quiver(0,0,vector[0],vector[1],
              angles = 'xy',
              scale_units = 'xy',
              scale = 1,
              color = RGB)
```

```python
T = np.matrix([[0.7,0.2],
               [0.3,0.8]])
```

```python
all_max = 1
all_min = 0
```

```python
x1 = np.linspace(-1.1,1.1,num = 201)
x2 = x1
xx1,xx2 = np.meshgrid(x1,x2)
zz = ((np.abs((xx1))**2) + (np.abs((xx2))**2)) **(1./2)
```

```python
pi_array = np.vstack((np.linspace(1,0,11),1 - np.linspace(1,0,11)))
pi_array = np.matrix(pi_array)
num_steps = 12
```

```python
colors = plt.cm.rainbow(np.linspace(0,1,num_steps + 1))
```

```python
for ini in np.arange(0,np.shape(pi_array)[1]):
    pi = pi_array[:,ini]
    fig,ax = plt.subplots(figsize = (10,10))

    plt.plot(x1,1 - x1,color = 'k',
    linestyle = '--')

    plt.contour(xx1,xx2,zz,levels = [1], #levels是列表时候,画出对应位置的等高线
               colors = 'k',linestyles = '--')

    for i in np.arange(0,num_steps + 1):
        draw_vector(pi / np.linalg.norm(pi),colors[i],ax) #归一化向量
        draw_vector(pi,colors[i],ax)#原始向量
        ax.tick_params(left = False,bottom = False)#隐藏刻度线
        ax.set_xlim(-1.1, 1.1)  # 设置x轴的范围为[-1.1, 1.1]
        ax.set_ylim(-1.1, 1.1)  # 设置y轴的范围为[-1.1, 1.1]
        ax.axvline(x=0, color='k')  # 绘制y轴
        ax.axhline(y=0, color='k')  # 绘制x轴
        ax.spines['top'].set_visible(False)  # 隐藏上边框
        ax.spines['right'].set_visible(False)  # 隐藏右边框
        ax.spines['bottom'].set_visible(False)  # 隐藏下边框
        ax.spines['left'].set_visible(False)  # 隐藏左边框
        ax.grid(color=[0.8, 0.8, 0.8])  # 设置网格颜色为灰色
        plt.xticks(np.linspace(-1, 1, 21))  # 设置x轴刻度，范围为[-1, 1]，共21个刻度
        plt.yticks(np.linspace(-1, 1, 21))  # 设置y轴刻度，范围为[-1, 1]，共21个刻度
        pi = T@pi
```
