```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1 = 10**x1
f2 = x1
f3 = np.log(x)#直接使用log自动计算ln自然对数
```

```python
fig,ax = plt.subplots()
plt.plot(x1,f1,color = 'r')
plt.plot(x1,f2,color = 'k')
plt.plot(x,f3,color = 'b')
plt.xlim(-1,10)
plt.ylim(-1,10)
ax.grid(which = 'major',ls = '-',lw = 0.5,color = '0.5')
ax.grid(which = 'minor',ls = '-',lw = 0.3,color = '0.8')
plt.tight_layout()
#这里保证绘图的box的外侧边长横纵比为1
#不完全等价但效果相同 ax.set_aspect('equal',adjustable = 'box')这是保证x,y单位等长
ax.set_box_aspect(1)
```

```python
#base是对数的底数,等比按照10倍增长
x_log_1 = np.logspace(np.log10(0.1),np.log10(3),num = 100,endpoint = True,base = 10.0)
#log10(base*k) 这里的base就是10,可以看出来log10(0.1) = -1>>>k = -1  log10(1000) = 3 k = 3 将k在-1到3分割100次
#因为人习惯线性感觉,将指数变化为了等差
#这里的输入min,max不是x本身,是x的指数
#这里的输出不再是等差的k而是x  x_i = base^(k_i)   -1>>>3是等差的 ▲k = 4/99
x_log_2 = np.logspace(np.log10(0.1),np.log10(1000),num = 100,endpoint = True,base = 10.0)
x_log_3 = np.logspace(np.log10(1),np.log10(1000),num = 100,endpoint = True,base = 10.0)
```

```python
f1 = 10**x_log_1
f2 = x_log_2
f3 = np.log(x_log_3)

fig,ax = plt.subplots()
plt.plot(x_log_1,f1,color = 'r')
plt.plot(x_log_2,f2,color = 'k')
plt.plot(x_log_3,f3,color = 'b')

plt.xscale('log')
plt.yscale('log')
plt.xlim(0.1,1000)
plt.ylim(0.1,1000)
ax.grid(which = 'major',ls = '-',lw = 0.5,color = '0.5')
ax.grid(which = 'minor',ls = '-',lw = 0.3,color = '0.8')
plt.tight_layout()
ax.set_box_aspect(1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy.abc import x
from sympy import exp,lambdify
```

```python
def plot_curve(x_array,y_array,x_array_new,y_array_new):
    fig,ax = plt.subplots()

    plt.plot(x_array,y_array,color = '#0070C0',label = 'Original')
    #不指定默认与y = 0之间填充,如果指定了第二条线作为之间的界限 参数就是x,y1,y2添加第二个y
    ax.fill_between(x_array,y_array,edgecolor = 'none',facecolor = '#0070C0',alpha = 0.2)
    plt.plot(x_array_new,y_array_new,color = 'r',label = 'Transformed')

    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(y = 0,color = 'k',linestyle = '-')
    plt.axvline(x = 0,color = 'k',linestyle = '-')
    plt.xticks(np.arange(-4,5,step = 1))
    plt.yticks(np.arange(-4,5,step = 1))
    plt.axis('scaled')

    ax.set_xlim(-4,4)
    ax.set_ylim(-4,4)
    ax.spines[['top','right','bottom','left']].set_visible(False)

    plt.legend()
    ax.grid(ls = '--',lw = 0.25,color = '0.5')
    #字符串只是键名,其中的'.'仅仅是命名规则
    plt.rcParams['figure.figsize'] = [7.5,3.5]
    plt.rcParams['figure.autolayout'] = True
```

```python
x_array = np.arange(-4,4+0.01,step = 0.01)
f_x = 2*exp(-(x - 1)**2)
f_x_fcn = lambdify([x],f_x)
f_x_array = f_x_fcn(x_array)
```

```python
for c in [2,-3]:#两个变换因子
    f_x_array_new = f_x_array + c
    plot_curve(x_array,f_x_array,x_array,f_x_array_new)
```

```python
for c in [3,-1]:
    f_x_new = 2*exp(-((x+c) - 1)**2)
    f_x_new_fcn = lambdify([x],f_x_new)
    f_x_array_new = f_x_new_fcn(x_array)
    plot_curve(x_array,f_x_array,x_array,f_x_array_new)
```

```python
for c in [1/2,2]:
    f_x_array_new = c*f_x_array
    plot_curve(x_array,f_x_array,x_array,f_x_array_new)
```

```python
for c in [1/2,2]:
    f_x_new = 2*exp(-(c*x - 1)**2)
    f_x_new_fcn = lambdify([x],f_x_new)
    f_x_array_new = f_x_new_fcn(x_array)
    plot_curve(x_array,f_x_array,x_array,f_x_array_new)
```

```python
f_x_array_new = -f_x_array
plot_curve(x_array,f_x_array,x_array,f_x_array_new)
```

```python
f_x_new = 2*exp(-(-x - 1)**2)
f_x_new_fcn = lambdify([x],f_x_new)
f_x_array_new = f_x_new_fcn(x_array)
plot_curve(x_array,f_x_array,x_array,f_x_array_new)
```
