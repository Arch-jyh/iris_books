```python
import matplotlib.pyplot as plt
#他已经实现封装好的参数化模板,直接用,里面存储了所需数据
from matplotlib.patches import RegularPolygon,Circle
import numpy as np
```

```python
for num_vertices in [3,4,5,6,7,8]:
    fig,ax = plt.subplots()#创建fig后,自动配置新的fig为当前fig(活跃的)
    ax.set_aspect('equal')
    #这里是快封装好的,收藏品自定义更多一些,但生成的还是pathc对象,但也好过从头创建
    hexagon_inner = RegularPolygon((0,0),numVertices = num_vertices,
                                  radius = 1,alpha = 0.2,edgecolor = 'k')#alpha取值范围为0-1
    #添加图形对象
        #前提是Patch类
        #收藏品是快配置好的Patch类,期间可以配置需要的参数,无需从头配置,配置好后还是Patch类,使用这个函数添加
    #这里添加的是已经配置好的
    ax.add_patch(hexagon_inner)
    plt.axis('off')

    plt.xlim(-1.5,1.5)
    plt.ylim(-1.5,1.5)

    plt.show()#这个只显示当前fig,但是juputer会将所有的fig都显示出来
```
