```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[1.25,-0.75],
              [-0.75,1.25]])
```

```python
#这里的xx2注意 -8,8这里的是从上到下的,所以第一行的坐标是x-8,8  y= -8
    #所以后面需要翻转角度数据
xx1,xx2 = np.meshgrid(np.linspace(-8,8,9),np.linspace(-8,8,9))
 #product 乘积的意思,意思是将所以的行数量和列的数量相乘,计算有多少个元素
num_vecs = np.prod(xx1.shape)
```

```python
thetas = np.linspace(0,2*np.pi,num_vecs)
thetas = np.reshape(thetas,(-1,9))
thetas = np.flipud(thetas) #flip up-down的意思上下翻转
```

```python
uu = np.cos(thetas)
vv = np.sin(thetas)
```

```python
fig,ax = plt.subplots()
ax.quiver(xx1,xx2,uu,vv,
          angles = 'xy',scale_units = 'xy',scale = 1,
          edgecolor = 'none',facecolor = 'b')

plt.ylabel('$x_2$')  # 设置 y 轴标签
plt.xlabel('$x_1$')  # 设置 x 轴标签
plt.axis('scaled')  # 保持坐标轴比例
ax.set_xlim([-10, 10])  # 设置 x 轴范围
ax.set_ylim([-10, 10])  # 设置 y 轴范围
ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])  # 添加网格线
ax.set_xticks(np.linspace(-10, 10, 11))  # 设置 x 轴刻度
ax.set_yticks(np.linspace(-10, 10, 11))  # 设置 y 轴刻度
plt.show()
```

```python
#flatten一定复制了一份数组ravel不一定
V = np.array([uu.flatten(),vv.flatten()]).T
```

```python
W = V@A
```

```python
uu_new = np.reshape(W[:,0],(-1,9))
vv_new = np.reshape(W[:,1],(-1,9))
```

```python
fig,ax = plt.subplots()
ax.quiver(xx1,xx2,uu,vv,
          angles = 'xy',scale_units = 'xy',scale = 1,
         edgecolor = 'none',facecolor = 'b')
ax.quiver(xx1,xx2,uu_new,vv_new,
          angles = 'xy',scale_units = 'xy',scale = 1,
          edgecolor = 'none',facecolor = 'r')

plt.ylabel('$x_2$')
plt.xlabel('$x_1$')
plt.axis('scaled')
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])
ax.set_xticks(np.linspace(-10, 10, 11))
ax.set_yticks(np.linspace(-10, 10, 11))
plt.show()
```

```python
fig,ax = plt.subplots()
ax.quiver(xx1*0,xx2*0,uu,vv,
          angles = 'xy',scale_units = 'xy',scale = 1,
         edgecolor = 'none',facecolor = 'b')
ax.quiver(xx1*0,xx2*0,uu_new,vv_new,
          angles = 'xy',scale_units = 'xy',scale = 1,
          edgecolor = 'none',facecolor = 'r')

plt.ylabel('$x_2$')
plt.xlabel('$x_1$')
plt.axis('scaled')
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])
ax.set_xticks(np.linspace(-2, 2, 5))
ax.set_yticks(np.linspace(-2, 2, 5))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def visualize(X_circle,X_vec,title_txt):
    fig,ax = plt.subplots()
    plt.plot(X_circle[0,:],X_circle[1,:],'k',linestyle = '--',linewidth = 0.5)

    plt.quiver(0,0,X_vec[0,0],X_vec[1,0],
              angles = 'xy',scale_units = 'xy',scale = 1,
              color = [0,0.4392,0.7529])

    plt.quiver(0,0,X_vec[0,1],X_vec[1,1],
              angles = 'xy',scale_units = 'xy',scale = 1,
              color = [1,0,0])

    plt.axvline(x=0, color='k', zorder=0)  # 绘制 y 轴
    plt.axhline(y=0, color='k', zorder=0)  # 绘制 x 轴

    plt.ylabel('$x_2$')  # y 轴标签
    plt.xlabel('$x_1$')  # x 轴标签

    ax.set_aspect(1)  # 设置坐标比例
    ax.set_xlim([-2.5, 2.5])  # x 轴范围
    ax.set_ylim([-2.5, 2.5])  # y 轴范围
    ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])  # 添加网格线
    ax.set_xticks(np.linspace(-2, 2, 5))  # 设置 x 轴刻度
    ax.set_yticks(np.linspace(-2, 2, 5))  # 设置 y 轴刻度
    plt.title(title_txt)  # 设置标题
    plt.show()
```

```python
theta = np.linspace(0,2*np.pi,100)

circle_x1 = np.cos(theta)
circle_x2 = np.sin(theta)

V_vec = np.array([[np.sqrt(2) / 2,-np.sqrt(2) / 2],
                  [np.sqrt(2) / 2,np.sqrt(2) / 2]])

X_circle = np.array([circle_x1,circle_x2])

visualize(X_circle,V_vec,'Original')
```

```python

A = np.array([[1.25,-0.75],
              [-0.75,1.25]])

visualize(A @ X_circle,A @ V_vec,'A')
```

```python
lambdas,V = np.linalg.eig(A)
```

```python
lambdas
```

```python
V
```

```python
D = np.diag(np.flip(lambdas)) #换方向,因为向量是右乘(默认的数据是左乘)
```

```python
V = V.T #取对应特征值的V
```

```python
print(D)
print(V)
```

```python
visualize(V.T @ X_circle,V.T @ V_vec,'V')
```

```python
visualize(D @ V.T @ X_circle,D @ V.T @ V_vec,'$\u039BV^T$')
```

```python
visualize(V@ D @ V.T @ X_circle,V @ D @ V.T @ V_vec,'$V\u039BV^T$')
```

```python
visualize(A@ X_circle,A @ V_vec,'A')
```
