```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import hsv_to_rgb
```

```python
thetas = np.linspace(0,np.pi,25)
```

```python
x = np.array([4,3])
```

```python
fig,ax = plt.subplots()

for i,theta in enumerate(thetas):
    v = np.array([np.cos(theta),np.sin(theta)])#单位向量

    proj = np.dot(v,x)

    hue = theta / np.pi #将弧度归一化
    color = hsv_to_rgb([hue,1,1])

    ax.plot([-v[0]*6,v[0]*6],[-v[1]*6,v[1]*6],color = color)

    proj_point = proj * v
    ax.plot([x[0],proj_point[0]],[x[1],proj_point[1]],color = color,linestyle = '-')

    ax.plot(proj_point[0],proj_point[1],marker = 'x',color = color)

    ax.quiver(0,0,v[0],v[1],angles = 'xy',scale_units = 'xy',scale = 1,color = color)

ax.plot(x[0],x[1],marker = 'x',color = 'r',label = 'Vector x')
ax.axhline(y = 0,color = 'k')
ax.axvline(x = 0,color = 'k')
ax.set_aspect('equal','box')
ax.axis(False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = plt.rcParams  # 获取默认绘图参数
p["font.weight"] = "light"  # 设置字体粗细为 light
p["ytick.minor.visible"] = True  # 显示 y 轴的小刻度
p["xtick.minor.visible"] = True  # 显示 x 轴的小刻度
p["axes.grid"] = True  # 启用坐标轴网格
p["grid.color"] = "0.5"  # 设置网格颜色为灰色
p["grid.linewidth"] = 0.5  # 设置网格线宽度
```

```python
thetas = np.array([0,15,30,45,60,75,90,120,135])
```

```python
x = np.array([[4],
              [3]])
```

```python
i = 1
```

```python
fig = plt.figure(figsize = (12,12))

for theta in thetas:

    theta = theta / 180 * np.pi
    ax = fig.add_subplot(3,3,i)

    v1 = np.array([[np.cos(theta)],[np.sin(theta)]])
    proj = v1.T @ x

    plt.plot([-v1[0,0] * 6,v1[0,0]*6],[-v1[1,0]*6,v1[1,0]*6])
    plt.plot([x[0,0],v1[0,0]*proj[0,0]],[x[1,0],v1[1,0]*proj[0,0]],color = 'k',linestyle = '--')
    plt.plot(v1[0,0]*proj[0,0],v1[1,0]*proj[0,0],color = 'k',marker = 'x')

    plt.quiver(0,0,v1[0,0],v1[1,0],angles = 'xy',scale_units = 'xy',scale = 1,color = 'b')

    v2 = np.array([[-np.sin(theta)],[np.cos(theta)]])
    proj = v2.T @ x
    
    plt.plot([-v2[0,0] * 6,v2[0,0]*6],[-v2[1,0]*6,v2[1,0]*6])
    plt.plot([x[0,0],v2[0,0]*proj[0,0]],[x[1,0],v2[1,0]*proj[0,0]],color = 'k',linestyle = '--')
    plt.plot(v2[0,0]*proj[0,0],v2[1,0]*proj[0,0],color = 'k',marker = 'x')

    plt.quiver(0,0,v2[0,0],v2[1,0],angles = 'xy',scale_units = 'xy',scale = 1,color = 'r')

    plt.axhline(y = 0,color = 'k')
    plt.axvline(x = 0,color = 'k')
    plt.plot(x[0,0],x[1,0],marker = 'x',color = 'r')
    plt.quiver(0,0,x[0,0],x[1,0],angles = 'xy',scale_units = 'xy',scale = 1,color = 'k')

    plt.axis('scaled')  # 使坐标轴比例相等
    ax.grid(linestyle='--', linewidth=0.25, color=[0.75, 0.75, 0.75])  # 设置网格样式
    plt.xlim([-6, 6])  # 设置 x 轴范围
    plt.ylim([-6, 6])  # 设置 y 轴范围
    plt.xticks(np.linspace(-6, 6, 13))  # 设置 x 轴刻度
    plt.yticks(np.linspace(-6, 6, 13))  # 设置 y 轴刻度

    i = i+1
```
