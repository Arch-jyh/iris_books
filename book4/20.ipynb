```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1.5
b = 1
```

```python
x1 = np.linspace(-3,3,200)
x2 = np.linspace(-3,3,200)
xx1,xx2 = np.meshgrid(x1,x2)
```

```python
fig,ax = plt.subplots()
theta_array = np.linspace(0,2*np.pi,100)
plt.plot(a*np.cos(theta_array),b*np.sin(theta_array),color = 'k') #
colors = plt.cm.RdYlBu(np.linspace(0,1,len(theta_array)))
for i in range(len(theta_array)):
    theta = theta_array[i]

    p1 = a*np.cos(theta)
    p2 = b*np.sin(theta)

    tangent = p1*xx1 / a**2 + p2*xx2 / b**2 - p1**2/a**2 - p2**2 / b**2 #切线方程

    colors_i = colors[int(i),:]

    #color和level的[]的元素需要对应
    ax.contour(xx1,xx2,tangent,levels = [0],colors = [colors_i])

## 设置图形显示属性
plt.axis('scaled')  # 保持坐标轴比例一致
ax.set_xlim(-3, 3)  # 设置 x 轴范围
ax.set_ylim(-3, 3)  # 设置 y 轴范围
ax.spines['top'].set_visible(False)  # 隐藏顶部边框
ax.spines['right'].set_visible(False)  # 隐藏右侧边框
ax.spines['bottom'].set_visible(False)  # 隐藏底部边框
ax.spines['left'].set_visible(False)  # 隐藏左侧边框
ax.axvline(x=0, color='k')  # 绘制 y 轴
ax.axhline(y=0, color='k')  # 绘制 x 轴
```
