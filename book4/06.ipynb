```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
def plot_heatmap(x,title):
    fig,ax = plt.subplots()
    ax = sns.heatmap(
        x,
        cmap = 'RdYlBu_r',
        cbar_kws = {'orientation':'horizontal'},
        vmin = -1,vmax = 1
    )
    ax.set_aspect('equal')
    plt.title(title)
```

```python
A = np.random.randint(0,40+1,size = (6,4))
B = np.random.randint(0,40+1,size = (4,3))

A = A/20 - 1
B = B/20 - 1
```

```python
plot_heatmap(A,'A')
```

```python
plot_heatmap(B,'B')
```

```python
C = A @ B
```

```python
plot_heatmap(C,'C = AB')
```

```python
C_rep = np.zeros_like(C)
```

```python
for i in np.arange(4):
    C_i = A[:,[i]] @ B[[i],:]
    title = 'C' + str(i + 1)
    plot_heatmap(C_i,title)
    C_rep = C_rep + C_i
```

```python
plot_heatmap(C_rep,'C reproduced')
```
