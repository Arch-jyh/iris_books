```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[1,0],
              [0,1]])
```

```python
L_T = np.array([[1,2/3],
                [0,1]],dtype = 'f')
B = np.array([[1.5,0],
              [0,np.sqrt(3)]],dtype = 'f')
```

```python
a_1 = A[:,0]
a_2 = A[:,1]

a_1_t1 = L_T @ a_1
a_1_t2 = B @ a_1_t1

a_2_t2 = L_T @ a_2
a_2_t2 = B @ a_2_t1
```

```python
fig,axs = plt.subplots(1,3,figsize = (15,5))

plt.sca(axs[0])
plt.quiver(0,0,a_1[0],a_1[1],color = 'b',angles = 'xy',scale_units = 'xy',scale = 1)
plt.quiver(0,0,a_2[0],a_2[1],color = 'r',angles = 'xy',scale_units = 'xy',scale = 1)
plt.axis('scaled')
plt.xlim(-2,2)
plt.ylim(-2,2)


plt.sca(axs[1])
plt.quiver(0,0,a_1_t1[0],a_1_t1[1],color = 'b',angles = 'xy',scale_units = 'xy',scale = 1)
plt.quiver(0,0,a_2_t1[0],a_2_t1[1],color = 'r',angles = 'xy',scale_units = 'xy',scale = 1)
plt.axis('scaled')
plt.xlim(-2,2)
plt.ylim(-2,2)


plt.sca(axs[2])
plt.quiver(0,0,a_1_t2[0],a_1_t2[1],color = 'b',angles = 'xy',scale_units = 'xy',scale = 1)
plt.quiver(0,0,a_2_t2[0],a_2_t2[1],color = 'r',angles = 'xy',scale_units = 'xy',scale = 1)
plt.axis('scaled')
plt.xlim(-2,2)
plt.ylim(-2,2)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D #现在不需要写了
import numpy as np
import math as m
```

```python
cos_theta_12 = np.cos(m.radians(120))
cos_theta_13 = np.cos(m.radians(120))
#角度转换为弧度再输入
cos_theta_23 = np.cos(m.radians(90))
```

```python
P = np.array([[1,cos_theta_12,cos_theta_13],
              [cos_theta_12,1,cos_theta_23],
              [cos_theta_13,cos_theta_23,1]])
```

```python
L = np.linalg.cholesky(P)
R = L.T
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

plt.plot(0,0,0,color = 'r',marker = 'x',
         markersize = 12)
colors = ['b','r','g']
for i in np.arange(0,3):
    vector = R[:,i]
    v = np.array([vector[0],vector[1],vector[2]])
    vlength = np.linalg.norm(v)
    ax.quiver(0,0,0,v[0],v[1],v[2],
             length = vlength,color = colors[i])

ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
ax.set_zlim([-1,1])

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.zaxis.set_ticklabels([])
ax.view_init(35,60)
ax.set_proj_type('ortho')
ax.set_box_aspect([1,1,1])
```
