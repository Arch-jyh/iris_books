```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris_sns = sns.load_dataset('iris')
iris = load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
feature_names = ['Sepal length,x1','Sepal width,x2',
                 'Petal length,x3','Petal width x4']
```

```python
X_df = pd.DataFrame(X,columns = feature_names)
```

```python
sns.set_style('ticks')
```

```python
X = X_df.to_numpy()
```

```python
fig,ax = plt.subplots()
ax = sns.heatmap(X,
                 cmap = 'RdYlBu_r',
                 xticklabels = list(X_df.columns),
                 cbar_kws = {'orientation':'vertical'},
                 vmin = -1,vmax = 9)
plt.title('X')
```

```python
v_1 = np.ones((len(X),1))
E_X = v_1.T @ X / len(X)
```

```python
X_demean = X_df.sub(X_df.mean())#mean默认axis = 0,sub减法自动广播对齐
```

```python
fig,ax = plt.subplots()
ax = sns.heatmap(X_demean,
                 cmap = 'RdYlBu_r',
                 xticklabels = list(X_df.columns),
                 cbar_kws = {'orientation':'vertical'},
                 vmin = -3,vmax = 3)
plt.title('$X_{demean}$')
```

```python
SIGMA = X_df.cov()
```

```python
SIGMA_inv = pd.DataFrame(data = np.linalg.inv(SIGMA.to_numpy()),columns = SIGMA.columns)
```

```python
SIGMA_inv
```

```python
X_stand = X_demean.to_numpy() @ SIGMA_inv.to_numpy()
```

```python
fig,ax = plt.subplots()
ax = sns.heatmap(X_stand,
                 cmap = 'RdYlBu_r',
                 xticklabels = list(X_df.columns),
                 cbar_kws = {'orientation':'vertical'})
plt.title('$X_{stand}$')
```

```python
SSD = (np.linalg.norm(X - E_X,axis = 1) ** 2).sum()
```

```python
fig,ax = plt.subplots()
sns.kdeplot(data = X_demean,fill = True,
            common_norm = False,
            alpha = 0.3,linewidth = 1,
            palette = 'viridis')
plt.title('Distribution of $X_{demean}$ columns')
```

```python
SIGMA = X_df.cov()
```

```python
fig,axs = plt.subplots()
h = sns.heatmap(SIGMA,cmap = 'RdBu_r',linewidth = .05,annot = True)
h.set_aspect('equal')
h.set_title(r'$\Sigma$')
```

```python
RHO = X_df.corr()
```

```python
fig,axs = plt.subplots()
h = sns.heatmap(RHO,cmap = 'RdBu_r',linewidth = .05,annot = True)
h.set_aspect('equal')
h.set_title('$\u03A1$')
```

```python
f,(ax1,ax2,ax3) = plt.subplots(1,3,sharey = True)
g1 = sns.heatmap(X_df[y == 0].cov(),cmap = 'RdYlBu_r',
                 annot = True,cbar = False,ax = ax1,square = True,
                 vmax = 0.4,vmin = 0)
ax1.set_title('Y = 0,setosa')

g2 = sns.heatmap(X_df[y == 1].cov(),cmap = 'RdYlBu_r',
                 annot = True,cbar = False,ax = ax2,square = True,
                 vmax = 0.4,vmin = 0)
ax2.set_title('Y = 1,versicolor')

g2 = sns.heatmap(X_df[y == 2].cov(),cmap = 'RdYlBu_r',
                 annot = True,cbar = False,ax = ax3,square = True,
                 vmax = 0.4,vmin = 0)
ax3.set_title('Y = 2,virginica')
```

```python
f,(ax1,ax2,ax3) = plt.subplots(1,3,sharey = True)
g1 = sns.heatmap(X_df[y == 0].corr(),cmap = 'RdYlBu_r',
                 annot = True,cbar = False,ax = ax1,square = True,
                 vmax = 0.4,vmin = 0)
ax1.set_title('Y = 0,setosa')

g2 = sns.heatmap(X_df[y == 1].corr(),cmap = 'RdYlBu_r',
                 annot = True,cbar = False,ax = ax2,square = True,
                 vmax = 0.4,vmin = 0)
ax2.set_title('Y = 1,versicolor')

g2 = sns.heatmap(X_df[y == 2].corr(),cmap = 'RdYlBu_r',
                 annot = True,cbar = False,ax = ax3,square = True,
                 vmax = 0.4,vmin = 0)
ax3.set_title('Y = 2,virginica')
```
