```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_vector(vector,RGB):
    array = np.array([[0,0,vector[0],vector[1]]])
    X,Y,U,V = zip(*array)
    plt.quiver(X,Y,U,V,angles = 'xy',scale_units = 'xy',scale = 1,color = RGB)
```

```python
fig,ax = plt.subplots()

draw_vector([4,3],np.array([0,112,192])/255)
draw_vector([-3,4],np.array([255,0,0])/255)

## 图形显示设置
plt.ylabel('$x_2$')  # 设置y轴标签为x_2
plt.xlabel('$x_1$')  # 设置x轴标签为x_1
plt.axis('scaled')  # 设置坐标轴比例
ax.set_xlim([-5, 5])  # 设置x轴显示范围
ax.set_ylim([-5, 5])  # 设置y轴显示范围
ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])  # 添加网格线，设置样式

plt.show()  # 显示图形
```

```python
x1 = np.linspace(-10,10,num = 201)
x2 = x1
```

```python
xx1,xx2 = np.meshgrid(x1,x2)
```

```python
p = 2
zz = ((np.abs((xx1))**p) + (np.abs((xx2))**p)) **(1./p)
```

```python
fig,ax = plt.subplots(figsize = (8,8))

ax.contour(xx1,xx2,zz,levels = np.arange(11),cmap = 'RdYlBu_r')

## 绘制坐标轴和边框设置
ax.axhline(y=0, color='k', linewidth=0.25)  # 绘制y轴
ax.axvline(x=0, color='k', linewidth=0.25)  # 绘制x轴
ax.set_xlim(-12, 12)  # 设置x轴范围为-12到12
ax.set_ylim(-12, 12)  # 设置y轴范围为-12到12
ax.spines['top'].set_visible(False)  # 隐藏顶部边框
ax.spines['right'].set_visible(False)  # 隐藏右侧边框
ax.spines['bottom'].set_visible(False)  # 隐藏底部边框
ax.spines['left'].set_visible(False)  # 隐藏左侧边框
ax.set_xlabel('$x_1$')  # 设置x轴标签
ax.set_ylabel('$x_2$')  # 设置y轴标签
ax.set_aspect('equal', adjustable='box')  # 设置坐标轴比例为相等
plt.show()  # 显示绘图
```

```python
a,b = np.array([[4],[3]]) , np.array([[5],[-2]])
```

```python
cos_theta = (a.T @ b) / (np.linalg.norm(a,2) * np.linalg.norm(b,2))
cos_theta
```

```python
cos_radian = np.arccos(cos_theta)
cos_radian
```

```python
cos_degree = cos_radian * (180 / np.pi)
cos_degree
```
