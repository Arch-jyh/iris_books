```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
PRECISION = 3
```

```python
def svd(X,full_matrices):
    #注意svd返回的是Vh = v.T,不是V
    U,s,Vt = np.linalg.svd(X,full_matrices = full_matrices)

    if full_matrices:
        S = np.zeros(X.shape)
        np.fill_diagonal(S,s)
    else:
        S = np.diag(s)
    #这里将Vt.T转置 得到V
    return np.round(U,PRECISION),np.round(S,PRECISION),np.round(Vt.T,PRECISION)
```

```python
np.random.seed(1)
X = np.random.randn(6,4)
```

```python
all_max = 2
all_min = -2
```

```python
U,S,V = svd(X,full_matrices = True)
```

```python
fig,axs = plt.subplots(1,4,figsize = (12,3))

plt.sca(axs[0])
ax = sns.heatmap(X,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('X')

plt.sca(axs[1])
ax = sns.heatmap(U,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('U')

plt.sca(axs[2])
ax = sns.heatmap(S,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('S')

plt.sca(axs[3])
ax = sns.heatmap(V.T,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('V.T')
```

```python
U,S,V = svd(X,full_matrices = False)
fig,axs = plt.subplots(1,4,figsize = (12,3))

plt.sca(axs[0])
ax = sns.heatmap(X,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('X')

plt.sca(axs[1])
ax = sns.heatmap(U,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('U')

plt.sca(axs[2])
ax = sns.heatmap(S,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('S')

plt.sca(axs[3])
ax = sns.heatmap(V.T,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('V.T')
```

```python
import copy
X_rank_3 = copy.deepcopy(X)
X_rank_3[:,3] = X_rank_3[:,0] + X_rank_3[:,1]
```

```python
U_rank_3,S_rank_3,V_rank_3 = svd(X_rank_3,full_matrices = False)
```

```python
fig,axs = plt.subplots(1,4,figsize = (12,3))

plt.sca(axs[0])
ax = sns.heatmap(X_rank_3,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('X_rank_3')

plt.sca(axs[1])
ax = sns.heatmap(U_rank_3,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('U_rank_3')

plt.sca(axs[2])
ax = sns.heatmap(S_rank_3,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('S_rank_3')

plt.sca(axs[3])
ax = sns.heatmap(V_rank_3.T,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('V_rank_3.T')
```

```python
num_p = 2
```

```python
U_truc = U[:,0:num_p]
S_truc = S[0:num_p,0:num_p]
V_truc = V[:,0:num_p]

X_hat = U_truc @ S_truc @ (V_truc.T)
```

```python
fig,axs = plt.subplots(1,4,figsize = (12,3))

plt.sca(axs[0])
ax = sns.heatmap(X_hat,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('X_hat')

plt.sca(axs[1])
ax = sns.heatmap(U_truc,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('U_truc')

plt.sca(axs[2])
ax = sns.heatmap(S_truc,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('S_truc')

plt.sca(axs[3])
ax = sns.heatmap(V_truc.T,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('V_truc.T')
```

```python
u1_outer_v1 = np.outer(U[:,0][:,None],V[:,0][:,None])
u2_outer_v2 = np.outer(U[:,1][:,None],V[:,1][:,None])
u3_outer_v3 = np.outer(U[:,2][:,None],V[:,2][:,None])
u4_outer_v4 = np.outer(U[:,3][:,None],V[:,3][:,None])
```

```python
fig,axs = plt.subplots(1,4,figsize = (12,3))

plt.sca(axs[0])
ax = sns.heatmap(u_1_outer_v1,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('u_1_outer_v1')

plt.sca(axs[1])
ax = sns.heatmap(u_2_outer_v2,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('u_2_outer_v2')

plt.sca(axs[2])
ax = sns.heatmap(u_3_outer_v3,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('u_3_outer_v3')

plt.sca(axs[3])
ax = sns.heatmap(u_4_outer_v4,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('u_4_outer_v4')
```

```python
X_1 = S[0,0] * u1_outer_v1
X_2 = S[1,1] * u2_outer_v2
X_3 = S[2,2] * u3_outer_v3
X_4 = S[3,3] * u4_outer_v4
```

```python
fig,axs = plt.subplots(1,4,figsize = (12,3))

plt.sca(axs[0])
ax = sns.heatmap(X_1,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('X_1')

plt.sca(axs[1])
ax = sns.heatmap(X_2,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('X_2')

plt.sca(axs[2])
ax = sns.heatmap(X_3,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('X_3')

plt.sca(axs[3])
ax = sns.heatmap(X_4,cmap = 'RdBu_r',vmax = all_max,vmin = all_min,
                 cbar_kws = {'orientation' : 'horizontal'})
ax.set_aspect('equal')
plt.title('X_4')
```
