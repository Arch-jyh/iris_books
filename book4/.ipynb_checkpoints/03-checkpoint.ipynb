```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
p_values = [0.05,0.2,0.5,1,1.5,2,4,8,np.inf]
```

```python
x1 = np.linspace(-2.5,2.5,num = 101)
x2 = x1

xx1,xx2 = np.meshgrid(x1,x2)
```

```python
fig,axes = plt.subplots(ncols = 3,nrows = 3,figsize = (12,12))

for p,ax in zip(p_values,axes.flat): #.flat本身就是一个创建axes时候配置好了的迭代器,这里直接调用
    if np.isinf(p):
        zz = np.maximum(np.abs(xx1),np.abs(xx2))
    else:
        zz = ((np.abs((xx1))**p) + (np.abs((xx2))**p))**(1./p)#1.是小数,1.0,0省略了

    ax.contourf(xx1,xx2,zz,20,cmap = 'RdYlBu_r')
    ax.contour(xx1,xx2,zz,[1],colors = 'k',linewidths = 2) #注意等高线绘制函数线的颜色和宽度是复数

    ## 图形装饰
    ax.axhline(y=0, color='k', linewidth=0.25)  # 绘制y=0的水平线
    ax.axvline(x=0, color='k', linewidth=0.25)  # 绘制x=0的垂直线
    ax.set_xlim(-2.5, 2.5)  # 设置x轴范围
    ax.set_ylim(-2.5, 2.5)  # 设置y轴范围
    ax.spines['top'].set_visible(False)  # 隐藏顶部边框
    ax.spines['right'].set_visible(False)  # 隐藏右侧边框
    ax.spines['bottom'].set_visible(False)  # 隐藏底部边框
    ax.spines['left'].set_visible(False)  # 隐藏左侧边框
    ax.set_xlabel('$x_1$')  # 设置x轴标签
    ax.set_ylabel('$x_2$')  # 设置y轴标签
    ax.set_title('p = ' + str(p))  # 设置子图标题，显示当前的p值
    ax.set_aspect('equal', adjustable='box')  # 设置坐标轴比例为相等

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
def plot_vectors(vectors,labels,ax_limits = (-2,8),grid = True):
    fig,ax = plt.subplots()

    x_coords = [v[0] for v in vectors]
    y_coords = [v[1] for v in vectors]
    dx_coords = [v[2] for v in vectors]
    dy_coords = [v[3] for v in vectors]

    plt.quiver(x_coords,y_coords,dx_coords,dy_coords,
              angles = 'xy',scale_units = 'xy',scale = 1,
              color = sns.color_palette())#传入一组颜色,是调色板不是colormap,n_colors = 可以预设color的数量

    plt.axvline(x = 0,color = 'grey')
    plt.axhline(y = 0,color = 'grey')

    for x_pos,y_pos,text,color_index in labels:
        plt.text(x_pos,y_pos,text,
                 color = sns.color_palette()[color_index],size = 12,
                 ha = 'center',va = 'center')

    # 设置图形细节
    plt.ylabel('$x_2$')  # y轴标签
    plt.xlabel('$x_1$')  # x轴标签
    plt.axis('scaled')  # 坐标轴比例相等
    ax.set_xticks(np.arange(ax_limits[0], ax_limits[1] + 1))  # 设置x轴刻度
    ax.set_yticks(np.arange(ax_limits[0], ax_limits[1] + 1))  # 设置y轴刻度
    ax.set_xlim(ax_limits[0], ax_limits[1])  # 设置x轴范围
    ax.set_ylim(ax_limits[0], ax_limits[1])  # 设置y轴范围

    if grid:
        ax.grid(linestyle = '--',linewidth = 0.25,color = '0.5')

    plt.show()
```

```python
u = [0,0,4,3]
v = [0,0,-2,4]
u_bis = [4,3,v[2],v[3]]
w = [0,0,2,7]
```

```python
vectors = [u,u_bis,w]
```
