```python
import numpy as np
import seaborn as sns
import scipy
from scipy.linalg import lu
from matplotlib import pyplot as plt
```

```python
A = np.array([[5,2,-2,3],
              [-2,5,-8,7],
              [7,-5,1,-6],
              [-5,4,-4,8]])
P,L,U = lu(A,permute_l = False) #permute_l表示是否把p合并进L
```

```python
fig,axs = plt.subplots(1,7,figsize = (12,3))

plt.sca(axs[0])
ax = sns.heatmap(A,cmap = 'RdBu_r',vmax = 10,vmin = -10,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('A')

plt.sca(axs[1])
plt.title(' = ')
plt.axis('off')

plt.sca(axs[2])
ax = sns.heatmap(P,cmap = 'RdBu_r',vmax = 10,vmin = -10,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('P')

plt.sca(axs[3])
plt.title(' @ ')
plt.axis('off')

plt.sca(axs[4])
ax = sns.heatmap(L,cmap = 'RdBu_r',vmax = 10,vmin = -10,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('L')

plt.sca(axs[5])
plt.title(' @ ')
plt.axis('off')

plt.sca(axs[6])
ax = sns.heatmap(U,cmap = 'RdBu_r',vmax = 10,vmin = -10,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('U')
```

```python
A = np.array([[5,2,-2,3],
              [-2,5,-8,7],
              [7,-5,1,-6],
              [-5,4,-4,8]])
L,D,a = scipy.linalg.ldl(A)
```

```python
fig,axs = plt.subplots(1,7,figsize = (12,3))

plt.sca(axs[0])
ax = sns.heatmap(A,cmap = 'RdBu_r',vmax = 10,vmin = -10,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('A')

plt.sca(axs[1])
plt.title(' = ')
plt.axis('off')

plt.sca(axs[2])
ax = sns.heatmap(L,cmap = 'RdBu_r',vmax = 10,vmin = -10,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('L')

plt.sca(axs[3])
plt.title(' @ ')
plt.axis('off')

plt.sca(axs[4])
ax = sns.heatmap(D,cmap = 'RdBu_r',vmax = 10,vmin = -10,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('D')

plt.sca(axs[5])
plt.title(' @ ')
plt.axis('off')

plt.sca(axs[6])
ax = sns.heatmap(L.T,cmap = 'RdBu_r',vmax = 10,vmin = -10,
                 cbar_kws = {'orientation':'horizontal'})
ax.set_aspect('equal')
plt.title('L_T')
```
