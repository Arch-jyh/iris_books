```python
A = np.matrix([[1.25,-0.75],
               [-0.75,1.25]])
```

```python
Lambda,V = np.linalg.eig(A)
Lambda
```

```python
B = V @ np.diag(np.sqrt(Lambda)) @ np.linalg.inv(V)
```

```python
initial_x_array = np.array([[1,0,0.5,0.4],
                            [0,1,0.5,0.6]]) #配置起始次数

num_iterations = 10
```

```python
for i in np.arange(0,4):
    initial_x = initial_x_array[:,i][:,None] #类似np.newaxis

    x_i = np.zeros_like(initial_x)
    x_i = initial_x #
    X = initial_x.T #初始化转置的状态,这里是第0次的数据

    for x in np.arange(0,num_iterations):
        x_i = T @ x_i
        X = np.concatenate([X,x_i.T],axis = 0) #这里的axis = 0表是按照最外层索引的方向拼接

    fig,ax = plt.subplots()

    itr = np.arange(0,num_iterations + 1)
    plt.plot(itr,X[:,0],marker = 'x',color = [1,0,0])
    plt.plot(itr,X[:,1],marker = 'x',color = [0,0.6,1])

    ax.grid(linestyle = '--',linewidth = 0.25,color = '0.5')
    ax.set_xlim(0,num_iterations)
    ax.set_ylim(0,1)
    ax.set_xlabel('Iteration,k')
    ax.set_ylabel('State')
```
