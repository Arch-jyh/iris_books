```python
import numpy as np
```

```python
A = np.matrix([[1.25,-0.75],
               [-0.75,1.25]])
```

```python
Lambda,V = np.linalg.eig(A)
Lambda
```

```python
V
```

```python
B = V @ np.diag(np.sqrt(Lambda)) @ np.linalg.inv(V)
```

```python
B
```

```python
A_reproduced = B @ B
A_reproduced
```
